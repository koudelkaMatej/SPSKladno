# 🌐 Práce s API a JSON

## Co je to API?

**API** (Application Programming Interface) je **rozhraní**, přes které spolu komunikují dva programy. Představte si API jako **číšníka v restauraci** – vy (klient) řeknete objednávku, číšník ji předá kuchyni (server) a přinese vám zpět jídlo (data).

### REST API

Nejčastější typ API na webu. Komunikace probíhá přes **HTTP požadavky** (stejný protokol jako webové stránky).

| HTTP metoda | K čemu slouží | Příklad |
|:---|:---|:---|
| **GET** | Získání dat | Zobrazení počasí |
| **POST** | Odeslání/vytvoření dat | Registrace uživatele |
| **PUT** | Aktualizace dat | Změna profilu |
| **DELETE** | Smazání dat | Odstranění příspěvku |

> 💡 V této lekci budeme pracovat hlavně s **GET** požadavky – tedy **získáváním dat**.

---

## 📋 Co je to JSON?

**JSON** (JavaScript Object Notation) je **textový formát pro výměnu dat**. Je to nejpoužívanější formát pro komunikaci s API.

JSON vypadá **velmi podobně** jako Python slovník:

```json
{
    "jmeno": "Anna",
    "vek": 20,
    "student": true,
    "predmety": ["Python", "SQL", "Flask"],
    "adresa": {
        "mesto": "Kladno",
        "psc": "272 01"
    }
}
```

### Mapování typů JSON → Python

| JSON | Python |
|:---|:---|
| `{}` objekt | `dict` |
| `[]` pole | `list` |
| `"text"` | `str` |
| `123` | `int` |
| `1.5` | `float` |
| `true` / `false` | `True` / `False` |
| `null` | `None` |

---

## 🔧 Modul `json`

Python má vestavěný modul `json` pro práci s JSON daty:

| Funkce | Směr | Popis |
|:---|:---|:---|
| `json.dumps(obj)` | Python → JSON řetězec | Serializace |
| `json.loads(text)` | JSON řetězec → Python | Deserializace |
| `json.dump(obj, soubor)` | Python → JSON soubor | Zápis do souboru |
| `json.load(soubor)` | JSON soubor → Python | Čtení ze souboru |

````python
import json

# Python slovník → JSON řetězec
student = {
    "jmeno": "Anna",
    "vek": 20,
    "predmety": ["Python", "SQL"]
}
json_text = json.dumps(student, ensure_ascii=False, indent=2)
print("JSON:")
print(json_text)

print("---")

# JSON řetězec → Python slovník
json_vstup = '{"mesto": "Kladno", "obyvatel": 69000}'
data = json.loads(json_vstup)
print("Python:", data)
print("Město:", data["mesto"])
````

---

## 🔗 JSON a soubory

````python
import json

# Zápis do JSON souboru
data = {
    "studenti": [
        {"jmeno": "Anna", "znamka": 1},
        {"jmeno": "Petr", "znamka": 2}
    ]
}

with open("studenti.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)
print("Soubor uložen!")

# Čtení z JSON souboru
with open("studenti.json", "r", encoding="utf-8") as f:
    nactena_data = json.load(f)
print("Načteno:", nactena_data)
````

---

### HTTP stavové kódy

| Kód | Význam |
|:---|:---|
| **200** | OK – vše v pořádku |
| **201** | Created – vytvořeno |
| **400** | Bad Request – chybný požadavek |
| **401** | Unauthorized – chybí autorizace |
| **404** | Not Found – nenalezeno |
| **500** | Internal Server Error – chyba serveru |

> 💡 Stavový kód **2xx** = úspěch, **4xx** = chyba klienta, **5xx** = chyba serveru

## 🎯 Praktický příklad – Veřejné API

Existuje mnoho **veřejných a bezplatných API**, se kterými si můžete hrát. Zde je několik oblíbených:

| API | URL | Popis |
|:---|:---|:---|
| JSONPlaceholder | `jsonplaceholder.typicode.com` | Falešná data (uživatelé, příspěvky) |
| PokeAPI | `pokeapi.co` | Data o Pokémonech |
| Open-Meteo | `open-meteo.com` | Předpověď počasí |
| Dog CEO | `dog.ceo` | Náhodné fotky psů |

````python
import requests

# Příklad: Získání dat o Pokémonovi
odpoved = requests.get("https://pokeapi.co/api/v2/pokemon/pikachu")

if odpoved.status_code == 200:
    pokemon = odpoved.json()   # Automaticky parsuje JSON → Python dict
    print(f"Jméno: {pokemon['name']}")
    print(f"Výška: {pokemon['height']}")
    print(f"Váha: {pokemon['weight']}")
    print(f"Typy: {[t['type']['name'] for t in pokemon['types']]}")
else:
    print("Chyba:", odpoved.status_code)
````

````python
import requests

# Příklad: Předpověď počasí pro Kladno
url = "https://api.open-meteo.com/v1/forecast"
parametry = {
    "latitude": 50.15,
    "longitude": 14.1,
    "current_weather": True
}

odpoved = requests.get(url, params=parametry)

if odpoved.status_code == 200:
    data = odpoved.json()
    pocasi = data["current_weather"]
    print(f"🌡️ Teplota: {pocasi['temperature']}°C")
    print(f"💨 Vítr: {pocasi['windspeed']} km/h")
else:
    print("Chyba:", odpoved.status_code)
````

## 🔑 API s parametry a klíči

Některá API vyžadují **parametry** v URL nebo **API klíč** pro autorizaci:

```python
# Parametry v URL
odpoved = requests.get("https://api.example.com/data", params={"city": "Prague", "lang": "cs"})
# Výsledná URL: https://api.example.com/data?city=Prague&lang=cs

# API klíč v hlavičce (headers)
headers = {"Authorization": "Bearer VAS_API_KLIC"}
odpoved = requests.get("https://api.example.com/data", headers=headers)
```

> ⚠️ **Nikdy nedávejte API klíče do veřejného kódu!** Používejte proměnné prostředí nebo konfigurační soubory mimo Git.
