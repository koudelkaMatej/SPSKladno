# ⌨️ PyQt6 – Lekce 3: Zpracování uživatelského vstupu

Aplikace, která jen zobrazuje pevně daný text, je málo užitečná.
V této lekci se naučíme, jak načítat hodnoty zadané uživatelem a jak je správně zpracovat.

---

## 📦 Přehled widgetů pro vstup

| Widget | Typ dat | Jak načíst hodnotu |
|---|---|---|
| `QLineEdit` | Jednořádkový text | `.text()` |
| `QTextEdit` | Víceřádkový text | `.toPlainText()` |
| `QSpinBox` | Celé číslo (se šipkami) | `.value()` |
| `QDoubleSpinBox` | Desetinné číslo | `.value()` |

---

## ⚠️ Ošetření vstupu – proč je důležité?

Uživatel může zadat cokoliv. Pokud chcete z textového pole číslo
a uživatel napíše `"ahoj"`, Python vyhodí `ValueError`.
Máme dvě řešení:

| Řešení | Princip |
|---|---|
| **Validátor** (`QIntValidator`) | Zabraňuje zadání špatných dat fyzicky při psaní |
| **`try / except`** | Ošetří chybu až ve chvíli zpracování |


---

## 🟢 Ukázka 1 – Základní textový vstup

Načteme text z pole a zobrazíme ho v nápisu.

````python
from PyQt6 import QtWidgets

app = QtWidgets.QApplication.instance() or QtWidgets.QApplication([])
okno = QtWidgets.QWidget()
okno.setWindowTitle("Textový vstup")
okno.resize(450, 80)

layout = QtWidgets.QHBoxLayout()
okno.setLayout(layout)

vstup = QtWidgets.QLineEdit()
vstup.setPlaceholderText("Napište vaše jméno…")  # Šedý hint text
tlacitko = QtWidgets.QPushButton("Pozdrav")
napis = QtWidgets.QLabel("Čekám na jméno…")

layout.addWidget(vstup)
layout.addWidget(tlacitko)
layout.addWidget(napis)

def pozdrav():
    jmeno = vstup.text()          # Načtení textu
    if jmeno.strip() == "":
        napis.setText("Zadej nejdřív jméno!")
    else:
        napis.setText(f"Ahoj, {jmeno}! 👋")
        vstup.clear()              # Vyčistit pole po odeslání

tlacitko.clicked.connect(pozdrav)

okno.show()
app.exec()
````

## 🟢 Ukázka 2 – Ošetření vstupu pomocí `try/except`

Sečteme čísla zadaná uživatelem oddělená mezerou.

````python
from PyQt6 import QtWidgets

app = QtWidgets.QApplication.instance() or QtWidgets.QApplication([])
okno = QtWidgets.QWidget()
okno.setWindowTitle("Součet čísel")
okno.resize(450, 80)

layout = QtWidgets.QHBoxLayout()
okno.setLayout(layout)

vstup = QtWidgets.QLineEdit()
vstup.setPlaceholderText("Zadejte čísla oddělená mezerou…")
tlacitko = QtWidgets.QPushButton("Sečti")
vysledek = QtWidgets.QLabel("—")

layout.addWidget(vstup)
layout.addWidget(tlacitko)
layout.addWidget(vysledek)

def soucet():
    try:
        cisla = [int(x) for x in vstup.text().split()]
        vysledek.setText(f"Součet = {sum(cisla)}")
    except ValueError:
        vysledek.setText("⛔ Chyba: zadal jsi nečíselnou hodnotu!")

tlacitko.clicked.connect(soucet)

okno.show()
app.exec()
````

## 🟡 Ukázka 3 – Validátor celých čísel

`QIntValidator` uživateli fyzicky zamezí zadat cokoliv jiného než celé číslo.

````python
from PyQt6 import QtWidgets, QtGui

app = QtWidgets.QApplication.instance() or QtWidgets.QApplication([])
okno = QtWidgets.QWidget()
okno.setWindowTitle("Validátor")
okno.resize(400, 80)

layout = QtWidgets.QHBoxLayout()
okno.setLayout(layout)

vstup = QtWidgets.QLineEdit()
vstup.setValidator(QtGui.QIntValidator(0, 1000))   # Povolena čísla 0–1000
vstup.setPlaceholderText("Zadejte číslo 0–1000")
tlacitko = QtWidgets.QPushButton("Zkontroluj")
napis = QtWidgets.QLabel("—")

layout.addWidget(vstup)
layout.addWidget(tlacitko)
layout.addWidget(napis)

def zkontroluj():
    hodnota = int(vstup.text())
    if hodnota > 500:
        napis.setText(f"{hodnota} je VELKÉ číslo.")
    else:
        napis.setText(f"{hodnota} je malé číslo.")

tlacitko.clicked.connect(zkontroluj)

okno.show()
app.exec()
````

## 🟡 Ukázka 4 – `QSpinBox` pro číselný vstup

`QSpinBox` je specializovaný widget pro celá čísla – má šipky nahoru/dolů.

````python
from PyQt6 import QtWidgets

app = QtWidgets.QApplication.instance() or QtWidgets.QApplication([])
okno = QtWidgets.QWidget()
okno.setWindowTitle("SpinBox ukázka")
okno.resize(350, 80)

layout = QtWidgets.QHBoxLayout()
okno.setLayout(layout)

napis_label = QtWidgets.QLabel("Věk:")
spin = QtWidgets.QSpinBox()
spin.setMinimum(0)
spin.setMaximum(120)
spin.setValue(18)

tlacitko = QtWidgets.QPushButton("Zkontroluj věk")
vysledek = QtWidgets.QLabel("—")

layout.addWidget(napis_label)
layout.addWidget(spin)
layout.addWidget(tlacitko)
layout.addWidget(vysledek)

def zkontroluj_vek():
    vek = spin.value()
    if vek >= 18:
        vysledek.setText("✅ Plnoletý")
    else:
        vysledek.setText(f"❌ Ještě {18 - vek} rok(ů) do plnoletosti")

tlacitko.clicked.connect(zkontroluj_vek)

okno.show()
app.exec()
````
