## ⚙️ Architektura: Signály a Sloty

PyQt používá návrhový vzor zvaný **Signál–Slot**:

```
Uživatel klikne na tlačítko
         │
         ▼
  Widget vyšle SIGNÁL  ──► clicked
         │
         ▼
  Signál je propojen se SLOTEM (naší funkcí)
         │
         ▼
  Funkce se zavolá a provede se
```

Propojení zajistíme takto:
```python
tlacitko.clicked.connect(moje_funkce)
#                                   ^
#             POZOR: bez závorek! Předáváme referenci na funkci,
#             ne výsledek jejího volání.
```

---

## 📦 Přehled metod pro práci s textem

| Metoda | Na čem | Co dělá |
|---|---|---|
| `.setText("text")` | `QLabel`, `QPushButton` | Změní text prvku |
| `.text()` | `QLabel`, `QLineEdit` | Vrátí aktuální text prvku |
| `.setWindowTitle("text")` | `QWidget` | Změní titulek okna |

## 🟢 Ukázka 1 – Tlačítko změní nápis

Po kliknutí se text `QLabel` změní na jiný.

````python
from PyQt5 import QtWidgets

app = QtWidgets.QApplication([])
okno = QtWidgets.QWidget()
okno.setWindowTitle("Funkce u tlačítka")
okno.resize(350, 80)

layout = QtWidgets.QHBoxLayout()
okno.setLayout(layout)

napis = QtWidgets.QLabel("Původní text")
tlacitko = QtWidgets.QPushButton("Změň text")

layout.addWidget(napis)
layout.addWidget(tlacitko)

def zmen_text():
    napis.setText("Text byl změněn! ✅")

tlacitko.clicked.connect(zmen_text)  # Propojení signálu clicked s funkcí

okno.show()
app.exec()
````

## 🟢 Ukázka 2 – Počítadlo kliknutí

Sledujeme, kolikrát uživatel klikl na tlačítko.

````python
from PyQt5 import QtWidgets

app = QtWidgets.QApplication([])
okno = QtWidgets.QWidget()
okno.setWindowTitle("Počítadlo")
okno.resize(300, 80)

layout = QtWidgets.QHBoxLayout()
okno.setLayout(layout)

pocet = 0  # Proměnná mimo funkci – sdílíme ji se slotem přes nonlocal

napis = QtWidgets.QLabel("Kliknuto: 0×")
tlacitko = QtWidgets.QPushButton("Klikni!")

layout.addWidget(napis)
layout.addWidget(tlacitko)

def kliknuto():
    global pocet       # Modifikujeme proměnnou z vnějšího scope
    pocet += 1
    napis.setText(f"Kliknuto: {pocet}×")

tlacitko.clicked.connect(kliknuto)

okno.show()
app.exec()
````

## 🟡 Ukázka 3 – Přepínání stavu (toggle)

Tlačítko přepíná mezi dvěma stavy – zapnuto/vypnuto.

````python
from PyQt5 import QtWidgets

app = QtWidgets.QApplication([])
okno = QtWidgets.QWidget()
okno.setWindowTitle("Přepínač")
okno.resize(300, 80)

layout = QtWidgets.QHBoxLayout()
okno.setLayout(layout)

zapnuto = False
napis = QtWidgets.QLabel("Stav: VYPNUTO 🔴")
tlacitko = QtWidgets.QPushButton("Přepnout")

layout.addWidget(napis)
layout.addWidget(tlacitko)

def prepni():
    global zapnuto
    zapnuto = not zapnuto
    if zapnuto:
        napis.setText("Stav: ZAPNUTO 🟢")
    else:
        napis.setText("Stav: VYPNUTO 🔴")

tlacitko.clicked.connect(prepni)

okno.show()
app.exec()
````

## 🟡 Ukázka 4 – Náhodný výběr ze seznamu

Po každém kliknutí se zobrazí náhodně vybraná položka.

````python
from PyQt5 import QtWidgets
import random

app = QtWidgets.QApplication([])
okno = QtWidgets.QWidget()
okno.setWindowTitle("Náhodný losovač")
okno.resize(400, 80)

layout = QtWidgets.QHBoxLayout()
okno.setLayout(layout)

jidla = ["Pizza 🍕", "Burgr 🍔", "Sushi 🍣", "Polévka 🍜", "Salát 🥗"]
napis = QtWidgets.QLabel("Co dnes k obědu?")
tlacitko = QtWidgets.QPushButton("Losuj!")

layout.addWidget(napis)
layout.addWidget(tlacitko)

def losuj():
    vybrane = random.choice(jidla)
    napis.setText(f"Dnes budeš jíst: {vybrane}")

tlacitko.clicked.connect(losuj)

okno.show()
app.exec()
````
