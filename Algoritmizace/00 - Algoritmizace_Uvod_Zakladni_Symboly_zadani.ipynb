# 🔄 Algoritmizace – Úvod a základní symboly

---

## Co je algoritmizace?

**Algoritmizace** je proces, při kterém graficky nebo textově znázorňujeme **algoritmy** – postupy řešení určitého problému.

Před tím, než začneme programovat, je důležité si:
- 📋 **Jasně vymezit** problém, který chceme vyřešit
- 🎯 **Navrhnout** postup řešení (algoritmus)
- 📐 **Vizuálně znázornit** tok programu

Algoritmizace nám pomáhá **myslet přesně** a **vyhnout se chybám** ještě před tím, než začneme psát kód.

---

## Proč je algoritmizace důležitá?

| Důvod | Popis |
|---|---|
| **Přehlednost** | Vidíme strukturu problému na první pohled |
| **Kontrola logiky** | Odhalíme chyby v logice ještě před programováním |
| **Komunikace** | Můžeme se o řešení snadno bavit s ostatními |
| **Opakovaná použitelnost** | Stejný algoritmus můžeme implementovat v různých jazycích |
| **Testování** | Snáze ověříme, zda algoritmus funguje správně |

---

## Algoritmus

**Algoritmus** = konečná posloupnost přesně definovaných kroků, která vede ke správnému výsledku.

### Příklad z běžného života
Algoritmus pro vaření čaje:
1. Vezmi konvici a naplň ji vodou
2. Zapni konvici a počkej, až se voda uvaří
3. Vlož čajový sáček do hrnku
4. Zalij horkou vodou
5. Počkej 3–5 minut
6. Vyjmi sáček, případně přidej cukr

---

### Vlastnosti algoritmu

Každý správný algoritmus musí mít tyto **základní vlastnosti**:

| Vlastnost | Popis |
|---|---|
| **Obecnost** | Funguje pro celou třídu problémů, ne jen pro jeden konkrétní případ. Např. algoritmus pro řešení kvadratické rovnice funguje pro *libovolnou* rovnici $ax^2 + bx + c = 0$, ne jen pro $3x^2 - 2x + 1 = 0$. |
| **Správnost (korektnost)** | Vždy dojde ke správnému výsledku pro libovolný platný vstup. |
| **Konečnost** | Algoritmus skončí po konečném počtu kroků – nikdy neběží do nekonečna. |
| **Elementárnost** | Skládá se z jednoduchých, elementárních kroků, kterým vykonavatel (počítač) rozumí. |
| **Determinovanost** | V každém okamžiku je jednoznačně dáno, co se má provést. Při stejném vstupu dostaneme vždy stejný výstup. |

---

## Metody znázornění algoritmů

### 1. **Vývojový diagram (flowchart)**
Grafické znázornění toku programu pomocí speciálních symbolů a šipek.

### 2. **Pseudokód**
Textový popis algoritmu napůl v přirozené řeči, napůl v programovacím stylu.

### 3. **Strukturovaný seznam kroků**
Jednoduše očíslovaný seznam instrukcí.

V této lekci se zaměříme na **vývojové diagramy** a **pseudokód**.

# 📊 Základní symboly vývojového diagramu

Existují standardizované symboly pro kreslení vývojových diagramů (norma ISO 5807). Zde jsou ty nejzákladnější:

## 1. **Terminál (Start/Konec)**

```
    ┌─────────┐
    │  START  │
    └────┬────┘
         │
        ...
         │
    ┌────┴────┐
    │  KONEC  │
    └─────────┘
```

**Použití:** Začátek a konec programu/algoritmu

**Tvar:** Zaoblený obdélník (elipsa nebo zaoblené rozhledy)

---

## 2. **Proces (Akce)**

```
    ┌──────────────┐
    │  Přiřazení   │
    │  proměnné    │
    └──────┬───────┘
```

**Použití:** Libovolná akce – přiřazení proměnné, výpočet, operace

**Tvar:** Obdélník

**Příklady:**
- `a = 5`
- `vysledek = a + b`
- `Přečti vstup od uživatele`

---

## 3. **Rozhodnutí (Podmínka)**

```
         /\
        /  \
       /    \
      | x>0  |
       \    /
        \  /
         \/
        ANO  NE
         │    │
```

**Použití:** Podmínka (if-else) – rozhoduje, kterou cestu se půjdeme

**Tvar:** Kosočtverec (diamant)

**Vlastnosti:**
- Má **dvě výstupní šipky**: ANO/NE (nebo PRAVDA/NEPRAVDA)
- Uvnitř se píše **podmínka** (výraz vrací TRUE nebo FALSE)

**Příklady podmínek:**
- `x > 0`
- `věk >= 18`
- `počet > 100`

---

## 4. **Vstup/Výstup (I/O)**

```
    ┌──────────────┐
    │     │  │     │
    │  Přečti: x   │
    │     └──┘     │
    └──────────────┘
```

**Použití:** Vstup od uživatele nebo výstup na obrazovku

**Tvar:** Rovnoběžník

**Příklady:**
- `Přečti: věk`
- `Vypiš: výsledek`

---

## 5. **Šipka (Tok)**

```
    ─────────►
```

**Použití:** Ukazuje směr toku programu – kterým směrem se pokračuje dál

---

# 📋 Shrnutí symbolů

| Symbol | Název | Použití |
|---|---|---|
| Elipsa/zaoblený obdélník | Terminál | Start/Konec |
| Obdélník | Proces | Akce, přiřazení |
| Kosočtverec | Rozhodnutí | Podmínka (if-else) |
| Rovnoběžník | Vstup/Výstup | Čtení/psaní dat |
| Šipka | Tok | Směr běhu programu |

## Stejný algoritmus v pseudokódu:

```

─────────────────────────────────

  1. Přečti číslo x
  2. POKUD x >= 0 POTOM
       Vypiš "Kladné nebo nula"
     JINAK
       Vypiš "Záporné"
     KONEC POKUD

KONEC
```

# 📝 Pseudokód – podrobněji

Pseudokód je **textový popis algoritmu**, který používá jednoduchý jazyk mezi přirozenou řečí a programovacím jazykem.

## Výhody pseudokódu:
- ✅ **Nezávisí** na konkrétním programovacím jazyce
- ✅ **Snadno čitelný** i pro lidi bez programovacích znalostí
- ✅ **Rychlejší** psaní než vlastní kód
- ✅ **Dobrý** pro plánování a komunikaci

## Základní klíčová slova v pseudokódu:

| Klíčové slovo | Popis | Příklad |
|---|---|---|
| START / KONEC | Začátek a konec algoritmu | START / KONEC |
| PŘEČTI | Vstup od uživatele | PŘEČTI: věk |
| VYPIŠ | Výstup na obrazovku | VYPIŠ: "Věk je", věk |
| ← (nebo =) | Přiřazení hodnoty | x ← 5 |
| POKUD...POTOM...JINAK | Podmínka | POKUD x > 0 POTOM ... JINAK ... |
| DOKUD | Cyklus s podmínkou | DOKUD počet < 10 |
| PRO | Cyklus s pevným počtem opakování | PRO i = 1 TO 5 |

## Příklad pseudokódu – Součet dvou čísel:

```
ALGORITMUS: Sčítání dvou čísel
─────────────────────────────
START

  1. PŘEČTI: prvníČíslo
  2. PŘEČTI: druhéČíslo
  3. součet ← prvníČíslo + druhéČíslo
  4. VYPIŠ: "Součet je:", součet

KONEC
```

# 🔗 Struktury algoritmů

Existují tři základní struktury, ze kterých se skládají všechny algoritmy:

## 1. **Sekvence (Posloupnost)**
Příkazy se vykonávají postupně za sebou, jeden po druhém.

```
Příkaz 1
   │
   ▼
Příkaz 2
   │
   ▼
Příkaz 3
```

## 2. **Větvení (Selektor)**
Na základě podmínky se program rozdělí do dvou cest.

```
      /\
     /  \
    /    \
   │ Podmínka?
    \    /
     \  /
      \/
    ANO  NE
     │    │
  Cesta A  Cesta B
```

## 3. **Cyklus (Iterace)**
Určitá část kódu se opakuje, dokud je splněna podmínka.

```
Inicializace
   │
   ▼
  /\
 /  \
│Podmínka?
 \  /
  \/
ANO │ NE
    ▼ └──► Konec
 Tělo cyklu
    │
    └──────► (zpět na Podmínka)
```

> 💡 **Důležité:** Každý algoritmus se dá vytvořit kombinací těchto tří základních struktur!
