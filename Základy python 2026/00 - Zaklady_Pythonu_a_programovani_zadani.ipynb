## Co je to programování?

**Programování** je proces, při kterém vytváříme sadu instrukcí (příkazů), kterým počítač rozumí a dokáže je vykonat. Výsledkem je **program** (aplikace, software), který řeší určitý problém nebo automatizuje nějakou činnost.

Programování je v dnešní době jednou z nejžádanějších dovedností. Programy jsou všude kolem nás – v mobilních telefonech, autech, bankomatech, webových stránkách, hrách i v lékařských přístrojích.

> 💡 **Zajímavost:** Jeden z prvních programů napsala **Ada Lovelace** v roce 1843 pro analytický stroj Charlese Babbage. Je považována za první programátorku na světě.

---

## Algoritmus

Než začneme programovat, musíme nejprve **pochopit problém**, který chceme řešit, a navrhnout **algoritmus** – tedy přesný postup kroků, které vedou k řešení.

**Algoritmus** = konečná posloupnost přesně definovaných kroků, která vede ke správnému výsledku.

### Příklad z běžného života
Algoritmus pro vaření čaje:
1. Vezmi konvici a naplň ji vodou
2. Zapni konvici a počkej, až se voda uvaří
3. Vlož čajový sáček do hrnku
4. Zalij horkou vodou
5. Počkej 3–5 minut
6. Vyjmi sáček, případně přidej cukr

### Vlastnosti algoritmu

Každý správný algoritmus musí mít tyto vlastnosti:

| Vlastnost | Popis |
|---|---|
| **Obecnost** | Funguje pro celou třídu problémů, ne jen pro jeden konkrétní případ. Např. algoritmus pro řešení kvadratické rovnice funguje pro *libovolnou* rovnici $ax^2 + bx + c = 0$, ne jen pro $3x^2 - 2x + 1 = 0$. |
| **Správnost (korektnost)** | Vždy dojde ke správnému výsledku pro libovolný platný vstup. |
| **Konečnost** | Algoritmus skončí po konečném počtu kroků – nikdy neběží do nekonečna. |
| **Elementárnost** | Skládá se z jednoduchých, elementárních kroků, kterým vykonavatel (počítač) rozumí. |
| **Determinovanost** | V každém okamžiku je jednoznačně dáno, co se má provést. Při stejném vstupu dostaneme vždy stejný výstup. |

## Sémantika vs. Syntax

Při programování se setkáváme se dvěma klíčovými pojmy:

### Syntax (syntaxe)
**Syntax** = pravidla, podle kterých musíme program psát, aby mu počítač rozuměl.

Podobně jako čeština má pravidla pravopisu a gramatiky, každý programovací jazyk má svá syntaktická pravidla. Pokud je porušíme, program nefunguje a dostaneme **syntaktickou chybu** (SyntaxError).

```python
# Správná syntax:
print("Ahoj")

# Špatná syntax (chybí závorka):
print("Ahoj"
```

### Sémantika
**Sémantika** = význam programu, tedy to, **co program skutečně dělá**.

Program může být syntakticky správný (bez chyb), ale sémanticky špatný (dělá něco jiného, než chceme). Odhalení sémantických chyb je mnohem těžší a tvoří hlavní náplň práce programátora.

```python
# Chtěli jsme sečíst dvě čísla, ale odečetli jsme je – sémantická chyba:
vysledek = a - b   # mělo být: a + b
```

> ⚠️ **Důležité:** Syntaktické chyby vám počítač oznámí sám. Sémantické chyby musíte najít vy – proto je tak důležité programu rozumět a testovat ho!

# 🐍 Programovací jazyk Python

## Proč Python?

Python je jeden z nejpopulárnějších programovacích jazyků na světě. Je oblíbený díky své **jednoduchosti**, **čitelnosti** a **univerzálnosti**. Používá se pro:

- 🌐 **Webové aplikace** (Django, Flask)
- 📊 **Datovou analýzu a vědu** (pandas, NumPy, matplotlib)
- 🤖 **Umělou inteligenci a strojové učení** (TensorFlow, PyTorch)
- 🎮 **Hry** (Pygame)
- 🔧 **Automatizaci a skriptování**
- 📱 **A mnoho dalšího...**

> 💡 Python používají firmy jako Google, Instagram, Spotify, Netflix nebo NASA.

## Interpretovaný jazyk

Python je **interpretovaný** programovací jazyk. Co to znamená?

| Typ | Popis | Příklady |
|---|---|---|
| **Interpretovaný** | Program se překládá a spouští **řádek po řádku** za běhu. Existuje program zvaný **interpret**, který čte náš kód a rovnou ho vykonává. | Python, JavaScript, Ruby |
| **Kompilovaný** | Celý program se nejprve přeloží (**zkompiluje**) do strojového kódu a teprve potom se spustí. | C, C++, Rust |

### Jak Python funguje?
1. Napíšeme zdrojový kód do textového souboru s příponou `.py`
2. **Interpret Pythonu** čte náš kód řádek po řádku
3. Každý řádek přeloží a **okamžitě vykoná**
4. Pokud narazí na chybu, zastaví se a vypíše chybovou hlášku

### Základní pravidla Pythonu
- Příkazy se vykonávají **odshora dolů** (sekvenčně)
- Píšeme **jeden příkaz na jeden řádek**
- Python je **case-sensitive** (rozlišuje velká a malá písmena) – `X` a `x` jsou různé hodnoty
- K oddělování bloků kódu se používá **odsazení** (indentace), nikoli závorky jako v jiných jazycích

# 📤 Funkce `print()` – výstup na obrazovku

Funkce `print()` je jednou z nejzákladnějších a nejpoužívanějších funkcí v Pythonu. Slouží k **zobrazení hodnot** na obrazovku (do konzole).

### Základní použití
```python
print("text, který chceme zobrazit")
```

### Co umí `print()`?
- Vypsat **text** (řetězec) – musí být v uvozovkách `"..."` nebo apostrofech `'...'`
- Vypsat **čísla** – bez uvozovek
- Vypsat **více hodnot najednou** – oddělených čárkou
- Automaticky přidá **nový řádek** na konci výpisu

### Důležité detaily
- Text (tzv. **řetězec/string**) vždy zapisujeme do uvozovek: `"Ahoj"` nebo `'Ahoj'`
- Čísla zapisujeme **bez uvozovek**: `42`, `3.14`
- Při vypsání více hodnot oddělených čárkou se mezi ně automaticky vloží **mezera**

> 🎯 Funkce `print()` je váš nejlepší přítel při ladění programu – kdykoli si nejste jisti, co se v programu děje, vypište si hodnoty pomocí `print()`!

````python
# Výpis textu (řetězce)
print("Ahoj světe!")

# Výpis více hodnot oddělených čárkou – Python mezi ně automaticky vloží mezeru
print("Jmenuji se", "Python")

# Výpis čísel
print(42)
print(3.14)

# Kombinace textu a čísel
print("Mám", 3, "jablka a", 2, "hrušky")

# Prázdný print() vytiskne pouze nový řádek
print()
print("slova", "oddělené", "jinak", "než",  "mezerou", sep="-")
print()
print("--- Konec ukázky ---")
````

# 📦 Proměnné

## Co je to proměnná?

**Proměnnou** si můžete představit jako **pojmenovanou krabičku** v paměti počítače, do které můžeme uložit libovolnou hodnotu (číslo, text, …) a později ji znovu použít.

Každá proměnná má tři základní vlastnosti:

| Vlastnost | Popis |
|---|---|
| **Název (identifikátor)** | Jméno, které proměnné dáme my – volíme ho tak, aby jasně popisovalo obsah |
| **Hodnota** | Konkrétní data uložená v proměnné (číslo, text, …) |
| **Datový typ** | Druh dat – Python ho určí **automaticky** podle přiřazené hodnoty |

## Vytvoření proměnné

Proměnnou vytvoříme pomocí operátoru přiřazení `=` (rovná se):

```python
nazev_promenne = hodnota
```

> ⚠️ Pozor: Znak `=` v programování neznamená „rovná se" jako v matematice! Znamená **„ulož hodnotu do proměnné"** (přiřazení). Čteme ho zprava doleva: *„hodnotu 10 ulož do proměnné číslo"*.

## Pravidla pro pojmenování proměnných

- Název **musí začínat písmenem** nebo podtržítkem `_`
- Může obsahovat **písmena, číslice a podtržítka**
- **Nesmí obsahovat mezery** ani speciální znaky (!, @, #, …)
- **Nesmí být klíčové slovo** Pythonu (`if`, `for`, `while`, `print`, `int` , …)
- Python **rozlišuje velká a malá písmena** – `vek`, `Vek` a `VEK` jsou tři různé proměnné!

### Konvence pojmenování
V Pythonu se proměnné pojmenovávají stylem **snake_case** – slova se oddělují podtržítky a píšou se malými písmeny:
```python
# ✅ Správně (snake_case):
jmeno_studenta = "Jan"
pocet_bodu = 42

# ❌ Špatně:
JmenoStudenta = "Jan"   # toto je camelCase – v Pythonu se nepoužívá pro proměnné
2cislo = 10              # začíná číslicí – syntaktická chyba!
```

## Základní datové typy

| Typ | Python název | Příklad | Popis |
|---|---|---|---|
| Celé číslo | `int` | `42`, `-7`, `0` | Celá čísla bez desetinné čárky |
| Desetinné číslo | `float` | `3.14`, `-0.5` | Čísla s desetinnou tečkou (ne čárkou!) |
| Text (řetězec) | `str` | `"Ahoj"`, `'Python'` | Text v uvozovkách nebo apostrofech |
| Logická hodnota | `bool` | `True`, `False` | Pravda / Nepravda (jen tyto dvě hodnoty) |

> 💡 Typ proměnné můžete kdykoliv zjistit pomocí funkce `type()`:
> ```python
> print(type(42))       # <class 'int'>
> print(type("Ahoj"))   # <class 'str'>
> ```

````python
# Vytvoření proměnných různých datových typů
vek = 17                    # int (celé číslo)
vyska = 1.75                # float (desetinné číslo)
jmeno = "Karel"             # str (řetězec/text)
je_student = True           # bool (logická hodnota)

# Vypsání hodnot proměnných
print("Jméno:", jmeno)
print("Věk:", vek)
print("Výška:", vyska)
print("Je student?", je_student)

# Zjištění datového typu
print()
print("Typ proměnné 'vek':", type(vek))
print("Typ proměnné 'jmeno':", type(jmeno))

# Změna hodnoty proměnné – do stejné proměnné můžeme uložit novou hodnotu
vek = 18
print()
print("Nový věk:", vek)
````

# 📥 Funkce `input()` – vstup od uživatele

Funkce `input()` umožňuje **přijmout data od uživatele** během běhu programu. Program se zastaví, počká na zadání textu uživatelem a po stisknutí `Enter` pokračuje dál.

### Základní použití
```python
# Bez výzvy – program čeká na vstup bez jakéhokoli textu
odpoved = input()

# S výzvou (promptem) – uživateli se zobrazí text
jmeno = input("Jak se jmenuješ? ")
```

### Důležité vlastnosti
- Funkce `input()` **vždy vrací text** (datový typ `str`) – i když uživatel zadá číslo!
- Pokud chceme se zadaným číslem počítat, musíme ho **převést** (o tom více za chvíli)
- Výsledek `input()` typicky ukládáme do **proměnné**, abychom s ním mohli dále pracovat

> ⚠️ **Častá chyba začátečníků:** Uživatel zadá číslo `5`, ale program ho vidí jako text `"5"`. Pokud se s ním pokusíme počítat bez převodu, dostaneme neočekávaný výsledek!

### 🧪 Vyzkoušejte si: Co vypíše následující blok kódu?
Spusťte buňku níže, zadejte požadované hodnoty a sledujte výstup:

````python
# Ukázka funkce input()
jmeno = input("Jak se jmenuješ? ")
jidlo = input("Jaké je tvé oblíbené jídlo? ")

print()
print("Ahoj,", jmeno + "!")
print("Tvé oblíbené jídlo je:", jidlo)
````

## 🔄 Převod datových typů (přetypování)

Jak jsme si řekli, `input()` vždy vrací **text** (`str`). Pokud chceme se vstupem pracovat jako s číslem, musíme ho převést:

| Funkce | Co dělá | Příklad |
|---|---|---|
| `int()` | Převede na celé číslo | `int("42")` → `42` |
| `float()` | Převede na desetinné číslo | `float("3.14")` → `3.14` |
| `str()` | Převede na text | `str(42)` → `"42"` |

### Příklad: Sčítání dvou čísel od uživatele
```python
# ❌ BEZ převodu – "spojí" texty místo sčítání:
a = input("Zadej číslo: ")   # uživatel zadá 5 → a = "5"
b = input("Zadej číslo: ")   # uživatel zadá 3 → b = "3"
print(a + b)                  # Výsledek: "53" (spojení textů!)

# ✅ S převodem – správně sečte čísla:
a = int(input("Zadej číslo: "))   # uživatel zadá 5 → a = 5
b = int(input("Zadej číslo: "))   # uživatel zadá 3 → b = 3
print(a + b)                       # Výsledek: 8
```

````python
# Ukázka problému bez převodu
cislo_text = "5"
print("Bez převodu (text + text):", cislo_text + cislo_text)  # "55" – spojení textů

# Správný převod na číslo
cislo = int(cislo_text)
print("S převodem (číslo + číslo):", cislo + cislo)           # 10 – matematické sčítání

# Praktický příklad: vstup od uživatele
rok_narozeni = int(input("Zadej svůj rok narození: "))
vek = 2026 - rok_narozeni
print("Je ti přibližně", vek, "let.")
````

# 💬 Komentáře

## K čemu slouží komentáře?

Komentáře jsou **text v kódu, který Python zcela ignoruje** – nemá žádný vliv na běh programu. Slouží výhradně pro **lidi**, kteří kód čtou (včetně vás samotných v budoucnu!).

### Proč psát komentáře?
- 📝 **Vysvětlení složitějších částí kódu** – co daný úsek dělá a proč
- 🔍 **Rychlejší orientace** – vy i ostatní se v kódu lépe vyznáte
- 🐛 **Ladění (debugging)** – můžete dočasně „zakomentovat" řádek kódu, aniž byste ho mazali
- 📖 **Dokumentace** – popis toho, co program/funkce dělá

## Typy komentářů v Pythonu

### 1. Jednořádkový komentář – symbol `#`
Vše za symbolem `#` až do konce řádku je komentář:

```python
# Toto je jednořádkový komentář
print("Ahoj")  # Komentář může být i za kódem na stejném řádku
```

### 2. Víceřádkový komentář – trojité uvozovky `"""..."""`
Pro delší komentáře přes více řádků použijeme trojité uvozovky:

```python
"""
Toto je víceřádkový komentář.
Může zabírat libovolný počet řádků.
Často se používá pro dokumentaci funkcí.
"""
```

### Dobrá praxe při psaní komentářů
- ✅ Pište komentáře, které vysvětlují **proč**, ne jen **co** (kód sám říká, co dělá)
- ✅ Komentáře pište **v češtině nebo angličtině** – konzistentně
- ❌ Nepřehánějte to – nekomentujte úplně zřejmé věci (`x = 5  # uložíme 5 do x` je zbytečné)
- ❌ Nepoužívejte komentáře jako náhradu za srozumitelný kód

> 💡 Klávesová zkratka pro zakomentování/odkomentování řádku ve VS Code: `Ctrl + /` (anglická klávesnice)
