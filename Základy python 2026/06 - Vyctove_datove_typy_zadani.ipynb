# 📦 Výčtové (kolekční) datové typy

## Co jsou kolekce?

Zatím jsme pracovali s proměnnými, které uchovávaly **jednu hodnotu** (`int`, `str`, `bool`, `float`). V praxi ale často potřebujeme uložit **více hodnot** pohromadě – seznam studentů, souřadnice bodu, databázi kontaktů...

K tomu slouží **kolekční datové typy**:

| Typ | Syntaxe | Měnitelný? | Indexovaný? | Duplicity? | Použití |
|:---:|:---:|:---:|:---:|:---:|---|
| `list` | `[1, 2, 3]` | ✅ Ano | ✅ Ano | ✅ Ano | Univerzální seznam |
| `tuple` | `(1, 2, 3)` | ❌ Ne | ✅ Ano | ✅ Ano | Neměnná data (souřadnice, konstanty) |
| `set` | `{1, 2, 3}` | ✅ Ano | ❌ Ne | ❌ Ne | Unikátní hodnoty |
| `dict` | `{"a": 1}` | ✅ Ano | ✅ Klíčem | ❌ Klíče | Páry klíč–hodnota |
| `Enum` | `class ...` | ❌ Ne | ✅ Jménem | ❌ Ne | Pojmenované konstanty |

> 💡 **Měnitelný (mutable)** = po vytvoření můžeme přidávat, odebírat a měnit prvky.  
> **Neměnitelný (immutable)** = po vytvoření už nelze obsah změnit.

## 0️⃣ Rychlé připomenutí: `range()`

`range()` jsme už poznali u `for` cyklu. Technicky také generuje posloupnost, ale **nevytváří seznam v paměti** – generuje čísla postupně (je to tzv. *lazy* objekt):

````python
# === range() – připomenutí ===
# range() sám o sobě nevypíše čísla, ale vytvoří objekt
print(range(5))          # range(0, 5)

# Hvězdička * "rozbalí" range do jednotlivých hodnot
print(*range(5))         # 0 1 2 3 4

# Převod na seznam pro lepší vizualizaci
print(list(range(5)))    # [0, 1, 2, 3, 4]

# Připomenutí: range(start, stop, krok)
print(list(range(2, 10, 3)))   # [2, 5, 8]
print(list(range(10, 0, -2)))  # [10, 8, 6, 4, 2]
````

## 1️⃣ List (seznam) – `[...]`

**List** je nejpoužívanější kolekce v Pythonu. Je **měnitelný** (mutable) a **uspořádaný** (zachovává pořadí prvků).

### Vytvoření listu

```python
prazdny = []                         # prázdný list
cisla = [1, 2, 3, 4, 5]             # list čísel
jmena = ["Adam", "Bára", "Cyril"]   # list stringů
mix = [1, "text", True, 3.14]       # list různých typů
```

````python
# === Vytvoření listu ===
muj_list = ["jablko", "hruška", "meloun"]
print(muj_list)           # ['jablko', 'hruška', 'meloun']
print(type(muj_list))     # <class 'list'>
print(len(muj_list))      # 3 – počet prvků
````

> ⚠️ **Pozor:** Pozor na jmenou konvenci! Vytvoření listu s názvem "list", vám rozbije práci s Listy!

### Indexování – přístup k prvkům

Prvky v listu jsou **indexované od 0** (stejně jako u stringů). K prvku přistupujeme pomocí `[index]`:

```
muj_list = ["jablko", "hruška", "meloun"]
  index:       0         1         2
záporný:      -3        -2        -1
```

> ⚠️ **Pozor:** Přístup k neexistujícímu indexu vyhodí `IndexError`!

````python
# === Indexování ===
#  index:       0         1         2
muj_list = ["jablko", "hruška", "meloun"]

print(muj_list[0])    # "jablko" (první prvek)
print(muj_list[1])    # "hruška" (druhý prvek)
print(muj_list[-1])   # "meloun" (poslední prvek)

# Slicing funguje stejně jako u stringů!
print(muj_list[0:2])  # ["jablko", "hruška"] (od 0 do 1)
````

### Úprava listu – přidání, odebrání, změna

Protože je list **měnitelný**, můžeme jeho obsah libovolně upravovat:

| Operace | Metoda | Příklad |
|---|---|---|
| Přidání na konec | `.append(x)` | `lst.append("nový")` |
| Vložení na pozici | `.insert(i, x)` | `lst.insert(1, "nový")` |
| Odebrání podle indexu | `.pop(i)` | `lst.pop(0)` |
| Odebrání podle hodnoty | `.remove(x)` | `lst.remove("hruška")` |
| Změna prvku | `lst[i] = x` | `lst[0] = "malina"` |
| Smazání všeho | `.clear()` | `lst.clear()` |

````python
# === Úprava listu ===
muj_list = ["jablko", "hruška", "meloun"]
print(f"Původní: {muj_list}")

# Přidání na konec
muj_list.append("banán")
print(f"Po append('banán'): {muj_list}")

# Vložení na konkrétní pozici
muj_list.insert(1, "třešeň")
print(f"Po insert(1, 'třešeň'): {muj_list}")

# Odebrání podle indexu (vrací odebraný prvek)
odebrany = muj_list.pop(2)
print(f"Po pop(2): {muj_list} (odebrán: '{odebrany}')")

# Odebrání podle hodnoty
muj_list.remove("meloun")
print(f"Po remove('meloun'): {muj_list}")

# Změna prvku na indexu
muj_list[0] = "malina"
print(f"Po lst[0] = 'malina': {muj_list}")
````

### 📏 Další užitečné metody a funkce

| Metoda / Funkce | Popis | Příklad |
|:---|:---|:---|
| `len(lst)` | Počet prvků | `len([1,2,3])` → `3` |
| `sorted(lst)` | Vrátí **nový** seřazený list | `sorted([3,1,2])` → `[1,2,3]` |
| `lst.sort()` | Seřadí list **na místě** (změní ho) | `lst.sort()` |
| `lst.reverse()` | Obrátí pořadí **na místě** | `lst.reverse()` |
| `lst.index(x)` | Index prvního výskytu `x` | `["a","b"].index("b")` → `1` |
| `lst.count(x)` | Počet výskytů `x` | `[1,1,2].count(1)` → `2` |
| `x in lst` | Je `x` v listu? (`True`/`False`) | `3 in [1,2,3]` → `True` |
| `lst.copy()` | Kopie listu (mělká) | `novy = lst.copy()` |

> ⚠️ **Pozor na rozdíl:** `sorted()` vytvoří **nový** list, zatímco `.sort()` změní **původní** list a vrátí `None`!

````python
# === Užitečné metody a funkce ===
cisla = [5, 2, 8, 1, 9, 2, 5]
print(f"Původní: {cisla}")
print(f"Délka:   len() = {len(cisla)}")
print(f"Počet 2: .count(2) = {cisla.count(2)}")
print(f"Index 8: .index(8) = {cisla.index(8)}")

# sorted() - NOVÝ list (původní se nezmění!)
serazeny = sorted(cisla)
print(f"\nsorted(cisla) = {serazeny}")
print(f"Původní cisla = {cisla}")  # nezměněn!

# .sort() - změní PŮVODNÍ list
cisla.sort()
print(f"\nPo cisla.sort(): {cisla}")

# .reverse() - obrátí pořadí
cisla.reverse()
print(f"Po .reverse():   {cisla}")

# in - kontrola přítomnosti
print(f"\n5 in cisla: {5 in cisla}")
print(f"3 in cisla: {3 in cisla}")
````

### ⚠️ Proč potřebuji `.copy()`? (Reference vs. kopie)

Tohle je **jeden z nejčastějších bugů** u začátečníků. Když přiřadíte list přes `=`, **nevytvoříte kopii** – vytvoříte jen **druhý odkaz (referenci)** na ten **samý** objekt v paměti:

```
# ❌ Tohle NEKOPÍRUJE!
a = [1, 2, 3]
b = a              ← b ukazuje na STEJNÝ list jako a

┌─────────┐
│ [1,2,3] │  ← jeden objekt v paměti
└─────────┘
   ↑    ↑
   a    b           ← dvě jména, stejná data!
```

Když teď změníte `b`, **změní se i `a`** (protože to je ten samý list):

```python
b.append(99)
print(a)   # → [1, 2, 3, 99]  😱 a se taky změnilo!
```

### ✅ Řešení: `.copy()`

```python
a = [1, 2, 3]
b = a.copy()       # ← vytvoří NOVÝ nezávislý list

┌─────────┐     ┌─────────┐
│ [1,2,3] │     │ [1,2,3] │   ← dva RŮZNÉ objekty
└─────────┘     └─────────┘
     ↑               ↑
     a               b
```

 Alternativní zápisy:
| Způsob | Příklad |
|:---|:---|
| `.copy()` | `b = a.copy()` |
| Slicing | `b = a[:]` |
| `list()` | `b = list(a)` |

> ⚠️ **Pozor – mělká kopie (shallow copy):** `.copy()` kopíruje jen **první úroveň**. Pokud list obsahuje vnořené listy, ty se stále **sdílejí**! Pro plné zkopírování vnořených struktur potřebujete `import copy; copy.deepcopy(a)`.

### 🔄 Procházení listu cyklem `for`

List a `for` cyklus jsou **nerozlučná dvojice** – většinu práce s listy budete dělat právě cyklem:

```python
for prvek in muj_list:
    print(prvek)
```

Pokud potřebujete i **index** prvku, použijte funkci `enumerate()`:

```python
for index, prvek in enumerate(muj_list):
    print(f"{index}: {prvek}")
```

> 💡 **Tip:** `enumerate()` je Pythonická alternativa k `for i in range(len(lst))`. Používejte ji!

````python
# === Procházení listu ===
ovoce = ["jablko", "hruška", "banán", "třešeň"]

# Základní for cyklus
print("--- Základní for ---")
for o in ovoce:
    print(f"  {o}")

# S indexem pomocí enumerate()
print("\n--- S enumerate() ---")
for i, o in enumerate(ovoce):
    print(f"  [{i}] {o}")

# Sečtení čísel v listu
cisla = [10, 20, 30, 40]
soucet = 0
for c in cisla:
    soucet += c
print(f"\nSoučet {cisla} = {soucet}")
print(f"Totéž přes sum(): {sum(cisla)}")  # vestavěná funkce!
````

### 🎨 Datové typy v listu + vnořené listy

List může obsahovat **libovolné datové typy** – dokonce i **jiné listy** (= vnořený/nested list):

```
┌──────────────────────────────────────────┐
│  [1, "ahoj", True, [10, 20]]            │
│   ↑    ↑      ↑       ↑                 │
│  int  str    bool    list (vnořený)      │
└──────────────────────────────────────────┘
```

> 💡 **List comprehension** (zkrácený zápis): Pythonický způsob, jak vytvořit nový list z existujícího:
> ```python
> # Místo:
> druhe_mocniny = []
> for x in range(5):
>     druhe_mocniny.append(x ** 2)
>
> # Napíšeme:
> druhe_mocniny = [x ** 2 for x in range(5)]
> ```

````python
# === Různé datové typy v listu ===
mix = [42, "Python", 3.14, True, None]
print(f"Mix: {mix}")
for prvek in mix:
    print(f"  {prvek!r:10} → typ: {type(prvek).__name__}")

# === Vnořené listy (2D list / matice) ===
matice = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
print(f"\nMatice: {matice}")
print(f"Řádek 0:      {matice[0]}")
print(f"Prvek [1][2]:  {matice[1][2]}")  # řádek 1, sloupec 2 → 6

# === List comprehension ===
druhe_mocniny = [x ** 2 for x in range(6)]
print(f"\nDruhé mocniny: {druhe_mocniny}")

# S podmínkou
suda = [x for x in range(10) if x % 2 == 0]
print(f"Sudá čísla:    {suda}")
````

## 📦 2. Tuple (n-tice)

**Tuple** je **neměnná** (immutable) uspořádaná kolekce. Vypadá podobně jako list, ale používá **kulaté závorky** `()`:

```python
muj_tuple = (1, 2, 3)
muj_tuple = 1, 2, 3      # závorky nejsou povinné (ale doporučené)
prazdny = ()
jediny = (42,)            # POZOR: čárka je nutná pro tuple s jedním prvkem!
```

> ⚠️ **Pozor na závorku bez čárky:** `(42)` je jen číslo `42` v závorkách, **NE** tuple! Pro tuple s jedním prvkem musíte napsat `(42,)`

### 🔒 Co to znamená „neměnný"?

Po vytvoření **nelze** měnit prvky, přidávat ani odebírat:

```
✅ Lze:   čtení, indexování, slicing, for cyklus, in
❌ Nelze:  tuple[0] = 5, .append(), .pop(), .remove()
```

### 🤔 Kdy použít tuple místo listu?

| Situace | Proč tuple? |
|:---|:---|
| Souřadnice `(x, y)` | Hodnoty spolu logicky patří, neměníme je |
| Barva `(R, G, B)` | Fixní sada hodnot |
| Vrácení více hodnot z funkce | `return jmeno, vek` |
| Klíč ve slovníku | Tuple je hashable (list ne!) |

> 💡 **Tip:** Tuple je **rychlejší** než list a zabírá **méně paměti**. Pokud data nepotřebujete měnit, použijte tuple.

````python
# === Vytvoření tuple ===
bod = (3, 7)
barva = (255, 128, 0)
jediny = (42,)   # čárka!
bez_zavorek = 1, 2, 3

print(f"Bod:         {bod}         typ: {type(bod).__name__}")
print(f"Barva:       {barva}       typ: {type(barva).__name__}")
print(f"Jednoprvkový:{jediny}       typ: {type(jediny).__name__}")
print(f"(42) =       {(42)}          typ: {type((42)).__name__}")  # to NENÍ tuple!

# === Indexování (stejné jako u listu) ===
print(f"\nbod[0] = {bod[0]}")
print(f"bod[-1] = {bod[-1]}")
print(f"barva[0:2] = {barva[0:2]}")

# === NELZE měnit! ===
try:
    bod[0] = 99
except TypeError as e:
    print(f"\n❌ bod[0] = 99 → {e}")

try:
    bod.append(5)
except AttributeError as e:
    print(f"❌ bod.append(5) → {e}")
````

### 🎁 Tuple unpacking (rozbalení)

Jedná z nejužitečnějších vlastností tuple – **rozbalení do proměnných**:

```python
bod = (3, 7)
x, y = bod        # x = 3, y = 7
```

```
   (3, 7)
    ↓  ↓
    x  y
```

> 💡 **Záměna dvou proměnných** v Pythonu jedním řádkem: `a, b = b, a` – tohle funguje díky tuple unpacking!

````python
# === Tuple unpacking ===
souradnice = (10, 20)
x, y = souradnice
print(f"Souřadnice: {souradnice}")
print(f"x = {x}, y = {y}")

# Záměna hodnot
a, b = 5, 10
print(f"\nPřed záměnou: a={a}, b={b}")
a, b = b, a
print(f"Po záměně:    a={a}, b={b}")

# Různé datové typy v tuple
mix_tuple = (42, "ahoj", True, 3.14)
print(f"\nMix tuple: {mix_tuple}")
for prvek in mix_tuple:
    print(f"  {prvek!r:10} → {type(prvek).__name__}")
````

### 🔄 Procházení tuple cyklem `for`

Funguje **úplně stejně** jako u listu:

```python
for prvek in muj_tuple:
    print(prvek)
```

| Metoda / Funkce | Popis |
|:---|:---|
| `len(t)` | Počet prvků |
| `t.count(x)` | Počet výskytů `x` |
| `t.index(x)` | Index prvního výskytu `x` |
| `x in t` | Je `x` v tuple? |
| `sorted(t)` | Vrátí **seřazený list** (ne tuple!) |
| `tuple(lst)` | Vytvoří tuple z listu |

> 💡 Tuple má jen 2 metody: `.count()` a `.index()` – protože je neměnný, nepotřebuje víc.

````python
# === Procházení tuple ===
dny = ("Po", "Út", "St", "Čt", "Pá")

for den in dny:
    print(den, end=" ")
print()

# enumerate() funguje i s tuple
for i, den in enumerate(dny, start=1):
    print(f"  Den {i}: {den}")

# Metody tuple
cisla_t = (3, 1, 4, 1, 5, 9, 2, 6, 5)
print(f"\nTuple:     {cisla_t}")
print(f"len():     {len(cisla_t)}")
print(f"count(1):  {cisla_t.count(1)}")
print(f"index(5):  {cisla_t.index(5)}")
print(f"5 in:      {5 in cisla_t}")
print(f"sorted():  {sorted(cisla_t)}")  # vrátí list!
````

## 🔢 3. Set (množina)

**Set** je **neuspořádaná** kolekce **unikátních** prvků. Odpovídá matematickému pojmu **množina**.

```python
muj_set = {1, 2, 3}
prazdny = set()          # POZOR: {} vytvoří prázdný SLOVNÍK, ne set!
z_listu = set([1, 2, 2, 3])  # → {1, 2, 3} (duplikáty odstraněny)
```

### 🔑 Klíčové vlastnosti

| Vlastnost | Popis |
|:---|:---|
| **Neuspořádaný** | Prvky nemají index – nelze `set[0]` |
| **Bez duplikátů** | Každý prvek se vyskytuje **maximálně jednou** |
| **Měnitelný** | Lze přidávat (`.add()`) a odebírat (`.remove()`, `.discard()`) |
| **Hashable prvky** | Může obsahovat jen neměnné typy (`int`, `str`, `tuple` – ale NE `list`!) |

> 💡 **Kdy použít set?** Když potřebujete **unikátní hodnoty** nebo **rychlé vyhledávání** (`in` je u setu mnohem **rychlejší** než u listu!).

### ✏️ Metody pro úpravu

| Metoda | Popis | Příklad |
|:---|:---|:---|
| `.add(x)` | Přidá prvek | `s.add(5)` |
| `.remove(x)` | Odebere prvek (❌ KeyError pokud neexistuje) | `s.remove(5)` |
| `.discard(x)` | Odebere prvek (✅ bez chyby pokud neexistuje) | `s.discard(5)` |
| `.pop()` | Odebere a vrátí **náhodný** prvek | `s.pop()` |
| `.clear()` | Vyprázdní set | `s.clear()` |

````python
# === Vytvoření setu ===
ovoce = {"jablko", "hruška", "banán"}
print(f"Set: {ovoce}")
print(f"Typ: {type(ovoce).__name__}")

# Přidání prvku
ovoce.add("třešeň")
print(f"\nPo add('třešeň'):   {ovoce}")

# Přidání duplikátu – nic se nestane
ovoce.add("jablko")
print(f"Po add('jablko'):   {ovoce}")  # nezměněn!

# Odebrání
ovoce.remove("hruška")
print(f"Po remove('hruška'): {ovoce}")

# discard vs remove
ovoce.discard("kiwi")  # neexistuje, ale žádná chyba
print(f"Po discard('kiwi'): {ovoce}")

try:
    ovoce.remove("kiwi")  # neexistuje → KeyError!
except KeyError as e:
    print(f"\n❌ remove('kiwi') → KeyError: {e}")
````

````python
# === Duplikáty jsou automaticky odstraněny ===
cisla = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
unikatni = set(cisla)
print(f"List:       {cisla}")
print(f"Set z listu: {unikatni}")
print(f"Počet unikátních: {len(unikatni)}")

# Praktický příklad: odstranění duplikátů ze seznamu
jmena = ["Anna", "Petr", "Anna", "Jana", "Petr", "Anna"]
unikatni_jmena = list(set(jmena))  # set → zpět na list
print(f"\nPůvodní:  {jmena}")
print(f"Unikátní: {unikatni_jmena}")
````

### ⚙️ Množinové operace

Set podporuje **matematické operace s množinami** – to je jeho hlavní síla!

```
    A = {1, 2, 3, 4}     B = {3, 4, 5, 6}

    Sjednocení (union):        A | B = {1, 2, 3, 4, 5, 6}
    Průnik (intersection):     A & B = {3, 4}
    Rozdíl (difference):       A - B = {1, 2}
    Sym. rozdíl (xor):         A ^ B = {1, 2, 5, 6}
```

| Operátor | Metoda | Popis |
|:---:|:---|:---|
| `A \| B` | `A.union(B)` | Všechny prvky z obou |
| `A & B` | `A.intersection(B)` | Jen společné prvky |
| `A - B` | `A.difference(B)` | Prvky v A, které **nejsou** v B |
| `A ^ B` | `A.symmetric_difference(B)` | Prvky, které jsou **jen v jednom** |
| `A <= B` | `A.issubset(B)` | Je A **podmnožinou** B? |

````python
# === Množinové operace ===
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print(f"A = {A}")
print(f"B = {B}")
print(f"A | B (sjednocení): {A | B}")
print(f"A & B (průnik):     {A & B}")
print(f"A - B (rozdíl):     {A - B}")
print(f"B - A (rozdíl):     {B - A}")
print(f"A ^ B (sym.rozdíl): {A ^ B}")

# Praktický příklad
python_studenti = {"Anna", "Petr", "Jana", "Karel"}
web_studenti = {"Jana", "Karel", "Eva", "Tomáš"}

print(f"\nStudenti Pythonu: {python_studenti}")
print(f"Studenti webu:   {web_studenti}")
print(f"Oba kurzy:       {python_studenti & web_studenti}")
print(f"Alespoň jeden:   {python_studenti | web_studenti}")
print(f"Jen Python:      {python_studenti - web_studenti}")
````

for prvek in muj_set:
    print(prvek)          # pořadí NENÍ zaručeno!

## 📖 4. Dictionary / Dict (slovník)

**Slovník** ukládá data jako páry **klíč → hodnota** (key-value). Je to **nejpoužívanější** datová struktura vedle listu.

```python
muj_dict = {"klíč1": "hodnota1", "klíč2": "hodnota2"}
prazdny = {}
```

```
┌─────────────────────────────────────┐
│  slovnik = {                        │
│      "jméno":  "Anna",     ← str   │
│      "věk":    17,         ← int   │
│      "student": True       ← bool  │
│  }                                  │
│                                     │
│  klíč ──────► hodnota               │
│  (unikátní)   (libovolný typ)       │
└─────────────────────────────────────┘
```

### 🔑 Klíčové vlastnosti

| Vlastnost | Popis |
|:---|:---|
| **Klíč → Hodnota** | Přístup přes klíč (ne přes index!) |
| **Unikátní klíče** | Každý klíč se může vyskytovat **jen jednou** |
| **Měnitelný** | Lze přidávat, měnit a odebírat páry |
| **Hashable klíče** | Klíč musí být neměnný typ (`str`, `int`, `tuple`...) |
| **Libovolné hodnoty** | Hodnota může být **cokoliv** (i list, dict...) |

### ✏️ Přístup a úpravy

| Operace | Syntaxe | Popis |
|:---|:---|:---|
| Přístup | `d["klíč"]` | ❌ KeyError pokud neexistuje |
| Bezpečný přístup | `d.get("klíč", default)` | ✅ Vrátí default (nebo `None`) |
| Přidání / Změna | `d["klíč"] = hodnota` | Přidá nový nebo přepíše existující |
| Odebrání | `d.pop("klíč")` | Odebere a vrátí hodnotu |
| Smazání | `del d["klíč"]` | Smaže pár |
| Všechny klíče | `d.keys()` | Vrátí view klíčů |
| Všechny hodnoty | `d.values()` | Vrátí view hodnot |
| Všechny páry | `d.items()` | Vrátí view (klíč, hodnota) |

> 💡 **Tip:** Vždy používejte `.get()` místo `d["klíč"]`, pokud si nejste 100% jistí, že klíč existuje – vyhnete se `KeyError`!

````python
# === Vytvoření slovníku ===
student = {
    "jméno": "Anna",
    "věk": 17,
    "škola": "SPŠ Kladno",
    "průměr": 1.5
}
print(f"Slovník: {student}")
print(f"Typ:     {type(student).__name__}")

# === Přístup k hodnotám ===
print(f"\nJméno:  {student['jméno']}")
print(f"Věk:    {student['věk']}")

# .get() – bezpečný přístup
print(f"Email:  {student.get('email', 'neuvedeno')}")  # klíč neexistuje → default

# Přímý přístup na neexistující klíč → KeyError!
try:
    print(student["email"])
except KeyError as e:
    print(f"❌ KeyError: {e}")

# === Přidání a změna ===
student["email"] = "anna@skola.cz"  # přidání nového klíče
student["věk"] = 18                  # změna existujícího
print(f"\nPo úpravách: {student}")

# === Odebrání ===
prumer = student.pop("průměr")
print(f"Odebraný průměr: {prumer}")
print(f"Po pop:     {student}")

# === Duplikátní klíče ===
test = {"a": 1, "b": 2, "a": 99}  # druhá "a" přepíše první!
print(f"\nDuplikátní klíče: {test}")  # → {'a': 99, 'b': 2}
````

### 🔄 Procházení slovníku cyklem `for`

Slovník nabízí **tři způsoby** procházení:

```python
# 1. Jen klíče (výchozí)
for klic in slovnik:

# 2. Jen hodnoty
for hodnota in slovnik.values():

# 3. Klíče i hodnoty současně (nejčastější!)
for klic, hodnota in slovnik.items():
```

> 💡 **Tip:** `.items()` vrací tuple `(klíč, hodnota)` – proto se dá rozbalit do dvou proměnných díky tuple unpacking!

````python
# === Procházení slovníku ===
student = {"jméno": "Petr", "věk": 16, "třída": "IT2"}

# Jen klíče
print("--- Klíče ---")
for klic in student:
    print(f"  {klic}")

# Jen hodnoty
print("\n--- Hodnoty ---")
for hodnota in student.values():
    print(f"  {hodnota}")

# Klíče + hodnoty (nejužitečnější!)
print("\n--- Klíče + Hodnoty ---")
for klic, hodnota in student.items():
    print(f"  {klic}: {hodnota}")

# Kontrola existence klíče
print(f"\n'jméno' in student: {'jméno' in student}")
print(f"'email' in student: {'email' in student}")
````

### 🎨 Datové typy + vnořené slovníky

**Klíče** musí být hashable (`str`, `int`, `tuple`...), ale **hodnoty** mohou být cokoliv – včetně listů a dalších slovníků:

```python
# Slovník se seznamem jako hodnotou
predmety = {
    "IT": ["Python", "HTML", "SQL"],
    "mat": ["algebra", "geometrie"]
}

# Slovník ve slovníku (nested dict)
skola = {
    "student1": {"jméno": "Anna", "věk": 17},
    "student2": {"jméno": "Petr", "věk": 16}
}
```

> 💡 **Dict comprehension:** Podobně jako list comprehension:
> ```python
> ctverec = {x: x**2 for x in range(5)}
> # → {0: 0, 1: 1, 2: 4, 3: 9, 4: 16}
> ```

````python
# === Vnořené slovníky ===
trida = {
    "Anna": {"věk": 17, "průměr": 1.2, "kroužky": ["Python", "šachy"]},
    "Petr": {"věk": 16, "průměr": 2.1, "kroužky": ["fotbal"]},
    "Jana": {"věk": 17, "průměr": 1.5, "kroužky": ["Python", "keramika"]}
}

# Přístup k vnořeným datům
print(f"Anna věk:     {trida['Anna']['věk']}")
print(f"Petr kroužky: {trida['Petr']['kroužky']}")

# Procházení vnořeného slovníku
print("\n--- Přehled třídy ---")
for jmeno, info in trida.items():
    print(f"  {jmeno}: věk {info['věk']}, průměr {info['průměr']}")

# Dict comprehension
ctverec = {x: x**2 for x in range(6)}
print(f"\nČtverce: {ctverec}")

# List nelze jako klíč!
try:
    spatny = {[1, 2]: "hodnota"}
except TypeError as e:
    print(f"\n❌ List jako klíč: {e}")
````

## 🏷️ 5. Enum (výčtový typ)

**Enum** slouží k definici **pojmenovaných konstant** – sady pevných hodnot, ze kterých si vybíráte. Musíte ho **importovat** z modulu `enum`:

```python
from enum import Enum

class Barva(Enum):
    CERVENA = 1
    ZELENA = 2
    MODRA = 3
```

### 🤔 Proč nepoužít jen proměnné?

```python
# ❌ Bez Enum – chybové, těžko čitelné
barva = "červená"      # překlep → chyba
barva = "čirvená"      # Python to neodhalí!

# ✅ S Enum – bezpečné, přehledné
barva = Barva.CERVENA   # překlep → okamžitá chyba
barva = Barva.CIRVENA   # AttributeError!
```

> 💡 **Enum se hodí pro:** dny v týdnu, směry (sever/jih/...), stavy (aktivní/neaktivní), herní obtížnosti, barvy...

### 📋 Přístup k členům Enum

| Přístup | Příklad | Výsledek |
|:---|:---|:---|
| Přes jméno | `Barva.CERVENA` | `Barva.CERVENA` |
| Přes hodnotu | `Barva(1)` | `Barva.CERVENA` |
| Jméno členu | `Barva.CERVENA.name` | `"CERVENA"` |
| Hodnota členu | `Barva.CERVENA.value` | `FF0000` |

````python
from enum import Enum

# === Definice Enum ===
class Den(Enum):
    PONDELI = 1
    UTERY = 2
    STREDA = 3
    CTVRTEK = 4
    PATEK = 5
    SOBOTA = 6
    NEDELE = 7

# Přístup k členům
dnes = Den.STREDA
print(f"Dnes:    {dnes}")
print(f"Jméno:   {dnes.name}")
print(f"Hodnota: {dnes.value}")
print(f"Typ:     {type(dnes).__name__}")

# Přístup přes hodnotu
print(f"\nDen(5):  {Den(5)}")
print(f"Den(5).name: {Den(5).name}")

# Porovnávání
print(f"\ndnes == Den.STREDA: {dnes == Den.STREDA}")
print(f"dnes == Den.PATEK:  {dnes == Den.PATEK}")

# Enum NELZE měnit!
try:
    Den.PONDELI = 99
except AttributeError as e:
    print(f"\n❌ Změna: {e}")
````

````python
# === Procházení Enum cyklem for ===
print("Všechny dny:")
for den in Den:
    print(f"  {den.name:10} = {den.value}")

# Praktické použití v podmínce
dnes = Den.SOBOTA
if dnes in (Den.SOBOTA, Den.NEDELE):
    print(f"\n{dnes.name} → Víkend! 🎉")
else:
    print(f"\n{dnes.name} → Pracovní den.")
````

### 🔢 Datové typy hodnot v Enum

Hodnoty v Enum mohou být **libovolného typu** – nejen čísla:

```python
class Barva(Enum):
    CERVENA = "red"       # str hodnoty
    ZELENA = "green"
    MODRA = "blue"
```

> 💡 **`auto()`** – automatické přiřazení hodnot:
> ```python
> from enum import Enum, auto
> class Smer(Enum):
>     SEVER = auto()   # 1
>     JIH = auto()     # 2
>     VYCHOD = auto()  # 3
>     ZAPAD = auto()   # 4
> ```

````python
from enum import Enum, auto

# Enum se string hodnotami
class Barva(Enum):
    CERVENA = "red"
    ZELENA = "green"
    MODRA = "blue"

print(f"Barva.CERVENA.value = '{Barva.CERVENA.value}'")
print(f"Barva('green') = {Barva('green')}")

# Enum s auto()
class Obtiznost(Enum):
    LEHKA = auto()
    STREDNI = auto()
    TEZKA = auto()
    NIGHTMARE = auto()

print("\nObtížnosti:")
for o in Obtiznost:
    print(f"  {o.name:12} = {o.value}")
````

### 🧩 Enum v kombinaci s jinými kolekcemi

Enum se často používá jako **klíč ve slovníku** nebo **prvek v listu/setu**:

````python
# === Enum v kolekcích ===
class Predmet(Enum):
    PYTHON = "Python"
    WEB = "Web design"
    SQL = "Databáze"

# Enum jako klíč ve slovníku
ucebna = {
    Predmet.PYTHON: "B203",
    Predmet.WEB: "A101",
    Predmet.SQL: "B205"
}
print("Učebny:")
for predmet, mistnost in ucebna.items():
    print(f"  {predmet.value:15} → {mistnost}")

# Enum v listu
dnesni_rozvrh = [Predmet.PYTHON, Predmet.SQL, Predmet.PYTHON]
print(f"\nDnešní rozvrh: {[p.value for p in dnesni_rozvrh]}")

# Enum v setu
unikatni_predmety = set(dnesni_rozvrh)
print(f"Unikátní:      {[p.value for p in unikatni_predmety]}")
````

## 📋 Souhrnná tabulka

### Porovnání kolekcí

| Vlastnost | `list` | `tuple` | `set` | `dict` | `Enum` |
|:---|:---:|:---:|:---:|:---:|:---:|
| **Syntaxe** | `[1,2,3]` | `(1,2,3)` | `{1,2,3}` | `{"a":1}` | `class X(Enum)` |
| **Měnitelný?** | ✅ Ano | ❌ Ne | ✅ Ano | ✅ Ano | ❌ Ne |
| **Indexování?** | ✅ `lst[0]` | ✅ `tpl[0]` | ❌ Ne | ✅ `d["k"]` | ✅ `E.JMENO` |
| **Duplikáty?** | ✅ Ano | ✅ Ano | ❌ Ne | ❌ klíče Ne | ❌ Ne |
| **Uspořádaný?** | ✅ Ano | ✅ Ano | ❌ Ne | ✅ (od 3.7) | ✅ Ano |
| **For cyklus?** | ✅ | ✅ | ✅ | ✅ | ✅ |
| **`in` operátor?** | ✅ (pomalý) | ✅ (pomalý) | ✅ (rychlý!) | ✅ klíče | ✅ |

### Kdy co použít?

| Potřebuji... | Použij |
|:---|:---|
| Upravitelný seznam prvků | `list` |
| Neměnnou skupinu hodnot (souřadnice, barvy) | `tuple` |
| Unikátní hodnoty, rychlé vyhledávání | `set` |
| Párování klíč → hodnota | `dict` |
| Pojmenované konstanty (stavy, směry, typy) | `Enum` |

### Konverze mezi typy

```python
lst = [1, 2, 2, 3]
tpl = tuple(lst)      # → (1, 2, 2, 3)
st  = set(lst)        # → {1, 2, 3}
lst2 = list(st)       # → [1, 2, 3]
lst3 = list(tpl)      # → [1, 2, 2, 3]
```

> 🎯 **Zapamatuj si:**
> - **List** = když potřebuješ měnit a záleží na pořadí
> - **Tuple** = když se data nemění (rychlejší, bezpečnější)
> - **Set** = unikátní hodnoty + množinové operace
> - **Dict** = nejflexibilnější – mapování čehokoliv na cokoliv
> - **Enum** = pojmenované konstanty pro čistší a bezpečnější kód
