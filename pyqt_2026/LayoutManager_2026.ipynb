# 📐 PyQt5 – Lekce 4: Layout Managery

Layout Manager rozhoduje o tom, jak jsou prvky v okně **rozmístěny a zvětšovány**.
Bez něj bychom museli každému prvku ručně nastavit souřadnice X, Y – což je při
změně velikosti okna noční můra.

---

## 📦 Přehled Layout Managerů

| Layout | Rozmístění | Typické použití |
|---|---|---|
| `QHBoxLayout` | Vedle sebe → | Panel nástrojů, tlačítka v řadě |
| `QVBoxLayout` | Pod sebou ↓ | Formulář, seznam položek |
| `QGridLayout` | Mřížka (řádky × sloupce) | Kalkulačka, piškvorky, formulář |
| `QFormLayout` | Dvojsloupcový formulář | Přihlašování, nastavení |

> 💡 Layouts lze do sebe **zanořovat** – do jednoho přidáme jiný přes `.addLayout()`.


---

## 🟢 Ukázka 1 – `QHBoxLayout` (horizontální)

Prvky jsou seřazeny vedle sebe v jedné řadě.

````python
from PyQt5 import QtWidgets

app = QtWidgets.QApplication([])
okno = QtWidgets.QWidget()
okno.setWindowTitle("QHBoxLayout")
layout = QtWidgets.QHBoxLayout()
okno.setLayout(layout)

for text in ["Tlačítko 1", "Tlačítko 2", "Tlačítko 3"]:
    layout.addWidget(QtWidgets.QPushButton(text))

okno.show()
app.exec()
````

## 🟢 Ukázka 2 – `QVBoxLayout` (vertikální)

Prvky jsou seřazeny pod sebou.

````python
from PyQt5 import QtWidgets

app = QtWidgets.QApplication([])
okno = QtWidgets.QWidget()
okno.setWindowTitle("QVBoxLayout")
layout = QtWidgets.QVBoxLayout()
okno.setLayout(layout)

for text in ["Nový soubor", "Otevřít soubor", "Uložit", "Ukončit"]:
    layout.addWidget(QtWidgets.QPushButton(text))

okno.show()
app.exec()
````

## 🟡 Ukázka 3 – `QGridLayout` (mřížka)

Každý prvek umístíme na konkrétní `(řádek, sloupec)`.

````python
from PyQt5 import QtWidgets

app = QtWidgets.QApplication([])
okno = QtWidgets.QWidget()
okno.setWindowTitle("QGridLayout – Kalkulačka (jen vzhled)")
okno.resize(250, 300)

grid = QtWidgets.QGridLayout()
okno.setLayout(grid)

# Displej – prvek přes celou šířku (row, col, rowspan, colspan)
displej = QtWidgets.QLabel("0")
displej.setStyleSheet("font-size: 24px; background: #eee; padding: 5px;")
grid.addWidget(displej, 0, 0, 1, 4)

# Tlačítka
tlacitka = [
    ["7", "8", "9", "/"],
    ["4", "5", "6", "*"],
    ["1", "2", "3", "-"],
    ["0", ".", "=", "+"],
]

for r, radek in enumerate(tlacitka):
    for s, text in enumerate(radek):
        grid.addWidget(QtWidgets.QPushButton(text), r + 1, s)

okno.show()
app.exec()
````

## 🟡 Ukázka 4 – `QFormLayout` (formulář)

Ideální pro formuláře – každý řádek má vlevo popisek, vpravo vstup.

````python
from PyQt5 import QtWidgets

app = QtWidgets.QApplication([])
okno = QtWidgets.QWidget()
okno.setWindowTitle("Registrace")
okno.resize(320, 200)

hlavni = QtWidgets.QVBoxLayout()
okno.setLayout(hlavni)

nadpis = QtWidgets.QLabel("📋 Registrační formulář")
nadpis.setStyleSheet("font-size: 16px; font-weight: bold;")
hlavni.addWidget(nadpis)

formular = QtWidgets.QFormLayout()
vstup_jmeno = QtWidgets.QLineEdit()
vstup_email = QtWidgets.QLineEdit()
vstup_heslo = QtWidgets.QLineEdit()
vstup_heslo.setEchoMode(QtWidgets.QLineEdit.Password)  # Maskování hesla

formular.addRow("Jméno:", vstup_jmeno)
formular.addRow("E-mail:", vstup_email)
formular.addRow("Heslo:", vstup_heslo)
hlavni.addLayout(formular)

hlavni.addWidget(QtWidgets.QPushButton("Registrovat"))

okno.show()
app.exec()
````

## 🔴 Ukázka 5 – Zanořené Layouty (Nesting)

Zanořením Layoutů do sebe dosáhneme libovolně složitého rozmístění.

````python
from PyQt5 import QtWidgets

app = QtWidgets.QApplication([])
okno = QtWidgets.QWidget()
okno.setWindowTitle("Zanořené layouty")
okno.resize(400, 250)

# Hlavní svislý layout
hlavni = QtWidgets.QVBoxLayout()
okno.setLayout(hlavni)

# Řádek 1: Nadpis
hlavni.addWidget(QtWidgets.QLabel("📧 Napište nám"))

# Řádek 2: Pole pro zadání e-mailu
radek_email = QtWidgets.QHBoxLayout()
radek_email.addWidget(QtWidgets.QLabel("E-mail:"))
radek_email.addWidget(QtWidgets.QLineEdit())
hlavni.addLayout(radek_email)   # Přidáme layout do layoutu!

# Řádek 3: Textová oblast pro zprávu
hlavni.addWidget(QtWidgets.QTextEdit())

# Řádek 4: Tlačítka
radek_tlacitka = QtWidgets.QHBoxLayout()
radek_tlacitka.addWidget(QtWidgets.QPushButton("Odeslat"))
radek_tlacitka.addWidget(QtWidgets.QPushButton("Zrušit"))
hlavni.addLayout(radek_tlacitka)

okno.show()
app.exec()
````
