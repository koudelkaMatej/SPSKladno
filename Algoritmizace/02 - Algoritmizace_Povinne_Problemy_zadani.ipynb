## 5. Počítání kvadratických rovnic

Napište algoritmus, který vyřeší **kvadratickou rovnici** `a * x² + b * x + c = 0` pomocí diskriminantu.

**Diskriminant:** $D = b^2 - 4ac$

- $D < 0$ → žádné reálné řešení
- $D = 0$ → jedno řešení: $x = \frac{-b}{2a}$
- $D > 0$ → dvě řešení: $x_1 = \frac{-b + \sqrt{D}}{2a}$, $x_2 = \frac{-b - \sqrt{D}}{2a}$

**Testovací případy:** `a=1,b=2,c=-1` (dvě řešení), `a=1,b=2,c=1` (jedno), `a=1,b=2,c=8` (žádné)
