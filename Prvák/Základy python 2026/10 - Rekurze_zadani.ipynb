## 📚 Příklad 3 – Součet seznamu

Sečíst všechna čísla v seznamu – rekurzivně:

```python
def soucet_seznamu(seznam):
    if len(seznam) == 0:       # základní případ – prázdný seznam
        return 0
    return seznam[0] + soucet_seznamu(seznam[1:])  # první prvek + součet zbytku
```

```
soucet_seznamu([3, 1, 4, 2])
  → 3 + soucet_seznamu([1, 4, 2])
        → 1 + soucet_seznamu([4, 2])
              → 4 + soucet_seznamu([2])
                    → 2 + soucet_seznamu([])
                          → 0
                    → 2 + 0 = 2
              → 4 + 2 = 6
        → 1 + 6 = 7
  → 3 + 7 = 10
```

## ⚠️ Rekurze vs. cyklus

| | Rekurze | Cyklus (iterace) |
|:---|:---|:---|
| **Čitelnost** | U některých problémů přehlednější | U jednoduchých problémů přehlednější |
| **Paměť** | Každé volání zabírá místo na zásobníku | Konstantní paměť |
| **Rychlost** | Může být pomalejší (režie volání) | Obvykle rychlejší |
| **Limit** | Python má limit ~1000 volání | Bez limitu |

> 💡 **Pravidlo:** Každou rekurzi lze přepsat na cyklus a naopak. Používejte rekurzi tam, kde je řešení **přirozenější a přehlednější**.

## Úkol 1 – Mocnina rekurzivně

Napište rekurzivní funkci `mocnina(zaklad, exponent)`, která vypočítá $základ^{exponent}$.

- Základní případ: $n^0 = 1$
- Rekurzivní případ: $n^k = n \times n^{k-1}$

```python
print(mocnina(2, 5))   # 32
print(mocnina(3, 0))   # 1
print(mocnina(5, 3))   # 125
```

## Úkol 2 – Součet číslic

Napište rekurzivní funkci `soucet_cislic(n)`, která vrátí **součet číslic** celého čísla `n`.

- Základní případ: jednociferné číslo → vrátíme ho
- Rekurzivní případ: poslední číslice (`n % 10`) + součet zbytku (`n // 10`)

```python
print(soucet_cislic(123))    # 6  (1+2+3)
print(soucet_cislic(9))      # 9
print(soucet_cislic(4567))   # 22 (4+5+6+7)
```

## Úkol 3 – Fibonacci

**Fibonacciho posloupnost:** 0, 1, 1, 2, 3, 5, 8, 13, 21, …

Každé číslo je součtem dvou předchozích:
- $F(0) = 0$
- $F(1) = 1$
- $F(n) = F(n-1) + F(n-2)$

Napište rekurzivní funkci `fibonacci(n)`, která vrátí $n$-té Fibonacciho číslo.

```python
print(fibonacci(0))   # 0
print(fibonacci(1))   # 1
print(fibonacci(7))   # 13
print(fibonacci(10))  # 55
```

## Úkol 4 – Palindrom

Napište rekurzivní funkci `je_palindrom(text)`, která vrátí `True`, pokud je řetězec **palindrom** (čte se stejně odpředu i odzadu).

- Základní případ: řetězec délky 0 nebo 1 → `True`
- Rekurzivní případ: první a poslední znak se shodují **a** vnitřek je také palindrom

```python
print(je_palindrom("radar"))    # True
print(je_palindrom("python"))   # False
print(je_palindrom("abacaba"))  # True
print(je_palindrom("a"))        # True
```

- Základní případ: prázdný řetězec → vrátíme ho
