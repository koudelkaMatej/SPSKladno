# Jednoduchá aplikace, když kliknete přičte se čas

```python
import pygame
import sys

pygame.init()

# Nastavení obrazovky
screen = pygame.display.set_mode((800, 600))
pygame.display.set_caption("Bomb Timer")

# Barvy
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)

# Font
font = pygame.font.Font(None, 36)

# Čas pro zneškodnění bomby (v milisekundách)
BOMB_TIMER = 10000  # 10 sekund
TIME_BONUS = 5000   # 5 sekund

def main():
    clock.tick(1) # kolikrát za vteřinu toto udělám (znamená, že čas se mi bude posouvat o 1 sekundu -> znamená, že odpočet prodloužím když kliknu,
    # ale když kliknu v 0.5 vteřině, bomba bouchne za 0.9 sekundy tak bomba bouchne, jelikož se mi nestihne zkontrolovat kliknutí, které proběhne o 0,5 později než můj klik
    start_time = pygame.time.get_ticks()

    running = True
    bonus_applied = False  # Indikátor, zda byl bonus již přidán
    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
                pygame.quit()
                sys.exit()

            if event.type == pygame.MOUSEBUTTONDOWN:
                # Pokud uživatel kliknul na tlačítko a bonus ještě nebyl přidán
                start_time += TIME_BONUS  # Přičtení bonusu k začátkovému času

        current_time = pygame.time.get_ticks()  # Aktuální čas
        remaining_time = max(0, BOMB_TIMER - (current_time - start_time))  # Výpočet zbývajícího času

        screen.fill(BLACK)  # Vyplníme obrazovku černou barvou

        # Zobrazení zbývajícího času na obrazovce
        text = font.render(f"Zbývající čas: {remaining_time // 1000} s", True, WHITE)
        screen.blit(text, (200, 250))

        pygame.display.flip()

        # Omezení rychlosti smyčky
        pygame.time.Clock().tick(30)

if __name__ == "__main__":
    main()
    pygame.quit()
```

# Vykreslení objektů pomocí kliku

Kruh

```python
import pygame
from pygame.locals import *

pygame.init()

window = pygame.display.set_mode((600, 600))

window.fill((255, 255, 255))

circle_positions = []
circle_radius = 60
color = (0, 0, 255)

run = True
while run:
	for event in pygame.event.get():
		if event.type == QUIT:
			run = False
			
		elif event.type == MOUSEBUTTONDOWN:
			position = event.pos
			circle_positions.append(position)
		
	for position in circle_positions:
		pygame.draw.circle(window, color, position,
						circle_radius)
	pygame.display.update()
pygame.quit()
```

# Animace

```python
import pygame

pygame.init()
clock = pygame.time.Clock() # pro FPS
screen = pygame.display.set_mode((400, 600))

running = True
color = (255, 0, 0)
a = 50
b = 200 # vykreslím a už neměním (statické části)
vlevo = False

while running: # vše co je zde měním dynamicky dle ticku
    clock.tick(60) # FPS
    screen.fill((0, 0, 0))
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
    #                            barva     pozice   polomer
    pygame.draw.circle(screen, (255, 255, 255), (a, b), 20) 
    pygame.draw.line(screen, (255, 102, 0), [25, 220],  [370, 220], 5)
    if vlevo == False and a >350:
        vlevo = True
    elif vlevo and a < 45:
        vlevo = False
    if vlevo:
        a -= 1
    else:
        a += 1                
    
    pygame.display.update()

pygame.quit()
```

Obrazek a rotace

```python
import pygame

pygame.init()
clock = pygame.time.Clock()  # pro FPS
screen = pygame.display.set_mode((400, 600))

running = True
a = 50
b = 200  # vykreslím a už neměním (statické části)
vlevo = False
ball = pygame.image.load('ball_image.png')  # NAČTENÍ OBRÁZKU
ball = pygame.transform.scale(ball, (40, 40))  # menim velikost obrazku
rotation_angle = 0
while running:  # vše co je zde měním dynamicky dle ticku
    clock.tick(60)  # FPS
    screen.fill((0, 0, 0))
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
    # Rotate the ball image
    rotated_ball = pygame.transform.rotate(ball, rotation_angle)
    ball_rect = rotated_ball.get_rect(center=(a, b))

    screen.blit(rotated_ball, ball_rect)  # vykreslení rotovaného obrázku
    pygame.draw.line(screen, (255, 102, 0), [25, b+20], [370, b+20], 5)

    if vlevo:
        a -= 1
        rotation_angle += 15  # Rotate clockwise 
        if a <= 50:
            vlevo = False
    else:
        a += 1
        rotation_angle -= 15  # Rotate counterclockwise
        if a >= 350:
            vlevo = True

    pygame.display.update()

pygame.quit()
```

```python
# import pygame module
import pygame

# import sys library
import sys

# initializing pygame
pygame.init()

clock = pygame.time.Clock()

# Set the window screen size
display_screen = pygame.display.set_mode((500, 500))

# add font style and size
base_font = pygame.font.Font(None, 40)


# stores text taken by keyboard
user_text = ''

# set left, top, width, height in 
# Pygame.Rect()
input_rect = pygame.Rect(200, 200, 140, 32)
color_active = pygame.Color("lightskyblue")
color_passive = pygame.Color("gray15")
color = color_passive

active = False

while True:
	
	for event in pygame.event.get():
		if event.type == pygame.QUIT:
			pygame.quit()
			sys.exit()

		# when mouse collides with the rectangle
		# make active as true
		if event.type == pygame.MOUSEBUTTONDOWN:
			if input_rect.collidepoint(event.pos):
				active = True

		# if the key is physically pressed down
		if event.type == pygame.KEYDOWN:
			if event.key == pygame.K_BACKSPACE:
				
				# stores text except last letter
				user_text = user_text[0:-1]
			else:
				user_text += event.unicode

	display_screen.fill((0, 0, 0))

	if active:
		color = color_active
	else:
		color = color_passive

	pygame.draw.rect(display_screen, color, input_rect)
	
	# render the text
	text_surface = base_font.render(user_text, True, (255, 255, 255))
	display_screen.blit(text_surface, (input_rect.x + 5, input_rect.y + 5))
	input_rect.w = max(100, text_surface.get_width() + 10)
	pygame.display.flip()
	clock.tick(60)
```

# Bomba čas a klik

```python
import pygame
import sys
pygame.init()
clock = pygame.time.Clock()
screen = pygame.display.set_mode((500, 500))
base_font = pygame.font.Font(None, 40)
bomb_timer = 5  # seconds
text = 'Bomba vybuchne za: {}'.format(bomb_timer)
bomb_timer_text = base_font.render(text, True, (0, 0, 0))
bomb_timer_text_rect = bomb_timer_text.get_rect(center=(250, 250))
active = False
running = True
while running: 
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit() # quit pygame
            sys.exit() # exit the program
        if event.type == pygame.MOUSEBUTTONDOWN:
            bomb_timer += 5
    screen.fill((255, 255, 255)) # fill the screen with white color
    # time in seconds until the bomb explodes on two decimal
    text = 'Bomba vybuchne za: {:.2f}'.format(bomb_timer)       
    bomb_timer_text = base_font.render(text, True, (0, 0, 0))
    screen.blit(bomb_timer_text, bomb_timer_text_rect) # display the text on the screen
    bomb_timer -= 1/60
    if bomb_timer < 0:
        running = False
    pygame.display.flip()
    clock.tick(60)
```
