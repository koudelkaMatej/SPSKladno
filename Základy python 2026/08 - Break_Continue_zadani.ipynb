# 🛑 `break`, `continue` a `pass`

V minulé lekci jsme se naučili cykly `for` a `while`. Dnes se podíváme na tři klíčová slova, která nám dávají **jemnější kontrolu** nad průběhem cyklu:

| Příkaz | Co dělá | Funguje v |
|:---|:---|:---|
| `break` | **Okamžitě ukončí** celý cyklus | `for` i `while` |
| `continue` | **Přeskočí** zbytek aktuálního průchodu a jde na další | `for` i `while` |
| `pass` | **Nedělá nic** – zástupný příkaz | kdekoliv |

---

## 🔴 `break` – okamžité ukončení cyklu

Příkaz `break` **okamžitě vyskočí** z cyklu, ve kterém se nachází. Žádný další kód v těle cyklu se neprovede a program pokračuje **prvním řádkem za cyklem**.

```
while podmínka:           for prvek in kolekce:
    ...                       ...
    if něco:                  if něco:
        break  ──►  KONEC        break  ──►  KONEC
    ...  (přeskočeno)         ...  (přeskočeno)
                          
kód za cyklem  ◄──────    kód za cyklem  ◄──────
```

> 💡 `break` se často používá s `while True:` – vytvoříme „nekonečný" cyklus a ukončíme ho **zevnitř**, když nastane správná podmínka.

### Příklad: `break` ve `while`

````python
# === break ve while ===
x = 1
while True:  # podmínka je vždy True → cyklus by sám nikdy neskončil
    if x == 5:
        break  # ukončíme cyklus, když x dosáhne 5
    print(x)
    x += 1
print("Konec, x:", x)
````

Jak to funguje krok po kroku:

```
x = 1 → print(1) → x = 2 → print(2) → x = 3 → print(3) → x = 4 → print(4)
→ x = 5 → break! → skočí na "Konec, x: 5"
```

> ⚠️ Všimněte si, že číslo **5 se nevypíše** – `break` je **před** `print(x)`, takže se `print` u x=5 nikdy neprovede.

### Příklad: `break` ve `for`

````python
# === break ve for ===
for kamarad in ["Adam", "Bert", "Cyril", "Dan"]:
    print("Můj kamarád je:", kamarad)
    if kamarad == "Bert":
        break  # po Bertovi končíme
print("Konec")
````

> 💡 **break** ukončí **pouze nejbližší cyklus**, ve kterém se nachází. Pokud máte cyklus v cyklu (vnořené cykly), `break` ukončí jen ten vnitřní.

### `break` ve vnořených cyklech

Při vnořených cyklech `break` ukončí **pouze vnitřní** cyklus:

```python
for i in range(3):           # vnější cyklus
    for j in range(5):       # vnitřní cyklus
        if j == 2:
            break            # ukončí JEN vnitřní cyklus (for j)
        print(f"i={i}, j={j}")
    print(f"Vnitřní cyklus pro i={i} skončil")  # provede se vždy
```

````python
# === break ve vnořených cyklech ===
for i in range(3):
    for j in range(5):
        if j == 2:
            break  # ukončí JEN vnitřní cyklus
        print(f"  i={i}, j={j}")
    print(f"Vnitřní cyklus pro i={i} skončil")
````

## 🟡 `continue` – přeskočení aktuálního průchodu

Na rozdíl od `break` (který ukončí celý cyklus), `continue` **přeskočí zbytek aktuálního průchodu** a cyklus pokračuje dalším opakováním:

- U `for` → přejde na **další prvek** kolekce
- U `while` → skočí zpět na **kontrolu podmínky**

```
while podmínka:           for prvek in kolekce:
    ...                       ...
    if něco:                  if něco:
        continue  ──►  ↑         continue  ──►  další prvek
    ...  (přeskočeno)  │      ...  (přeskočeno)
    ...  (přeskočeno)  │      ...  (přeskočeno)
         ──────────────┘
```

### Příklad: `continue` ve `for`

````python
# === continue – přeskočíme Berta ===
for kamarad in ["Adam", "Bert", "Cyril", "Dan", "Emil"]:
    if kamarad == "Bert":
        continue  # přeskočí Berta, pokračuje Cyrilem
    print("Můj kamarád je:", kamarad)
print("Konec")
````

### Příklad: `continue` ve `while`

````python
# === continue – vypíšeme jen lichá čísla ===
i = 0
while i < 10:
    i += 1
    if i % 2 == 0:
        continue  # sudá čísla přeskočíme
    print(f"  Liché: {i}")
print("Hotovo")
````

> ⚠️ **Pozor u `while`:** příkaz, který mění řídicí proměnnou (např. `i += 1`), musí být **před** `continue`! Jinak ho `continue` přeskočí a vznikne **nekonečný cyklus**:
>
> ```python
> # ❌ ŠPATNĚ – nekonečný cyklus!
> i = 0
> while i < 10:
>     if i % 2 == 0:
>         continue   # přeskočí i += 1 → i zůstane 0 navždy!
>     print(i)
>     i += 1
> ```

## 🔀 Kombinace `break` a `continue`

`break` a `continue` se dají kombinovat v jednom cyklu – každý reaguje na jinou podmínku:

````python
# === kombinace break + continue ===
for kamarad in ["Adam", "Bert", "Cyril", "Dan", "Emil"]:
    if kamarad == "Bert":
        continue  # Berta přeskočíme
    print("Můj kamarád je:", kamarad)
    if kamarad == "Dan":
        break  # u Dana končíme celý cyklus
print("Konec")
````

Výstup:
```
Můj kamarád je: Adam      ← Adam projde normálně
                           ← Bert je přeskočen (continue)
Můj kamarád je: Cyril     ← Cyril projde normálně
Můj kamarád je: Dan       ← Dan se vypíše, pak break
Konec                      ← Emil se už neprovede
```

## ⚪ `pass` – prázdné tělo (zástupný příkaz)

V Pythonu **nemůže** být blok kódu prázdný. Pokud chcete cyklus, funkci nebo `if` větev zatím nechat prázdnou, použijte `pass`:

```python
while podmínka:
    pass    # zatím prázdné – dodělám později

for x in seznam:
    pass    # TODO: implementovat

if podmínka:
    pass    # zatím nic
```

> 💡 `pass` se hodí při **postupném vývoji** – vytvoříte strukturu programu a detaily doplníte později.

````python
# === pass – zástupný příkaz ===
for i in range(5):
    pass  # TODO: implementovat logiku

print("pass nic nedělá – program pokračuje dál")
````

## 🔁 Cyklus + `else`

Méně známá vlastnost Pythonu – za `for` i `while` cyklus můžete přidat blok `else`. Ten se provede **jen pokud cyklus skončil přirozeně** (podmínka přestala platit / kolekce došla), ale **NE pokud skončil přes `break`**:

```python
for prvek in kolekce:       while podmínka:
    if něco:                    if něco:
        break                       break
else:                       else:
    # jen bez break             # jen bez break
```

> 💡 Můžete si to přeložit jako: *"Pokud cyklus proběhl celý **aniž** jsem ho přerušil, udělej tohle."*

````python
# === for + else – hledání čísla ===
hledane = 7
cisla = [2, 4, 6, 8, 10]

for cislo in cisla:
    if cislo == hledane:
        print(f"Našel jsem {hledane}!")
        break
else:
    # Provede se JEN pokud cyklus doběhl BEZ break
    print(f"Číslo {hledane} nebylo v seznamu nalezeno")

# Zkuste změnit hledane na 6 a uvidíte rozdíl
````

````python
# === while + else ===
pokus = 1
max_pokusu = 3
heslo = "python"

while pokus <= max_pokusu:
    vstup = input(f"Pokus {pokus}/{max_pokusu} – Zadej heslo: ")
    if vstup == heslo:
        print("✅ Správné heslo!")
        break
    pokus += 1
else:
    # Provede se, pokud while skončil přirozeně (pokus > max_pokusu)
    print("❌ Příliš mnoho pokusů – přístup zamítnut!")
````

## 🧩 Typické vzory s `break` a `continue`

### Vzor 1: Hledání v kolekci

Projdeme kolekci a jakmile najdeme, co hledáme, **ukončíme** cyklus – nemá smysl procházet zbytek:

```python
for student in studenti:
    if student["prumer"] < 1.5:
        print(f"Nejlepší student: {student['jmeno']}")
        break
```

### Vzor 2: Filtrování (přeskočení nevhodných prvků)

Zpracujeme **jen některé** prvky kolekce – nevyhovující přeskočíme:

```python
for cislo in cisla:
    if cislo < 0:
        continue  # záporná čísla přeskočíme
    print(f"Odmocnina {cislo} = {cislo ** 0.5:.2f}")
```

### Vzor 3: Nekonečný cyklus s podmínkou ukončení

Klasický `while True` + `break` – běží, dokud nenastane speciální událost:

```python
while True:
    prikaz = input("Zadej příkaz: ")
    if prikaz == "konec":
        break
    zpracuj(prikaz)
```

## 📋 Shrnutí

### Přehled příkazů

| Příkaz | Význam | Funguje v |
|:---|:---|:---|
| `break` | Okamžitě **ukončí** celý cyklus | `for`, `while` |
| `continue` | **Přeskočí** zbytek průchodu, jde na další | `for`, `while` |
| `pass` | **Nedělá nic** – zástupný příkaz | kdekoliv |
| `else` (za cyklem) | Provede se, pokud cyklus **neskončil** přes `break` | `for`, `while` |

### Vizuální přehled

```
while podmínka:          ◄─────────────────┐
    příkaz_1                               │
    if podmínka_A:                         │
        continue   ────────────────────────┘  (zpět na podmínku)
    příkaz_2                               
    if podmínka_B:                         
        break      ─────────┐                (ven z cyklu)
    příkaz_3                │
else:                       │
    toto_bez_break          │  (jen při přirozeném konci)
                            │
kód_za_cyklem  ◄────────────┘
```

### Kdy co použít?

| Potřebuji… | Použij |
|:---|:---|
| Ukončit cyklus při splnění podmínky | `break` |
| Přeskočit nevhodné prvky | `continue` |
| Zatím prázdný blok kódu | `pass` |
| Zjistit, zda cyklus proběhl celý | `for`/`while` + `else` |
| Nekonečný cyklus s podmínkou ukončení | `while True:` + `break` |

### Časté chyby

| Chyba | Problém | Řešení |
|:---|:---|:---|
| `continue` za aktualizací proměnné | Proměnná se nezmění → nekonečný cyklus | Přesuňte `i += 1` **před** `continue` |
| `break` ve vnořeném cyklu | Ukončí jen vnitřní cyklus | Použijte vlajku (`found = True`) + `break` ve vnějším |
| Záměna `break` a `continue` | Ukončí cyklus místo přeskočení | Zkontrolujte, co chcete – skončit vs. přeskočit |

> 🎯 **Zapamatuj si:** `break` = „dost, končím celý cyklus", `continue` = „tohle přeskakuji, ale pokračuji dál". Oba příkazy fungují ve `for` i `while`!
