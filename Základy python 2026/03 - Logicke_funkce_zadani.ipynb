# 🔗 Logické operátory

## Co jsou logické operátory?

V minulé lekci jsme se naučili větvení pomocí `if-elif-else`. Podmínky, které jsme zatím používali, byly jednoduché – porovnávali jsme jednu hodnotu s druhou. V praxi ale často potřebujeme **kombinovat více podmínek dohromady**.

K tomu slouží **logické operátory** (logical operators):

| Operátor | Význam | Stručně |
|:---:|---|---|
| `and` | **A zároveň** | Obě podmínky musí platit |
| `or` | **Nebo** | Stačí, aby platila alespoň jedna |
| `not` | **Negace (opak)** | Obrátí `True` ↔ `False` |

> 💡 Logické operátory pracují s hodnotami typu **`bool`** (`True`/`False`) a vrací opět `bool`.

---

## 1️⃣ Operátor `and` – „A zároveň"

Operátor `and` vrací `True` **pouze tehdy, když jsou obě podmínky pravdivé**. Pokud je alespoň jedna `False`, celý výraz je `False`.

### Pravdivostní tabulka

| Podmínka A | Podmínka B | A `and` B |
|:---:|:---:|:---:|
| `True` | `True` | ✅ `True` |
| `True` | `False` | ❌ `False` |
| `False` | `True` | ❌ `False` |
| `False` | `False` | ❌ `False` |

### Příklad z praxe
Chcete jet autem → musíte být **dospělí** (`vek >= 18`) **a zároveň** mít **řidičský průkaz** (`ma_ridicak == True`):

```python
if vek >= 18 and ma_ridicak:
    print("Můžeš řídit!")
```

````python
# === Jednoduchá ukázka AND ===
# Jaký bude výsledek? Nejprve si tipněte, pak spusťte!
if (7 < 5) and (8 > 7):
    print("Podmínka je splněna")
else:
    print("Podmínka NENÍ splněna")

# Vysvětlení:
# (7 < 5) → False
# (8 > 7) → True  
# False and True → False  ← stačí, aby jedna byla False!
````

````python
# === Praktický příklad AND ===
# Obě čísla musí být >= minimální hodnota
a = 0
b = 2
minimalni_hodnota = 0

if a >= minimalni_hodnota and b >= minimalni_hodnota:
    print(f"✅ Obě čísla ({a}, {b}) jsou >= {minimalni_hodnota}")
else:
    print(f"❌ Alespoň jedno z čísel ({a}, {b}) je menší než {minimalni_hodnota}")

# 🧪 Zkuste změnit hodnoty a a b a spusťte znovu!
# Např.: a = -1 → co se změní?
````

---

## 2️⃣ Operátor `or` – „Nebo"

Operátor `or` vrací `True`, **pokud je alespoň jedna podmínka pravdivá**. Vrací `False` pouze tehdy, když jsou **obě** podmínky `False`.

### Pravdivostní tabulka

| Podmínka A | Podmínka B | A `or` B |
|:---:|:---:|:---:|
| `True` | `True` | ✅ `True` |
| `True` | `False` | ✅ `True` |
| `False` | `True` | ✅ `True` |
| `False` | `False` | ❌ `False` |

### Příklad z praxe
Nemusíte do práce, pokud je **víkend** *nebo* máte **dovolenou** (stačí jedno):

```python
if je_vikend or ma_dovolenou:
    print("Nemusíš do práce!")
```

### Operátor `in` – „Je obsaženo v..."
V příkladech uvidíte také operátor `in`, který zjišťuje, **zda se prvek nachází v kolekci** (textu, seznamu, …):

```python
"a" in "ahoj"       # True – písmeno "a" je v textu "ahoj"
"z" in "ahoj"       # False – písmeno "z" tam není
3 in [1, 2, 3, 4]   # True – číslo 3 je v seznamu
```

````python
# === Jednoduchá ukázka OR ===
# Jaký bude výsledek? Nejprve si tipněte!
if (7 < 5) or (8 > 7):
    print("Podmínka je splněna")
else:
    print("Podmínka NENÍ splněna")

# Vysvětlení:
# (7 < 5) → False
# (8 > 7) → True
# False or True → True  ← stačí, aby jedna byla True!
````

````python
# === Praktický příklad OR + operátor "in" ===
# Hledáme, zda se alespoň jeden ze znaků nachází ve slově
hledany_znak_1 = "a"
hledany_znak_2 = "w"
slovo = "ahoj"

if hledany_znak_1 in slovo or hledany_znak_2 in slovo:
    print(f"✅ Alespoň jeden ze znaků ('{hledany_znak_1}', '{hledany_znak_2}') je ve slově '{slovo}'")
else:
    print(f"❌ Ani '{hledany_znak_1}', ani '{hledany_znak_2}' se v '{slovo}' nenachází")

# Vysvětlení:
# "a" in "ahoj" → True
# "w" in "ahoj" → False
# True or False → True

# 🧪 Zkuste změnit znaky na "x" a "z" – co se stane?
````

## 3️⃣ Operátor `not` – „Negace (opak)"

Operátor `not` **obrací** logickou hodnotu – z `True` udělá `False` a naopak.

### Pravdivostní tabulka

| Podmínka A | `not` A |
|:---:|:---:|
| `True` | ❌ `False` |
| `False` | ✅ `True` |

### Kdy se hodí?
- Když chcete zkontrolovat, že něco **neplatí**
- Čitelnější alternativa k `!=` v některých případech

```python
# Tyto dva zápisy dělají totéž:
if not je_zavreno:        # přirozenější čtení: "pokud NENÍ zavřeno"
    print("Otevřeno!")

if je_zavreno == False:   # funguje, ale méně elegantní
    print("Otevřeno!")
```

### `not in` – „Není obsaženo v..."
Kombinace `not` a `in` je velmi užitečná pro zjištění, zda se prvek **nenachází** v kolekci:

```python
"z" not in "ahoj"    # True – "z" se v "ahoj" nenachází
"a" not in "ahoj"    # False – "a" tam je
```

````python
# === Jednoduchý příklad s NOT ===
jmeno = input("Zadej své jméno: ")

# not obrací logickou hodnotu
# Pokud jmeno NENÍ "Matěj" → not False → True
if not jmeno == "Matěj":
    print(f"Ahoj {jmeno}, ty nejsi Matěj!")
else:
    print("Ahoj Matěji! 👋")

# 🧪 Krok za krokem:
# Např. jmeno = "Petr"
# jmeno == "Matěj" → False
# not False → True → vstoupíme do if
````

````python
# === Praktický příklad – kontrola záporného čísla ===
cislo = int(input("Zadej číslo: "))

# not (cislo > 0)  →  číslo NENÍ kladné
if not cislo > 0:
    print(f"Číslo {cislo} NENÍ kladné (je záporné nebo nula).")
else:
    print(f"Číslo {cislo} je kladné. ✅")

# 💡 Alternativní zápis (totéž, jinak):
# if cislo <= 0:
````

### Příklad s `not in`
Operátor `not in` je přehledný způsob, jak ověřit **nepřítomnost** prvku:

````python
# === Příklad s not in ===
hledany_znak = input("Zadej znak, který chceš hledat: ")
retezec = input("Zadej text, ve kterém hledat: ")

if hledany_znak not in retezec:
    print(f"Znak '{hledany_znak}' se v textu '{retezec}' NENACHÁZÍ. ❌")
else:
    print(f"Znak '{hledany_znak}' se v textu '{retezec}' nachází. ✅")

# 🧪 Vyzkoušej:
# hledany_znak = "x", retezec = "Python"  →  "x" not in "Python" → True → NENACHÁZÍ
# hledany_znak = "P", retezec = "Python"  →  "P" not in "Python" → False → nachází
````

## 4️⃣ Operátor `==` – „Je rovno"

Operátor `==` porovnává **hodnoty** dvou proměnných. Vrací `True`, pokud jsou hodnoty **stejné**.

> ⚠️ **Pozor na záměnu!**  
> `=` je **přiřazení** (uložení hodnoty do proměnné), `==` je **porovnání** (zjištění, zda jsou hodnoty stejné).

### Co vše lze porovnávat?

| Typ | Příklad | Výsledek |
|---|---|---|
| Čísla | `5 == 5` | `True` |
| Texty (stringy) | `"ahoj" == "ahoj"` | `True` |
| Seznamy (listy) | `[1, 2] == [1, 2]` | `True` |
| Různé typy | `5 == "5"` | `False` ⚠️ |
| Bool a číslo | `True == 1` | `True` 🤔 |

> 💡 Python porovnává **obsah/hodnotu**, ne datový typ (kromě případů jako `5 == "5"`, kde se typy liší).

````python
# === Porovnání čísel ===
a = 2
b = 2

if a == b:
    print(f"✅ Proměnné a={a} a b={b} mají STEJNOU hodnotu")
else:
    print(f"❌ Proměnné a={a} a b={b} mají RŮZNOU hodnotu")

# 🧪 Zkuste změnit b na 3 a spusťte znovu!
````

### Porovnání textů (stringů)
Operátor `==` funguje i na textové řetězce – porovnává **celý obsah** znak po znaku:

> ⚠️ **Pozor na velká/malá písmena!** `"Ahoj" == "ahoj"` → `False`

````python
# === Porovnání textů ===
a = "text"
b = "text"

if a == b:
    print(f"✅ '{a}' a '{b}' jsou stejné texty")
else:
    print(f"❌ '{a}' a '{b}' jsou různé texty")

# 🧪 Zkuste: a = "Text", b = "text" → co se stane?
# Velká a malá písmena se ROZLIŠUJÍ!
````

### Porovnání seznamů (listů)
U seznamů `==` porovnává **každý prvek** na stejné pozici. Seznamy musí mít **stejnou délku i obsah**:

````python
# === Porovnání seznamů ===
a = [1, "2", True]
b = [2, "2", True]

if a == b:
    print(f"✅ Seznamy {a} a {b} mají stejný obsah")
else:
    print(f"❌ Seznamy {a} a {b} mají různý obsah")

# Výsledek: ❌ – liší se první prvek (1 vs 2)
# 🧪 Změňte b na [1, "2", True] a spusťte znovu
````

## 5️⃣ Operátor `is` – „Je to tentýž objekt?"

Operátor `is` je **podobný** operátoru `==`, ale funguje jinak:

| Operátor | Co porovnává | Otázka |
|:---:|---|---|
| `==` | **Hodnotu** (obsah) | „Mají stejný obsah?" |
| `is` | **Identitu** (místo v paměti) | „Je to tentýž objekt v paměti?" |

### Analogie 🏠
- `==` → Dva domy mají **stejný počet pokojů** (stejná hodnota)
- `is` → Dva klíče otevírají **ten samý dům** (stejný objekt v paměti)

### Kdy `==` a `is` dávají různé výsledky?
U **seznamů** (a jiných měnitelných objektů): dva seznamy mohou mít stejný obsah, ale být uloženy **na různých místech v paměti**.

> 💡 U jednoduchých typů (`int`, `str`) Python optimalizuje paměť – stejné hodnoty často sdílí stejné místo. Proto se `is` u čísel a textů chová podobně jako `==`, ale **nespoléhejte na to!**

````python
# === Příklad: b = a (odkaz na stejný objekt) ===
a = [1, "2", True]
b = a  # b nyní ODKAZUJE na stejný objekt jako a

# a i b ukazují na TOTÉŽ místo v paměti
print(f"a == b: {a == b}")   # True – stejná hodnota
print(f"a is b: {a is b}")   # True – stejný objekt!
print(f"id(a): {id(a)}")     # id() vrací adresu v paměti
print(f"id(b): {id(b)}")     # stejná adresa jako a

if a is b:
    print("✅ Proměnné 'a' a 'b' jsou identické (stejný objekt)")
else:
    print("❌ Proměnné 'a' a 'b' nejsou identické")

# 💡 Protože b = a, změna v b změní i a:
b.append(99)
print(f"\nPo přidání 99 do b:")
print(f"a = {a}")  # [1, '2', True, 99] – změnilo se i a!
````

### Dva různé objekty se stejnou hodnotou
Pokud vytvoříme dva seznamy se **stejným obsahem** samostatně, budou na **různých místech** v paměti:

````python
# === Příklad: dva samostatné seznamy ===
a = [1, "2", True]
b = [1, "2", True]  # stejný obsah, ale NOVÝ objekt

print(f"a == b: {a == b}")   # True – stejná hodnota
print(f"a is b: {a is b}")   # False – různé objekty!
print(f"id(a): {id(a)}")     # různé adresy
print(f"id(b): {id(b)}")

if a is b:
    print("✅ Proměnné 'a' a 'b' jsou identické (stejný objekt)")
else:
    print("❌ Proměnné 'a' a 'b' NEJSOU identické (různé objekty, ale stejný obsah)")

# 🧪 Shrnutí:
# == porovnává HODNOTU → True (obsah je stejný)
# is porovnává IDENTITU → False (jsou to dva různé objekty v paměti)
````

## 6️⃣ Kombinace logických operátorů

Logické operátory lze samozřejmě **kombinovat** v jedné podmínce. Je ale důležité znát **priority** (pořadí vyhodnocení):

### Priorita operátorů (od nejvyšší po nejnižší)

| Pořadí | Operátor | Příklad |
|:---:|:---:|---|
| 1. | `not` | `not True` → `False` |
| 2. | `and` | `True and False` → `False` |
| 3. | `or` | `False or True` → `True` |

> ⚠️ **Důležité pravidlo:** `and` má **přednost** před `or`!  
> Výraz `A or B and C` se vyhodnotí jako `A or (B and C)`, nikoli jako `(A or B) and C`.

### Tip: **Používejte závorky!** 🎯
I když znáte priority, závorky dělají kód **čitelnějším** pro vás i ostatní:

```python
# Bez závorek – méně přehledné:
if vek >= 18 and ma_ridicak or je_vyjimka:

# Se závorkami – jasné, co se děje:
if (vek >= 18 and ma_ridicak) or je_vyjimka:
```

````python
# === Příklad kombinace operátorů ===
a = 1
b = 2
soucet = a + b  # = 3

# Složená podmínka se závorkami pro přehlednost
# Vyhodnocení: nejdříve and, pak or
if soucet == 3 and (soucet > 0 or soucet == 3) and soucet > 5:
    print("Podmínka je splněna")
else:
    print("Podmínka NENÍ splněna")

# 🧪 Krok za krokem:
# soucet == 3      → True
# soucet > 0       → True
# soucet == 3      → True
# (True or True)   → True
# soucet > 5       → False
# True and True and False → False  ← stačí jedno False u and!
````

## 7️⃣ Uložení podmínky do proměnné

Podmínky vrací `True` nebo `False` – tedy datový typ **`bool`**. Můžeme si je proto **uložit do proměnné** a použít později:

```python
je_dospely = vek >= 18          # True nebo False
ma_pristup = je_dospely and ma_heslo

if ma_pristup:
    print("Vstup povolen!")
```

### Proč to dělat?
- **Přehlednost** – pojmenovaná podmínka se lépe čte
- **Znovupoužitelnost** – nemusíme psát stejnou podmínku vícekrát
- **Ladění** – můžeme si vypsat hodnotu proměnné a zjistit, proč podmínka (ne)platí

````python
# === Uložení podmínky do proměnné ===

# Podmínka vrací bool – můžeme si ověřit:
vysledek = 1 + 2 == 3 or 4 > 2
print(f"Typ podmínky: {type(vysledek)}")  # <class 'bool'>
print(f"Hodnota: {vysledek}")              # True

# Uložíme podmínku do proměnné s popisným názvem
podminka = 1 + 2 == 3 or 4 > 2

# A použijeme ji v if
if podminka:
    print(f"✅ Podmínka je splněna (hodnota: {podminka})")
else:
    print(f"❌ Podmínka není splněna (hodnota: {podminka})")

# 💡 Praktičtější příklad:
vek = 20
ma_ridicak = True
muze_ridit = vek >= 18 and ma_ridicak

print(f"\nVěk: {vek}, má řidičák: {ma_ridicak}")
print(f"Může řídit: {muze_ridit}")
````

## 📋 Shrnutí

| Operátor | Význam | Příklad | Výsledek |
|:---:|---|---|:---:|
| `and` | A zároveň | `True and False` | `False` |
| `or` | Nebo | `False or True` | `True` |
| `not` | Negace | `not True` | `False` |
| `in` | Je obsaženo v | `"a" in "ahoj"` | `True` |
| `not in` | Není obsaženo v | `"z" not in "ahoj"` | `True` |
| `==` | Stejná hodnota | `[1,2] == [1,2]` | `True` |
| `!=` | Různá hodnota | `5 != 3` | `True` |
| `is` | Stejný objekt | `a is b` (po `b = a`) | `True` |

### Priority (od nejvyšší)
`not` → `and` → `or`

> 🎯 **Tip na závěr:** Pokud si nejste jisti prioritami, **vždy používejte závorky**. Kód bude přehlednější a vyhnete se záludným chybám!
