## 2. Quick Sort (Rychlé třídění)

**Princip:**
1. Vyber **pivot** (např. poslední prvek)

3. Rekurzivně seřaď obě části

```

```

## 4. Vyhledávání

Napište algoritmy pro **lineární** a **binární** vyhledávání.

### A) Lineární vyhledávání
Prochází pole od začátku do konce, dokud nenajde hledaný prvek.
- Funguje na **neseřazeném** i seřazeném poli
- Časová složitost: **O(n)**

### B) Binární vyhledávání
Pole musí být **seřazeno**. Porovnáváme hledaný prvek se středním prvkem:
- Je-li hledaný prvek menší → prohledáváme levou polovinu
- Je-li větší → prohledáváme pravou polovinu
- Opakujeme, dokud prvek nenajdeme (nebo nezjistíme, že není v poli)
- Časová složitost: **O(log n)**

**Příklad:**
```
Pole:     [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]
Hledám:   23
Výstup:   Index 5
```

## 6. Fronta (Queue)

Implementujte datovou strukturu **fronta** (FIFO – First In, First Out).

**Princip:** První přidaný prvek je první odebraný (jako fronta u pokladny).

```

vypis()     → [10, 20, 30]
```

## 7. Zásobník (Stack)

Implementujte datovou strukturu **zásobník** (LIFO – Last In, First Out).

**Princip:** Poslední přidaný prvek je první odebraný (jako zásobník talířů).

**Požadované operace:**
1. `push(hodnota)` – vloží prvek na vrchol zásobníku
2. `pop()` – odebere a vrátí prvek z vrcholu
3. `peek()` – vrátí vrchní prvek bez odebrání
4. `je_prazdny()` – vrátí `True`, pokud je zásobník prázdný
5. `vypis()` – vypíše obsah zásobníku

**Příklad:**
```
push(1) → push(2) → push(3)
vypis() → [1, 2, 3]  (vrchol = 3)
pop()   → 3
vypis() → [1, 2]
```
