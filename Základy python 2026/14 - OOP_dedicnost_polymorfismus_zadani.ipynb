## ğŸ“ Syntaxe dÄ›diÄnosti


### Co potomek **automaticky zÃ­skÃ¡** od rodiÄe:
- âœ… VÅ¡echny **metody** (pokud je nepÅ™epÃ­Å¡e)
- âœ… VÅ¡echny **atributy** nastavenÃ© v `__init__` rodiÄe (pokud zavolÃ¡ `super().__init__()`)
- âœ… Metodu `__str__` (pokud ji nepÅ™epÃ­Å¡e)

````python
# === Syntaxe dÄ›diÄnosti ===
class Rodic:
    def __init__(self, atribut):
        self.atribut = atribut

    def metoda(self):
        print(f"RodiÄovskÃ¡ metoda, atribut = {self.atribut}")


class Potomek(Rodic):                     # â† zÃ¡vorky = dÄ›dÃ­ od Rodic
    def __init__(self, atribut, novy):
        super().__init__(atribut)          # â† zavolÃ¡ __init__ rodiÄe
        self.novy = novy

# === Test ===
p = Potomek("ahoj", 42)
print(p.atribut)     # ahoj   â† zdÄ›dÄ›nÃ½ atribut
print(p.novy)        # 42     â† vlastnÃ­ atribut
p.metoda()           # RodiÄovskÃ¡ metoda, atribut = ahoj  â† zdÄ›dÄ›nÃ¡ metoda
````

## ğŸ¾ PÅ™Ã­klad â€“ RodiÄovskÃ¡ tÅ™Ã­da `Zvire`

ZaÄneme rodiÄovskou tÅ™Ã­dou, kterÃ¡ obsahuje **spoleÄnÃ© vlastnosti** vÅ¡ech zvÃ­Å™at:

````python
# === RodiÄovskÃ¡ tÅ™Ã­da Zvire ===
class Zvire:
    def __init__(self, jmeno, hmotnost):
        if hmotnost < 0:
            raise ValueError("Hmotnost nemÅ¯Å¾e bÃ½t zÃ¡pornÃ¡")
        self.jmeno = jmeno
        self.hmotnost = hmotnost

    def zarvi(self):
        print("HuÃ¡Ã¡Ã¡!")

    def __str__(self):
        return f"{self.jmeno} ({self.hmotnost} kg)"

# === Test ===
z = Zvire("AniÄka", 30)
print(z)           # AniÄka (30 kg)
z.zarvi()          # HuÃ¡Ã¡Ã¡!
````

## ğŸ• Potomci â€“ `Pes` a `Kocka`

Pes i koÄka jsou zvÃ­Å™ata, ale kaÅ¾dÃ½ **Å™ve jinak** a mÃ¡ **vlastnÃ­ atribut navÃ­c**:

````python
# === Potomci ===
class Pes(Zvire):
    def __init__(self, jmeno, hmotnost, hlasitost):
        super().__init__(jmeno, hmotnost)   # zavolÃ¡ Zvire.__init__
        self.hlasitost = hlasitost          # novÃ½ atribut

    def zarvi(self):                        # pÅ™epsÃ¡nÃ­ (override)
        print(f"Haf haf! (hlasitost {self.hlasitost})")

    def __str__(self):
        return f"Pes {super().__str__()}, hlasitost {self.hlasitost}"


class Kocka(Zvire):
    def __init__(self, jmeno, hmotnost, roztomilost):
        super().__init__(jmeno, hmotnost)
        self.roztomilost = roztomilost

    def zarvi(self):
        print(f"MÅˆau! (roztomilost {self.roztomilost})")

    def __str__(self):
        return f"KoÄka {super().__str__()}, roztomilost {self.roztomilost}"


# === Test ===
p = Pes("BÃ¡ra", 40, 90)
k = Kocka("Cecilka", 3, 100)

print(p)            # Pes BÃ¡ra (40 kg), hlasitost 90
p.zarvi()           # Haf haf! (hlasitost 90)

print(k)            # KoÄka Cecilka (3 kg), roztomilost 100
k.zarvi()           # MÅˆau! (roztomilost 100)
````

## ğŸ”‘ `super()` â€“ volÃ¡nÃ­ rodiÄovskÃ© tÅ™Ã­dy

`super()` vracÃ­ odkaz na **rodiÄovskou tÅ™Ã­du**. NejÄastÄ›ji ho pouÅ¾Ã­vÃ¡me v `__init__()`.

### Dva zpÅ¯soby zÃ¡pisu:

| ZpÅ¯sob | ZÃ¡pis | DoporuÄeno |
|:-------|:------|:-----------|
| **`super()`** | `super().__init__(jmeno, hmotnost)` | âœ… Ano |
| **PÅ™Ã­mÃ© volÃ¡nÃ­** | `Zvire.__init__(self, jmeno, hmotnost)` | âš ï¸ Funguje, ale mÃ©nÄ› flexibilnÃ­ |

> ğŸ¯ **PouÅ¾Ã­vejte `super()`** â€“ je kratÅ¡Ã­, ÄitelnÄ›jÅ¡Ã­ a sprÃ¡vnÄ› funguje i pÅ™i vÃ­ceÃºrovÅˆovÃ© dÄ›diÄnosti.

````python
# === super() v metodÄ› __str__ ===
class Pes(Zvire):
    def __init__(self, jmeno, hmotnost, hlasitost):
        super().__init__(jmeno, hmotnost)
        self.hlasitost = hlasitost

    def __str__(self):
        # super().__str__() vrÃ¡tÃ­ "BÃ¡ra (40 kg)" od rodiÄe
        zaklad = super().__str__()
        return f"Pes {zaklad}, hlasitost {self.hlasitost}"

# === Test ===
p = Pes("BÃ¡ra", 40, 90)
print(p)             # Pes BÃ¡ra (40 kg), hlasitost 90
print(super(Pes, p).__str__())   # BÃ¡ra (40 kg)  â† jen rodiÄovskÃ½ __str__
````

ExistujÃ­ dva pÅ™Ã­stupy:

````python
# === 2. RozÅ¡Ã­Å™enÃ­ â€“ rodiÄ + vlastnÃ­ kÃ³d ===
class PesRozsireny(Zvire):
    def __init__(self, jmeno, hmotnost):
        super().__init__(jmeno, hmotnost)

    def zarvi(self):
        super().zarvi()            # nejprve rodiÄovskÃ©: "HuÃ¡Ã¡Ã¡!"
        print("...a pak Haf!")     # potom vlastnÃ­

p2 = PesRozsireny("Aria", 30)
p2.zarvi()
# HuÃ¡Ã¡Ã¡!
# ...a pak Haf!
````

> ğŸ’¡ **RozÅ¡Ã­Å™enÃ­** je uÅ¾iteÄnÃ©, kdyÅ¾ chceme zachovat chovÃ¡nÃ­ rodiÄe a jen k nÄ›mu nÄ›co pÅ™idat.

V naÅ¡em pÅ™Ã­kladu: metoda `zarvi()` existuje u `Zvire`, `Pes` i `Kocka` â€“ ale kaÅ¾dÃ¡ verze dÄ›lÃ¡ nÄ›co jinÃ©ho:
