# 🖥️ PyQt5 – Lekce 1: Okno, nápis a tlačítko

V této lekci si poprvé spustíme vlastní **grafickou aplikaci** v Pythonu.
Na rozdíl od konzolové aplikace (která skončí hned po provedení posledního řádku),
GUI aplikace běží neustále a čeká na akce uživatele – říká se tomu **událostní smyčka (Event loop)**.

---

## ⚙️ Jak funguje PyQt5 aplikace?

```
START skriptu
    │
    ▼
Vytvoříme QApplication ─── správce celé aplikace, musí být PRVNÍ
    │
    ▼
Vytvoříme QWidget ──────── prázdné okno
    │
    ▼
Přidáme prvky (QLabel, QPushButton, …) do Layout managera
    │
    ▼
okno.show() ───────────── okno se zobrazí na obrazovce
    │
    ▼
app.exec() ────────────── spustí událostní smyčku (čeká na uživatele)
    │                          ↑
    ▼                          │ každá akce uživatele (klik, psaní…)
  KONEC ◄──── zavření okna ────┘
```

---

## 📦 Přehled základních objektů

| Objekt | Co dělá |
|---|---|
| `QApplication([])` | Správce aplikace. Vytváříme ho **vždy jako první**. |
| `QWidget()` | Prázdné okno. Základ pro vše ostatní. |
| `QLabel("text")` | Zobrazuje text (nebo obrázek). Uživatel ho nemůže editovat. |
| `QPushButton("text")` | Klikatelné tlačítko. |
| `QHBoxLayout()` | Seřadí vložené prvky **vedle sebe** (horizontálně). |

## 🟢 Ukázka 1 – Nejjednodušší možné okno

Vytvoříme pouze prázdné okno. Žádné prvky uvnitř.

````python
from PyQt5 import QtWidgets

app = QtWidgets.QApplication([])        # 1. Správce aplikace (VŽDY PRVNÍ)

okno = QtWidgets.QWidget()              # 2. Prázdné okno
okno.setWindowTitle("Moje první okno") # 3. Titulek v záhlaví
okno.resize(400, 300)                   # 4. Šířka x Výška (v pixelech)

okno.show()                             # 5. Zobrazit okno
app.exec()                              # 6. Spustit událostní smyčku
````

## 🟢 Ukázka 2 – Okno s textem (`QLabel`)

Aby bylo možné přidávat prvky do okna, musíme nejprve nastavit **Layout Manager** –
správce rozvržení, který se postará o správné umístění prvků i při zvětšování okna.

````python
from PyQt5 import QtWidgets, QtCore

app = QtWidgets.QApplication([])
okno = QtWidgets.QWidget()
okno.setWindowTitle("Okno s textem")
okno.resize(350, 150)

# Layout manager – prvky budou vedle sebe (H = Horizontal)
layout = QtWidgets.QHBoxLayout()
okno.setLayout(layout)

napis = QtWidgets.QLabel("Vítejte v PyQt5!")
napis.setAlignment(QtCore.Qt.AlignCenter)  # Zarovnání textu na střed prvku
layout.addWidget(napis)                    # Přidání do layoutu

okno.show()
app.exec()
````

## 🟢 Ukázka 3 – Okno s textem a tlačítkem

Přidáme `QPushButton`. Zatím nic nedělá – pouze ho zobrazíme.
Věnovat se funkcím u tlačítka budeme v příští lekci.

````python
from PyQt5 import QtWidgets

app = QtWidgets.QApplication([])
okno = QtWidgets.QWidget()
okno.setWindowTitle("Text a tlačítko")
okno.resize(400, 120)

layout = QtWidgets.QHBoxLayout()
okno.setLayout(layout)

napis = QtWidgets.QLabel("Nějaký text...")
tlacitko = QtWidgets.QPushButton("Klikni na mě")

layout.addWidget(napis)     # Prvky se přidávají ZLEVA DOPRAVA
layout.addWidget(tlacitko)

okno.show()
app.exec()
````

## 🟡 Ukázka 4 – Více prvků a nastavení velikosti okna

Okno lze libovolně pojmenovat a změnit jeho minimální velikost.

````python
from PyQt5 import QtWidgets

app = QtWidgets.QApplication([])
okno = QtWidgets.QWidget()
okno.setWindowTitle("Ukázka více prvků")
okno.resize(500, 80)
okno.setMinimumSize(300, 60)   # Uživatel okno nezmenší pod tuto mez

layout = QtWidgets.QHBoxLayout()
okno.setLayout(layout)

layout.addWidget(QtWidgets.QLabel("Jméno:"))
layout.addWidget(QtWidgets.QLabel("Tomáš Novák"))
layout.addWidget(QtWidgets.QPushButton("Upravit"))
layout.addWidget(QtWidgets.QPushButton("Smazat"))

okno.show()
app.exec()
````
