## ⚙️ Jak funguje stylování v PyQt?

PyQt5 používá systém nazvaný **Qt Style Sheets (QSS)** – syntaxe je téměř totožná s webovým CSS.
Styl nastavíme metodou `.setStyleSheet("CSS pravidla")`.

```python
widget.setStyleSheet("background-color: #2b2b2b; color: white;")
```

---

## 📦 Nejužitečnější CSS vlastnosti v QSS

| Vlastnost | Příklad hodnoty | Efekt |
|---|---|---|
| `background-color` | `#2b2b2b`, `red`, `rgb(0,0,255)` | Barva pozadí |
| `color` | `white`, `#4CAF50` | Barva textu |
| `font-size` | `16px` | Velikost písma |
| `font-weight` | `bold` | Tučné písmo |
| `border-radius` | `8px` | Zaoblené rohy |
| `padding` | `10px` | Vnitřní odsazení |
| `border` | `2px solid red` | Rámeček |

> 💡 Stav `:hover` funguje také: `QPushButton:hover { background: blue; }`

## 🟢 Ukázka 1 – Základní barevné stylování

````python
from PyQt5.QtWidgets import QApplication, QWidget, QVBoxLayout, QPushButton, QLabel

app = QApplication([])
okno = QWidget()
okno.setWindowTitle("Stylování")
okno.resize(350, 200)

# Styl celého okna
okno.setStyleSheet("background-color: #1e1e2e; color: white;")

layout = QVBoxLayout()
okno.setLayout(layout)

nadpis = QLabel("Tmavý režim 🌙")
nadpis.setStyleSheet("font-size: 22px; font-weight: bold; color: #cba6f7; padding: 8px;")
layout.addWidget(nadpis)

btn_primarni = QPushButton("Primární tlačítko")
btn_primarni.setStyleSheet('''
    QPushButton {
        background-color: #89b4fa;
        color: #1e1e2e;
        font-size: 14px;
        font-weight: bold;
        border-radius: 8px;
        padding: 10px 20px;
    }
    QPushButton:hover { background-color: #74c7ec; }
    QPushButton:pressed { background-color: #89dceb; }
''')

btn_sekundarni = QPushButton("Sekundární tlačítko")
btn_sekundarni.setStyleSheet('''
    QPushButton {
        background-color: transparent;
        color: #89b4fa;
        border: 2px solid #89b4fa;
        border-radius: 8px;
        padding: 8px 18px;
    }
    QPushButton:hover { background-color: rgba(137,180,250,0.1); }
''')

layout.addWidget(btn_primarni)
layout.addWidget(btn_sekundarni)

okno.show()
app.exec()
````

## 🟡 Ukázka 2 – Stylování různých widgetů

````python
from PyQt5.QtWidgets import QApplication, QWidget, QVBoxLayout, QLineEdit, QComboBox, QLabel

app = QApplication([])
okno = QWidget()
okno.setWindowTitle("Stylování widgetů")
okno.resize(320, 220)
okno.setStyleSheet("background-color: #f5f5f5;")

layout = QVBoxLayout()
layout.setContentsMargins(20, 20, 20, 20)
layout.setSpacing(10)
okno.setLayout(layout)

styl_label = "font-size: 12px; font-weight: bold; color: #555;"
styl_input = '''
    QLineEdit {
        border: 2px solid #ddd;
        border-radius: 6px;
        padding: 6px;
        font-size: 14px;
        background: white;
    }
    QLineEdit:focus { border-color: #4A90E2; }
'''

layout.addWidget(QLabel("Jméno:"))
vstup = QLineEdit()
vstup.setStyleSheet(styl_input)
vstup.setPlaceholderText("Zadejte jméno…")
layout.addWidget(vstup)

layout.addWidget(QLabel("Jazyk:"))
combo = QComboBox()
combo.addItems(["Python", "Java", "C++"])
combo.setStyleSheet('''
    QComboBox {
        border: 2px solid #ddd;
        border-radius: 6px;
        padding: 5px;
        font-size: 14px;
        background: white;
    }
''')
layout.addWidget(combo)

okno.show()
app.exec()
````

## 🟡 Ukázka 3 – Obrázky pomocí `QPixmap`

Obrázek se zobrazí vložením do `QLabel` přes `QPixmap`.

````python
from PyQt5.QtWidgets import QApplication, QWidget, QVBoxLayout, QLabel
from PyQt5.QtGui import QPixmap
from PyQt5.QtCore import Qt

app = QApplication([])
okno = QWidget()
okno.setWindowTitle("Zobrazení obrázku")
okno.resize(400, 350)

layout = QVBoxLayout()
okno.setLayout(layout)

obrazek_label = QLabel()
pixmap = QPixmap("obrazek.png")   # Cesta k vašemu souboru

if not pixmap.isNull():
    # Přizpůsobíme velikost obrázku, zachováme poměr stran
    pixmap = pixmap.scaled(380, 280, Qt.KeepAspectRatio, Qt.SmoothTransformation)
    obrazek_label.setPixmap(pixmap)
    obrazek_label.setAlignment(Qt.AlignCenter)
else:
    obrazek_label.setText("📷 Soubor 'obrazek.png' nebyl nalezen.
Zkopírujte obrázek do stejné složky jako tento skript.")
    obrazek_label.setAlignment(Qt.AlignCenter)
    obrazek_label.setStyleSheet("color: gray; font-size: 13px;")

layout.addWidget(obrazek_label)

okno.show()
app.exec()
````

## 🔴 Ukázka 4 – Téma aplikace (světlé/tmavé)

````python
from PyQt5.QtWidgets import QApplication, QWidget, QVBoxLayout, QHBoxLayout, QPushButton, QLabel, QLineEdit

TMAVY = {
    "okno": "background-color: #1e1e2e; color: #cdd6f4;",
    "tlacitko": "QPushButton { background: #313244; color: #cdd6f4; border-radius: 6px; padding: 8px; } QPushButton:hover { background: #45475a; }",
    "vstup": "QLineEdit { background: #313244; color: #cdd6f4; border: 1px solid #45475a; border-radius: 5px; padding: 5px; }",
}

SVETLY = {
    "okno": "background-color: #eff1f5; color: #4c4f69;",
    "tlacitko": "QPushButton { background: #dce0e8; color: #4c4f69; border-radius: 6px; padding: 8px; } QPushButton:hover { background: #ccd0da; }",
    "vstup": "QLineEdit { background: white; color: #4c4f69; border: 1px solid #ccd0da; border-radius: 5px; padding: 5px; }",
}

app = QApplication([])
okno = QWidget()
okno.setWindowTitle("Přepínání tématu")
okno.resize(360, 200)

layout = QVBoxLayout()
okno.setLayout(layout)

nadpis = QLabel("Vítejte!")
vstup = QLineEdit()
vstup.setPlaceholderText("Zadejte cokoliv…")
btn_prepnout = QPushButton("Přepnout téma 🌙/☀️")
layout.addWidget(nadpis)
layout.addWidget(vstup)
layout.addWidget(btn_prepnout)

tmavy_rezim = True

def prepnout():
    global tmavy_rezim
    tematica = TMAVY if tmavy_rezim else SVETLY
    okno.setStyleSheet(tematica["okno"])
    btn_prepnout.setStyleSheet(tematica["tlacitko"])
    vstup.setStyleSheet(tematica["vstup"])
    tmavy_rezim = not tmavy_rezim

prepnout()   # Nastavit výchozí téma hned při startu
btn_prepnout.clicked.connect(prepnout)

okno.show()
app.exec()
````
