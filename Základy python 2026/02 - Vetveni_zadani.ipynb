# 🌿 Větvení (podmínky)

## Co je to větvení?

Dosud naše programy běžely **lineárně** – řádek po řádku, odshora dolů, bez výjimky. To je ale pro reálné programy nedostatečné. Často potřebujeme, aby program **reagoval na různé situace různě**.

**Větvení** (anglicky *branching* nebo *conditional statements*) je **řídící struktura**, která umožňuje programu **rozhodovat se** – vykonat různé příkazy podle toho, zda je splněna určitá podmínka.

### Analogie ze života

Představte si křižovatku: podle směru, který zvolíte, se dostanete na jiné místo. Podobně program na základě **podmínky** zvolí jednu z „větví" kódu.

```
         ┌─ podmínka splněna → příkazy A
Program ─┤
         └─ podmínka nesplněna → příkazy B
```

### Tři základní typy větvení v Pythonu

| Struktura | Kdy použít |
|---|---|
| `if` | Chceme vykonat příkazy **pouze pokud** je podmínka splněna |
| `if-else` | Chceme vybrat **jednu ze dvou** možností |
| `if-elif-else` | Chceme vybrat **jednu z více** možností |

## 🔑 Podmínka

Než se pustíme do samotného větvení, je důležité rozumět tomu, co je **podmínka**.

**Podmínka** je jakýkoliv výraz, který Python dokáže vyhodnotit jako **`True`** (pravda) nebo **`False`** (nepravda). Nejčastěji používáme:

| Operátor | Význam | Příklad | Výsledek |
|:---:|---|---|---|
| `==` | Rovná se | `5 == 5` | `True` |
| `!=` | Nerovná se | `5 != 3` | `True` |
| `<` | Menší než | `3 < 5` | `True` |
| `>` | Větší než | `3 > 5` | `False` |
| `<=` | Menší nebo rovno | `5 <= 5` | `True` |
| `>=` | Větší nebo rovno | `3 >= 5` | `False` |

> 💡 Výsledkem podmínky je vždy **`bool`** – tedy `True` nebo `False`. Právě tuto hodnotu Python používá k rozhodnutí, jakou větví se vydat.

# 1️⃣ Podmínka `if` – „pokud platí, udělej"

Nejjednodušší forma větvení. Příkazy v **těle** `if` se provedou **pouze tehdy**, když je podmínka splněna (`True`). Pokud podmínka splněna není (`False`), příkazy se **přeskočí** a program pokračuje dál.

### Syntax (zápis)

```python
if podmínka:
    příkaz_1        # ← tělo if (odsazené o 4 mezery!)
    příkaz_2        # ← stále v těle if

příkaz_mimo_if      # ← tento se provede VŽDY (není odsazený)
```

### ⚠️ Klíčová pravidla

1. **Za podmínkou musí být dvojtečka `:`** – pokud ji zapomenete, dostanete `SyntaxError`
2. **Tělo musí být odsazené** – standardně o **4 mezery** (nebo 1 tabulátor). Python pozná, co patří do `if`, právě podle odsazení!
3. **Příkazy bez odsazení** už do `if` nepatří a provedou se vždy

> 🎯 **Odsazení (indentace)** je v Pythonu zásadní! Na rozdíl od jiných jazyků (C, Java), kde se bloky kódu ohraničují závorkami `{...}`, Python používá **odsazení**. Špatné odsazení = nefunkční program.

### Vizualizace

```
podmínka == True?
    ├── ANO → vykonej tělo if
    └── NE  → přeskoč tělo if
pokračuj dál...
```

````python
# === Ukázka if ===
obdobi = "léto"

if obdobi == "zima":
    print("Máma ti vzkazuje: Vezmi si kulicha!")

# Tento print se provede VŽDY – není v těle if (není odsazený)
print("Hezký den!")  

# Zkuste změnit proměnnou obdobi na "zima" a spustit znovu – uvidíte rozdíl
````

# 2️⃣ Podmínka `if-else` – „buď ... anebo"

Často nechceme jen „něco udělat, pokud platí podmínka", ale chceme **vybrat jednu ze dvou možností**. K tomu slouží `if-else`.

- **`if` větev** – provede se, když je podmínka `True`
- **`else` větev** – provede se, když je podmínka `False`

**Vždy se provede přesně jedna z těchto dvou větví** – nikdy obě, nikdy žádná.

### Syntax

```python
if podmínka:
    příkazy_když_True       # větev A
else:
    příkazy_když_False      # větev B
```

### Vizualizace

```
podmínka == True?
    ├── ANO → větev A (tělo if)
    └── NE  → větev B (tělo else)
pokračuj dál...
```

> ⚠️ Za `else` se **nepíše žádná podmínka** – jednoduše `else:` (s dvojtečkou). Else zachytí **všechno ostatní**.

````python
# === Ukázka if-else ===
obdobi = "léto"

if obdobi == "zima":
    print("Máma ti vzkazuje: Vezmi si kulicha!")
else:
    print("Máma ti vzkazuje: Stejně si vezmi kulicha!")

# Praktičtější příklad:
print()
vek = int(input("Kolik ti je let? "))

if vek >= 18:
    print("Jsi dospělý – můžeš volit! 🗳️")
else:
    print("Jsi nezletilý – ještě chvíli počkej 😊")
````

> 🤔 **Kontrolní otázka:** V prvním příkladu – co máma vzkazuje v létě? A co v zimě? Zkuste si to nejprve odvodit z kódu, pak spusťte buňku a ověřte.

# 3️⃣ Podmínka `if-elif-else` – „výběr z více možností"

Někdy potřebujeme vybírat ne mezi dvěma, ale mezi **více možnostmi**. K tomu slouží klíčové slovo **`elif`** (zkratka od „**else if**" = „jinak pokud").

### Jak to funguje?

1. Python zkontroluje podmínku u `if`
2. Pokud je `False`, zkontroluje podmínku u prvního `elif`
3. Pokud je opět `False`, zkontroluje další `elif` (a tak dále)
4. Pokud **žádná** podmínka není `True`, provede se `else` (pokud existuje)

**Klíčová vlastnost:** Provede se **pouze první větev**, jejíž podmínka je splněna! Jakmile Python najde `True`, zbylé `elif`/`else` přeskočí.

### Syntax

```python
if podmínka_1:
    příkazy_1           # provede se, když podmínka_1 je True
elif podmínka_2:
    příkazy_2           # provede se, když podmínka_2 je True (a podmínka_1 je False)
elif podmínka_3:
    příkazy_3           # provede se, když podmínka_3 je True (a předchozí jsou False)
else:
    příkazy_jinak       # provede se, když ŽÁDNÁ podmínka výše nebyla True
```

### Důležité poznámky

- `elif` větví může být **libovolný počet**
- `else` je **nepovinný** – ale pokud ho vynecháte, může se stát, že se neprovede žádná větev
- Pořadí podmínek **záleží** – kontrolují se odshora dolů

### Vizualizace

```
podmínka_1 == True? ── ANO → větev 1
    │ NE
podmínka_2 == True? ── ANO → větev 2
    │ NE
podmínka_3 == True? ── ANO → větev 3
    │ NE
    └── else větev (záchytná síť)
```

````python
# === Ukázka if-elif-else ===
obdobi = "léto"

if obdobi == "zima":
    print("Máma ti vzkazuje: Vezmi si kulicha! 🧣")
elif obdobi == "jaro":
    print("Máma ti vzkazuje: Vezmi si pořádný boty! 👢")
elif obdobi == "léto":
    print("Máma ti vzkazuje: Vezmi si opalovací krém! ☀️")
elif obdobi == "podzim":
    print("Máma ti vzkazuje: Vezmi si šálu! 🍂")
else:
    print("Máma ti vzkazuje: Neznám toto období, ale vezmi si svačinu! 🥪")

# Zkuste měnit hodnotu proměnné "obdobi" a sledujte, jak se mění výstup
````

### ⚠️ Pozor: `if-elif` vs. více samostatných `if`

Toto je **častá chyba začátečníků**! Podívejte se na rozdíl:

```python
# Varianta A: if-elif – provede se jen JEDNA větev
x = 10
if x > 5:
    print("Větší než 5")     # ✅ provede se
elif x > 3:
    print("Větší než 3")     # ❌ přeskočí (už se provedla první větev!)

# Varianta B: dva samostatné if – provede se KAŽDÝ splněný
x = 10
if x > 5:
    print("Větší než 5")     # ✅ provede se
if x > 3:
    print("Větší než 3")     # ✅ také se provede (je to samostatný if!)
```

> 💡 Používejte `elif`, když chcete **vzájemně se vylučující možnosti** (jako roční období). Používejte více `if`, když chcete kontrolovat **nezávislé podmínky**.

# 🔗 Logické operátory – kombinování podmínek

Často potřebujeme zkontrolovat **více podmínek najednou**. K tomu slouží **logické operátory**:

| Operátor | Význam | Příklad | Výsledek |
|:---:|---|---|---|
| `and` | **A zároveň** – obě musí být True | `5 > 3 and 2 < 10` | `True` |
| `or` | **Nebo** – alespoň jedna musí být True | `5 > 3 or 2 > 10` | `True` |
| `not` | **Negace** – obrátí True ↔ False | `not True` | `False` |

### Pravdivostní tabulky

**`and`** – obě podmínky musí platit:
| A | B | A `and` B |
|:---:|:---:|:---:|
| True | True | **True** |
| True | False | False |
| False | True | False |
| False | False | False |

**`or`** – stačí jedna:
| A | B | A `or` B |
|:---:|:---:|:---:|
| True | True | True |
| True | False | **True** |
| False | True | **True** |
| False | False | False |

**`not`** – obrátí hodnotu:
| A | `not` A |
|:---:|:---:|
| True | False |
| False | True |

### Příklady z praxe

```python
vek = 20
ma_ridicak = True

# Může řídit, pokud je dospělý A má řidičák
if vek >= 18 and ma_ridicak:
    print("Můžeš řídit!")

# Je víkend, nebo má dovolenou → nemusí do práce
je_vikend = True
ma_dovolenou = False
if je_vikend or ma_dovolenou:
    print("Nemusíš do práce!")

# Otevřený obchod → pokud NENÍ zavřeno
je_zavreno = False
if not je_zavreno:
    print("Obchod je otevřený!")
```

````python
# === Ukázka logických operátorů ===
vek = 20
ma_ridicak = True

# and – obě podmínky musí platit
if vek >= 18 and ma_ridicak:
    print("✅ Můžeš řídit!")
else:
    print("❌ Nemůžeš řídit.")

# or – stačí jedna podmínka
print()
teplota = 35
prsi = False
if teplota > 30 or prsi:
    print("🏠 Zůstaň doma – je příliš horko nebo prší.")
else:
    print("🚶 Hezké počasí, můžeš ven!")

# not – negace
print()
je_zavreno = False
if not je_zavreno:
    print("🛒 Obchod je otevřený!")

# Kombinace operátorů
print()
cislo = 15
if cislo > 0 and cislo % 2 == 0:
    print(f"{cislo} je kladné sudé číslo")
elif cislo > 0 and cislo % 2 != 0:
    print(f"{cislo} je kladné liché číslo")
else:
    print(f"{cislo} není kladné číslo")
````

# 🪆 Vnořené podmínky (nesting)

Podmínky můžeme **vnořovat** (vkládat jednu do druhé). Vnořená podmínka se kontroluje **pouze tehdy**, když je nadřazená podmínka splněna.

### Syntax

```python
if podmínka_1:
    příkazy_1
    if podmínka_2:          # vnořená podmínka – kontroluje se jen když podmínka_1 je True
        příkazy_vnořené
    else:
        jiné_příkazy
else:
    příkazy_když_False
```

### Kdy vnořovat?

- Když potřebujeme **zpřesnit rozhodování** v rámci jedné větve
- Např. „Je zima? → Pokud ano, jaká je teplota?"

### Kdy raději nevnořovat?

- Příliš hluboké zanoření (3+ úrovně) dělá kód **nepřehledným**
- Často lze nahradit **logickými operátory** (`and`, `or`):

```python
# Vnořené – méně přehledné:
if obdobi == "zima":
    if teplota < 0:
        print("Mrzne!")

# Ekvivalentní s and – přehlednější:
if obdobi == "zima" and teplota < 0:
    print("Mrzne!")
```

> 💡 **Pravidlo palce:** Pokud je vnořená podmínka jediná věc v těle `if`, zvažte spojení pomocí `and`.

````python
# === Ukázka vnořených podmínek ===
obdobi = "zima"
teplota = -5

if obdobi == "zima":
    print("Máma ti vzkazuje: Vezmi si kulicha! 🧣")
    
    # Vnořená podmínka – upřesňuje situaci v zimě
    if teplota < -10:
        print("🥶 Pozor! Venku je krutá zima!")
    elif teplota < 0:
        print("❄️ Mrzne! Oblékni se pořádně.")
    else:
        print("☁️ Zima, ale nad nulou – snesitelné.")

elif obdobi == "jaro":
    print("Máma ti vzkazuje: Vezmi si pořádný boty! 👢")
elif obdobi == "léto":
    print("Máma ti vzkazuje: Vezmi si opalovací krém! ☀️")
elif obdobi == "podzim":
    print("Máma ti vzkazuje: Vezmi si šálu! 🍂")
else:
    print("❓ Neznámé období roku!")

# Zkuste měnit hodnoty "obdobi" a "teplota" a sledujte výstup
````

# ⚠️ Časté chyby a tipy

### 1. Zapomenutá dvojtečka
```python
# ❌ Chybí dvojtečka za podmínkou:
if x > 5
    print("Ahoj")    # SyntaxError!

# ✅ Správně:
if x > 5:
    print("Ahoj")
```

### 2. Záměna `=` a `==`
```python
# ❌ Přiřazení místo porovnání:
if x = 5:             # SyntaxError!

# ✅ Správně (porovnání):
if x == 5:
```

### 3. Špatné odsazení
```python
# ❌ Chybné odsazení:
if x > 5:
print("Ahoj")        # IndentationError!

# ✅ Správně:
if x > 5:
    print("Ahoj")    # 4 mezery odsazení
```

### 4. Pořadí podmínek u `elif`
```python
# ❌ Špatné pořadí – "nad 10" se nikdy neprovede:
znamka = 1
if znamka >= 0:
    print("Kladné")       # provede se vždy pro kladné číslo
elif znamka >= 10:
    print("Nad 10")       # nikdy se neprovede!

# ✅ Správné pořadí – specifičtější podmínka první:
if znamka >= 10:
    print("Nad 10")
elif znamka >= 0:
    print("Kladné (0-9)")
```

> 🎯 **Tip:** Při `elif` vždy pište **specifičtější (přísnější)** podmínky **nahoře** a obecnější dole.

# 📋 Shrnutí

| Struktura | Počet větví | Popis |
|---|---|---|
| `if` | 0 nebo 1 | Provede příkazy jen při splnění podmínky |
| `if-else` | přesně 1 | Vybere jednu ze dvou větví |
| `if-elif-else` | přesně 1 | Vybere jednu z mnoha větví |
| Vnořené `if` | záleží | Podmínka uvnitř podmínky – zpřesňuje rozhodování |

| Logický operátor | Význam | Příklad |
|---|---|---|
| `and` | obě musí být True | `x > 0 and x < 100` |
| `or` | alespoň jedna True | `x < 0 or x > 100` |
| `not` | obrátí True/False | `not je_zavreno` |
