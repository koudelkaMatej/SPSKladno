# 💬 PyQt5 – Lekce 7: Dialogy a Upozornění

Kdykoli chcete uživateli zobrazit rychlou zprávu, varování, nebo se ho zeptat
na potvrzení, nemusíte programovat novou třídu okna. PyQt má připravené dialogy.

---

## 📦 Přehled typů dialogů

| Dialog | Použití | Příkaz |
|---|---|---|
| `information` | Informace, úspěch | `QMessageBox.information(...)` |
| `warning` | Varování, upozornění | `QMessageBox.warning(...)` |
| `critical` | Závažná chyba | `QMessageBox.critical(...)` |
| `question` | Otázka Ano/Ne | `QMessageBox.question(...)` |
| `QInputDialog` | Vstupní dialog (text, číslo) | `QInputDialog.getText(...)` |
| `QFileDialog` | Výběr souboru/složky | `QFileDialog.getOpenFileName(...)` |
| `QColorDialog` | Výběr barvy | `QColorDialog.getColor(...)` |


---

## 🟢 Ukázka 1 – Základní typy `QMessageBox`

```python
from PyQt5.QtWidgets import QApplication, QWidget, QVBoxLayout, QPushButton, QMessageBox

app = QApplication([])
okno = QWidget()
okno.setWindowTitle("Typy dialogů")
layout = QVBoxLayout()
okno.setLayout(layout)

def info():
    QMessageBox.information(okno, "Informace", "Soubor byl úspěšně uložen! ✅")

def varovani():
    QMessageBox.warning(okno, "Varování", "Disk je téměř plný. Uvolněte místo!")

def chyba():
    QMessageBox.critical(okno, "Chyba", "Připojení k databázi selhalo!")

for text, fce in [("Ukáž info", info), ("Ukáž varování", varovani), ("Ukáž chybu", chyba)]:
    btn = QPushButton(text)
    btn.clicked.connect(fce)
    layout.addWidget(btn)

okno.show()
app.exec()
```

## 🟡 Ukázka 2 – Otázka Ano/Ne a zachycení odpovědi

```python
from PyQt5.QtWidgets import QApplication, QWidget, QVBoxLayout, QPushButton, QLabel, QMessageBox

app = QApplication([])
okno = QWidget()
okno.setWindowTitle("Otázka Ano/Ne")
layout = QVBoxLayout()
okno.setLayout(layout)

info = QLabel("Žádná odpověď.")
btn = QPushButton("Smazat soubor")
layout.addWidget(btn)
layout.addWidget(info)

def potvrdit_smazani():
    odpoved = QMessageBox.question(
        okno,
        "Potvrzení",
        "Opravdu chcete soubor trvale smazat?",
        QMessageBox.Yes | QMessageBox.No
    )
    if odpoved == QMessageBox.Yes:
        info.setText("🗑️ Soubor byl smazán.")
    else:
        info.setText("✅ Mazání bylo zrušeno.")

btn.clicked.connect(potvrdit_smazani)
okno.show()
app.exec()
```

## 🟡 Ukázka 3 – `QInputDialog` (vstupní dialogy)

```python
from PyQt5.QtWidgets import QApplication, QWidget, QVBoxLayout, QPushButton, QLabel, QInputDialog

app = QApplication([])
okno = QWidget()
okno.setWindowTitle("QInputDialog")
layout = QVBoxLayout()
okno.setLayout(layout)

info = QLabel("Zatím žádný vstup.")
btn_text = QPushButton("Zadej jméno")
btn_cislo = QPushButton("Zadej věk")
layout.addWidget(btn_text)
layout.addWidget(btn_cislo)
layout.addWidget(info)

def zadej_jmeno():
    jmeno, ok = QInputDialog.getText(okno, "Jméno", "Zadejte vaše jméno:")
    if ok and jmeno.strip():
        info.setText(f"Ahoj, {jmeno}!")

def zadej_vek():
    vek, ok = QInputDialog.getInt(okno, "Věk", "Zadejte váš věk:", 18, 0, 120)
    if ok:
        info.setText(f"Věk: {vek} let")

btn_text.clicked.connect(zadej_jmeno)
btn_cislo.clicked.connect(zadej_vek)
okno.show()
app.exec()
```

## 🔴 Ukázka 4 – `QFileDialog` a `QColorDialog`

```python
from PyQt5.QtWidgets import QApplication, QWidget, QVBoxLayout, QPushButton, QLabel, QFileDialog, QColorDialog

app = QApplication([])
okno = QWidget()
okno.setWindowTitle("FileDialog + ColorDialog")
okno.resize(400, 150)
layout = QVBoxLayout()
okno.setLayout(layout)

info = QLabel("Zatím nic nevybráno.")
btn_soubor = QPushButton("Vybrat soubor")
btn_barva = QPushButton("Vybrat barvu")
layout.addWidget(btn_soubor)
layout.addWidget(btn_barva)
layout.addWidget(info)

def vyber_soubor():
    cesta, _ = QFileDialog.getOpenFileName(okno, "Otevřít soubor", "", "Všechny soubory (*)")
    if cesta:
        info.setText(f"Soubor: {cesta}")

def vyber_barvu():
    barva = QColorDialog.getColor()
    if barva.isValid():
        info.setText(f"Vybraná barva: {barva.name()}")
        info.setStyleSheet(f"background-color: {barva.name()}; padding: 4px;")

btn_soubor.clicked.connect(vyber_soubor)
btn_barva.clicked.connect(vyber_barvu)
okno.show()
app.exec()
```
