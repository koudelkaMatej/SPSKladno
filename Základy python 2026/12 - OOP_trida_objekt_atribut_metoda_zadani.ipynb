````python
# === SprÃ¡va vÃ­ce objektÅ¯ ===
class Zak:
    def __init__(self, jmeno, rocnik):
        self.jmeno = jmeno
        self.rocnik = rocnik
        self.znamky = []

    def pridej_znamku(self, znamka):
        self.znamky.append(znamka)

    def prumer(self):
        if len(self.znamky) == 0:
            return 0
        return sum(self.znamky) / len(self.znamky)

    def __str__(self):
        z = ", ".join(str(x) for x in self.znamky) if self.znamky else "Å¾Ã¡dnÃ©"
        return f"{self.jmeno} ({self.rocnik}. roÄnÃ­k) â€“ znÃ¡mky: {z}"

# === VytvoÅ™Ã­me seznam Å¾Ã¡kÅ¯ ===
zaci = [
    Zak("Adam", 1),
    Zak("BÃ¡ra", 2),
    Zak("Cyril", 1),
    Zak("Dana", 3),
]

# === PÅ™idÃ¡me znÃ¡mky ===
zaci[0].pridej_znamku(1)
zaci[0].pridej_znamku(2)
zaci[0].pridej_znamku(1)
zaci[1].pridej_znamku(3)
zaci[1].pridej_znamku(2)
zaci[2].pridej_znamku(1)
zaci[3].pridej_znamku(2)
zaci[3].pridej_znamku(2)

# === VÃ½pis ===
print("ğŸ“‹ Seznam Å¾Ã¡kÅ¯:")
for zak in zaci:
    print(f"  {zak}  â†’  prÅ¯mÄ›r: {zak.prumer():.2f}")
````

## ğŸ’¡ PraktickÃ½ pÅ™Ã­klad â€“ BankovnÃ­ ÃºÄet

````python
# === TÅ™Ã­da BankovniUcet ===
class BankovniUcet:
    def __init__(self, majitel, pocatecni_zustatek=0):
        self.majitel = majitel
        self.zustatek = pocatecni_zustatek
        self.historie = []     # seznam operacÃ­

    def vloz(self, castka):
        self.zustatek += castka
        self.historie.append(f"+{castka} KÄ")
        print(f"âœ… VloÅ¾eno {castka} KÄ â†’ zÅ¯statek: {self.zustatek} KÄ")

    def vyber(self, castka):
        if castka > self.zustatek:
            print(f"âŒ Nedostatek prostÅ™edkÅ¯! (poÅ¾adovÃ¡no: {castka}, mÃ¡te: {self.zustatek})")
            return
        self.zustatek -= castka
        self.historie.append(f"-{castka} KÄ")
        print(f"âœ… VybrÃ¡no {castka} KÄ â†’ zÅ¯statek: {self.zustatek} KÄ")

    def vypis_historii(self):
        print(f"\nğŸ“œ Historie ÃºÄtu ({self.majitel}):")
        for operace in self.historie:
            print(f"  {operace}")
        print(f"  AktuÃ¡lnÃ­ zÅ¯statek: {self.zustatek} KÄ")

    def __str__(self):
        return f"ÃšÄet: {self.majitel} | ZÅ¯statek: {self.zustatek} KÄ"

# === PouÅ¾itÃ­ ===
ucet = BankovniUcet("Petr NovÃ¡k", 1000)
print(ucet)

ucet.vloz(500)
ucet.vyber(200)
ucet.vyber(2000)      # nedostatek prostÅ™edkÅ¯
ucet.vloz(100)
ucet.vypis_historii()
````

## ğŸ”€ VÃ­ce zpÅ¯sobÅ¯ vytvoÅ™enÃ­ objektu (pÅ™etÄ›Å¾ovÃ¡nÃ­ konstruktoru)

V jazycÃ­ch jako **Java** nebo **C#** lze mÃ­t vÃ­ce konstruktorÅ¯ se stejnÃ½m nÃ¡zvem, ale rÅ¯znÃ½mi parametry (**pÅ™etÄ›Å¾ovÃ¡nÃ­** / *overloading*):

```java
// Java â€“ vÃ­ce konstruktorÅ¯
class Bod {
    Bod(int x, int y) { ... }       // ze souÅ™adnic
    Bod(String text)  { ... }       // z textu "3,5"
}
```

V Pythonu **to nejde** â€“ tÅ™Ã­da mÅ¯Å¾e mÃ­t pouze **jeden** `__init__()`. Pokud definujete druhÃ½, pÅ™epÃ­Å¡e ten prvnÃ­.

### Jak to Å™eÅ¡Ã­me v Pythonu?

### 1. VÃ½chozÃ­ hodnoty parametrÅ¯ (nejÄastÄ›jÅ¡Ã­)

ParametrÅ¯m v `__init__` dÃ¡me vÃ½chozÃ­ hodnotu â€“ pak je volÃ¡nÃ­ flexibilnÃ­:

````python
# === 1. VÃ½chozÃ­ hodnoty parametrÅ¯ ===
class Bod:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def __str__(self):
        return f"Bod({self.x}, {self.y})"

# RÅ¯znÃ© zpÅ¯soby vytvoÅ™enÃ­:
b1 = Bod(3, 5)       # oba argumenty
b2 = Bod(7)           # jen x, y = 0
b3 = Bod()            # oba vÃ½chozÃ­
b4 = Bod(y=10)        # jen y, x = 0

print(b1)   # Bod(3, 5)
print(b2)   # Bod(7, 0)
print(b3)   # Bod(0, 0)
print(b4)   # Bod(0, 10)
````

### 2. `None` jako vÃ½chozÃ­ hodnota + vÄ›tvenÃ­

Pokud chceme, aby `__init__` pÅ™ijÃ­mal **ÃºplnÄ› jinÃ© kombinace** argumentÅ¯, pouÅ¾ijeme `None` a pak rozhodujeme v kÃ³du:


````python
# === 2. None jako vÃ½chozÃ­ hodnota ===
class Kruh:
    def __init__(self, r=None, d=None):
        if r is not None:
            self.r = r
        elif d is not None:
            self.r = d / 2
        else:
            self.r = 1    # vÃ½chozÃ­ polomÄ›r

    def obsah(self):
        return 3.14159 * self.r ** 2

    def __str__(self):
        return f"Kruh(r={self.r})"

# === RÅ¯znÃ© zpÅ¯soby vytvoÅ™enÃ­ ===
k1 = Kruh(r=5)        # z polomÄ›ru
k2 = Kruh(d=10)       # z prÅ¯mÄ›ru â†’ r = 5
k3 = Kruh()           # vÃ½chozÃ­ r = 1

print(k1, f"â†’ obsah: {k1.obsah():.2f}")
print(k2, f"â†’ obsah: {k2.obsah():.2f}")
print(k3, f"â†’ obsah: {k3.obsah():.2f}")
````

### 3. `@classmethod` â€“ alternativnÃ­ konstruktor

Pro ÃºplnÄ› **jinÃ½ zpÅ¯sob** vytvoÅ™enÃ­ objektu (napÅ™. z textu, ze souboru) pouÅ¾ijeme **`@classmethod`**:

- DekorÃ¡tor `@classmethod` Å™Ã­kÃ¡ Pythonu, Å¾e metoda pracuje s **tÅ™Ã­dou** (ne s konkrÃ©tnÃ­m objektem)
- MÃ­sto `self` dostane parametr `cls` (odkaz na tÅ™Ã­du)
- UvnitÅ™ zavolÃ¡me `cls(...)`, coÅ¾ vytvoÅ™Ã­ novÃ½ objekt

````python
# === 3. @classmethod â€“ alternativnÃ­ konstruktor ===
class Barva:
    def __init__(self, r, g, b):
        self.r = r
        self.g = g
        self.b = b

    @classmethod
    def z_hexkodu(cls, hex_kod):
        """VytvoÅ™Ã­ Barvu z hex kÃ³du, napÅ™. '#FF8800'."""
        hex_kod = hex_kod.lstrip("#")
        r = int(hex_kod[0:2], 16)
        g = int(hex_kod[2:4], 16)
        b = int(hex_kod[4:6], 16)
        return cls(r, g, b)           # cls(...) = Barva(r, g, b)

    @classmethod
    def bila(cls):
        """VytvoÅ™Ã­ bÃ­lou barvu."""
        return cls(255, 255, 255)

    def __str__(self):
        return f"RGB({self.r}, {self.g}, {self.b})"

# === RÅ¯znÃ© zpÅ¯soby vytvoÅ™enÃ­ ===
b1 = Barva(255, 128, 0)           # normÃ¡lnÃ­ __init__
b2 = Barva.z_hexkodu("#FF8800")   # alternativnÃ­ konstruktor
b3 = Barva.bila()                  # pojmenovanÃ½ konstruktor

print(b1)   # RGB(255, 128, 0)
print(b2)   # RGB(255, 136, 0)
print(b3)   # RGB(255, 255, 255)
````

### SrovnÃ¡nÃ­

| PÅ™Ã­stup | Kdy pouÅ¾Ã­t | PÅ™Ã­klad |
|:--------|:-----------|:--------|
| **VÃ½chozÃ­ hodnoty** | NÄ›kterÃ© argumenty jsou volitelnÃ© | `Bod(x=0, y=0)` |
| **`None` + vÄ›tvenÃ­** | VzÃ¡jemnÄ› se vyluÄujÃ­cÃ­ argumenty | `Kruh(r=5)` nebo `Kruh(d=10)` |
