# 📦 Moduly a balíčky

## Co je to modul?

**Modul** je soubor s příponou `.py`, který obsahuje Python kód (funkce, třídy, proměnné). Moduly slouží k **organizaci kódu** do logických celků a k **znovupoužití** – místo kopírování kódu ho jednoduše **naimportujeme**.

> 💡 **Představte si:** Modul je jako krabička s nástroji. Místo toho, abyste si pokaždé vyráběli kladivo, vezmete ho z krabičky (naimportujete).

### Typy modulů

| Typ | Popis | Příklad |
|:---|:---|:---|
| **Vestavěné** | Součást Pythonu, vždy dostupné | `math`, `random`, `os`, `sys` |
| **Standardní knihovna** | Součást Pythonu, bohatá kolekce | `json`, `datetime`, `re`, `csv` |
| **Externí (třetí strany)** | Instalují se přes `pip` | `requests`, `numpy`, `flask` |
| **Vlastní** | Vaše vlastní `.py` soubory | `muj_modul.py` |

---

## 📥 Import – základní způsoby

### 1. `import modul`
Naimportuje celý modul. K funkcím přistupujeme přes **tečkovou notaci**:

```python
import math

print(math.sqrt(16))    # 4.0
print(math.pi)          # 3.141592653589793
print(math.factorial(5))# 120
```

### 2. `from modul import něco`
Naimportuje **konkrétní** funkci/třídu/proměnnou – nemusíme psát název modulu:

```python
from math import sqrt, pi

print(sqrt(16))   # 4.0
print(pi)         # 3.141592653589793
```

### 3. `import modul as alias`
Naimportuje modul pod **zkráceným názvem**:

```python
import random as rnd

print(rnd.randint(1, 10))
```

### 4. `from modul import *` ⚠️
Naimportuje **všechno** z modulu. **Nedoporučuje se** – nevíte, co přesně se naimportovalo, a může dojít ke konfliktu jmen.

```python
from math import *    # ⚠️ Raději nepoužívat
```

````python
# Vyzkoušejte si různé importy
import math
print("sqrt(25) =", math.sqrt(25))
print("pi =", math.pi)

from random import randint, choice
print("Náhodné číslo 1-10:", randint(1, 10))
print("Náhodný výběr:", choice(["kámen", "nůžky", "papír"]))

import os
print("Aktuální složka:", os.getcwd())
````

## 🔧 Užitečné vestavěné moduly

| Modul | K čemu slouží | Příklad funkce |
|:---|:---|:---|
| `math` | Matematické operace | `sqrt()`, `sin()`, `pi`, `ceil()` |
| `random` | Náhodná čísla | `randint()`, `choice()`, `shuffle()` |
| `os` | Práce s operačním systémem | `getcwd()`, `listdir()`, `path.exists()` |
| `sys` | Systémové informace | `sys.version`, `sys.argv` |
| `json` | Práce s JSON daty | `loads()`, `dumps()` |
| `csv` | Čtení/zápis CSV souborů | `reader()`, `writer()` |
| `datetime` | Datum a čas | `datetime.now()`, `timedelta()` |
| `re` | Regulární výrazy | `search()`, `findall()` |

````python
# Ukázky vestavěných modulů
import math
print("Zaokrouhlení nahoru:", math.ceil(4.2))   # 5
print("Zaokrouhlení dolů:", math.floor(4.8))    # 4

import random
seznam = [1, 2, 3, 4, 5]
random.shuffle(seznam)
print("Zamíchaný seznam:", seznam)

import os
print("Soubory v aktuální složce:", os.listdir(".")[:5])  # prvních 5
````

## 📦 Balíčky a pip

**Balíček (package)** = složka obsahující více modulů + soubor `__init__.py`.

**pip** = nástroj pro instalaci **externích balíčků** z [PyPI](https://pypi.org/) (Python Package Index).

### Základní příkazy pip

```bash
# Instalace balíčku
pip install requests

# Instalace konkrétní verze
pip install requests==2.28.0

# Zobrazení nainstalovaných balíčků
pip list

# Odinstalace
pip uninstall requests

# Uložení závislostí do souboru
pip freeze > requirements.txt

# Instalace ze souboru
pip install -r requirements.txt
```

> 💡 V Jupyter notebooku můžete pip spustit s vykřičníkem: `!pip install requests`

## 🏗️ Vytvoření vlastního modulu

Stačí vytvořit `.py` soubor a naimportovat ho!

### Příklad:

**Soubor `matematika.py`:**
```python
def secti(a, b):
    return a + b

def odecti(a, b):
    return a - b

PI = 3.14159
```

**Soubor `hlavni.py` (ve stejné složce):**
```python
import matematika

print(matematika.secti(3, 5))    # 8
print(matematika.PI)             # 3.14159
```

### `if __name__ == "__main__":`

Tento blok se spustí **pouze** když soubor spustíme přímo (ne když ho importujeme):

```python
# matematika.py
def secti(a, b):
    return a + b

if __name__ == "__main__":
    # Tento kód se spustí JEN při přímém spuštění
    print("Test:", secti(2, 3))  # 5
```
