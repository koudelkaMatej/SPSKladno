# 🧰 PyQt5 – Lekce 6: Složitější Widgety

Zatím jsme používali `QLabel`, `QPushButton` a `QLineEdit`.
PyQt nabízí mnohem více specializovaných prvků pro interakci s uživatelem.

---

## 📦 Přehled nových widgetů

| Widget | Typ vstupu | Jak získat hodnotu |
|---|---|---|
| `QCheckBox` | Zaškrtávátko (Ano/Ne) | `.isChecked()` → True/False |
| `QRadioButton` | Výběr jednoho z více | `.isChecked()` → True/False |
| `QButtonGroup` | Správce skupiny RadioButtonů | `.checkedButton().text()` |
| `QComboBox` | Rozbalovací seznam | `.currentText()` nebo `.currentIndex()` |
| `QSlider` | Posuvník | `.value()` |
| `QProgressBar` | Ukazatel průběhu | `.setValue(n)` (0–100) |
| `QListWidget` | Seznam položek | `.currentItem().text()` |

> 💡 `QRadioButton` správně funguje (volba jen jednoho) tehdy, jsou-li v jednom `QButtonGroup` nebo ve stejném přímém rodiči (layoutu).


---

## 🟢 Ukázka 1 – `QCheckBox` a `QComboBox`

```python
from PyQt5.QtWidgets import QApplication, QWidget, QVBoxLayout, QCheckBox, QComboBox, QPushButton, QLabel

app = QApplication([])
okno = QWidget()
okno.setWindowTitle("CheckBox + ComboBox")
okno.resize(300, 200)
layout = QVBoxLayout()
okno.setLayout(layout)

check_souhlas = QCheckBox("Souhlasím s podmínkami")
combo_jazyk = QComboBox()
combo_jazyk.addItems(["Python", "Java", "C#", "JavaScript", "Kotlin"])
vysledek = QLabel("—")
tlacitko = QPushButton("Odeslat")

layout.addWidget(QLabel("Oblíbený jazyk:"))
layout.addWidget(combo_jazyk)
layout.addWidget(check_souhlas)
layout.addWidget(tlacitko)
layout.addWidget(vysledek)

def odeslat():
    if not check_souhlas.isChecked():
        vysledek.setText("⛔ Musíte souhlasit s podmínkami!")
        return
    jazyk = combo_jazyk.currentText()
    vysledek.setText(f"✅ Skvělá volba: {jazyk}")

tlacitko.clicked.connect(odeslat)
okno.show()
app.exec()
```

## 🟢 Ukázka 2 – `QRadioButton` a `QButtonGroup`

```python
from PyQt5.QtWidgets import QApplication, QWidget, QVBoxLayout, QRadioButton, QButtonGroup, QPushButton, QLabel

app = QApplication([])
okno = QWidget()
okno.setWindowTitle("RadioButton + ButtonGroup")
okno.resize(300, 200)
layout = QVBoxLayout()
okno.setLayout(layout)

layout.addWidget(QLabel("Vyberte obtížnost:"))

skupina = QButtonGroup()
radios = []
for text in ["Lehká 🟢", "Střední 🟡", "Těžká 🔴"]:
    rb = QRadioButton(text)
    skupina.addButton(rb)
    layout.addWidget(rb)
    radios.append(rb)
radios[0].setChecked(True)   # Výchozí volba

vysledek = QLabel("—")
tlacitko = QPushButton("Potvrdit volbu")
layout.addWidget(tlacitko)
layout.addWidget(vysledek)

def potvrdit():
    vybrano = skupina.checkedButton()
    if vybrano:
        vysledek.setText(f"Zvolena obtížnost: {vybrano.text()}")

tlacitko.clicked.connect(potvrdit)
okno.show()
app.exec()
```

## 🟡 Ukázka 3 – `QSlider` a `QProgressBar`

```python
from PyQt5.QtWidgets import QApplication, QWidget, QVBoxLayout, QSlider, QProgressBar, QLabel
from PyQt5.QtCore import Qt

app = QApplication([])
okno = QWidget()
okno.setWindowTitle("Slider + ProgressBar")
okno.resize(350, 200)
layout = QVBoxLayout()
okno.setLayout(layout)

layout.addWidget(QLabel("Táhněte posuvník:"))

slider = QSlider(Qt.Horizontal)
slider.setMinimum(0)
slider.setMaximum(100)
slider.setValue(50)

progress = QProgressBar()
progress.setValue(50)

info = QLabel("Hodnota: 50")

layout.addWidget(slider)
layout.addWidget(progress)
layout.addWidget(info)

def aktualizuj(hodnota):
    progress.setValue(hodnota)
    info.setText(f"Hodnota: {hodnota}")
    if hodnota < 33:
        info.setStyleSheet("color: green;")
    elif hodnota < 66:
        info.setStyleSheet("color: orange;")
    else:
        info.setStyleSheet("color: red;")

slider.valueChanged.connect(aktualizuj)   # Signál při každé změně hodnoty
okno.show()
app.exec()
```

## 🟡 Ukázka 4 – `QListWidget` (seznam položek)

```python
from PyQt5.QtWidgets import QApplication, QWidget, QVBoxLayout, QHBoxLayout, QListWidget, QLineEdit, QPushButton, QLabel

app = QApplication([])
okno = QWidget()
okno.setWindowTitle("Seznam úkolů")
okno.resize(320, 300)

hlavni = QVBoxLayout()
okno.setLayout(hlavni)

seznam = QListWidget()
seznam.addItems(["Nakoupit", "Zavolat lékaři", "Dokončit projekt"])

vstup = QLineEdit()
vstup.setPlaceholderText("Nový úkol…")
btn_pridat = QPushButton("Přidat")
btn_smazat = QPushButton("Smazat vybraný")
info = QLabel("Klikněte na položku pro výběr")

radek = QHBoxLayout()
radek.addWidget(vstup)
radek.addWidget(btn_pridat)

hlavni.addWidget(seznam)
hlavni.addLayout(radek)
hlavni.addWidget(btn_smazat)
hlavni.addWidget(info)

def pridat():
    text = vstup.text().strip()
    if text:
        seznam.addItem(text)
        vstup.clear()

def smazat():
    polozka = seznam.currentItem()
    if polozka:
        seznam.takeItem(seznam.row(polozka))

def vybrat(polozka):
    info.setText(f"Vybráno: {polozka.text()}")

btn_pridat.clicked.connect(pridat)
btn_smazat.clicked.connect(smazat)
seznam.itemClicked.connect(vybrat)

okno.show()
app.exec()
```
