# Cyklus `while`

## Co je to cyklus?

Cyklus (anglicky *loop*) je základní programátorská konstrukce, která umožňuje **opakovaně vykonávat blok kódu**. Místo toho, abychom stejný kód psali znovu a znovu, zapíšeme jej jednou a necháme ho opakovat automaticky.

V Pythonu existují dva hlavní typy cyklů:

| Cyklus | Kdy ho použít |
|--------|--------------|
| `for` | Když **předem víme**, kolikrát se má kód opakovat (nebo procházíme kolekci) |
| `while` | Když **předem nevíme**, kolikrát se má kód opakovat – opakujeme **dokud platí podmínka** |

## Proč potřebujeme `while`?

Představte si tyto situace:
- Hra běží, **dokud** hráč neprohraje
- Program se ptá na heslo, **dokud** uživatel nezadá správné
- Číslo dělíme dvěma, **dokud** je větší než 1

Ve všech těchto případech nevíme předem, kolik opakování bude potřeba — záleží na okolnostech za běhu programu. Přesně na to slouží `while`.

## Syntaxe (jak se to píše)

```python
while podmínka:
    příkaz_1
    příkaz_2
    příkaz_3

příkaz_mimo_while
```

**Klíčové body:**
- Za klíčovým slovem `while` následuje **podmínka** (logický výraz, stejně jako u `if`)
- Za podmínkou musí být **dvojtečka** `:`
- Tělo cyklu (příkazy uvnitř) musí být **odsazené** (standardně 4 mezery)
- Příkazy na stejné úrovni odsazení jako `while` už **nejsou součástí cyklu**

## Sémantika (jak to funguje)

Průběh `while` cyklu krok po kroku:

1. **Vyhodnoť podmínku** → výsledek je `True` nebo `False`
2. Pokud je podmínka **`True`** → proveď všechny příkazy v těle cyklu, pak **se vrať na krok 1**
3. Pokud je podmínka **`False`** → **přeskoč** tělo cyklu a pokračuj příkazem za cyklem

### Vizuální schéma

```
        ┌──────────────────┐
        │  Vyhodnoť podmínku│
        └────────┬─────────┘
                 │
           ┌─────▼─────┐
           │  True?     │───── NE ────► Pokračuj za cyklem
           └─────┬──────┘
                 │ ANO
           ┌─────▼──────┐
           │ Proveď tělo │
           │   cyklu     │
           └─────┬──────┘
                 │
                 └──────────────► (zpět na vyhodnocení podmínky)
```

> **Důležité:** Podmínka se kontroluje **vždy na začátku** každého průchodu. Pokud je podmínka `False` hned na začátku, tělo cyklu se **neprovede ani jednou**.

## Jednoduché příklady

### Příklad 1: Odpočítávání
Jednoduchý odpočet od 5 do 1:

````python
pocitadlo = 5                    # počáteční hodnota
while pocitadlo > 0:             # opakuj, dokud je počítadlo kladné
    print(pocitadlo)             # vypiš aktuální hodnotu
    pocitadlo = pocitadlo - 1    # sniž hodnotu o 1 (KLÍČOVÉ — bez tohoto by cyklus nikdy neskončil!)
print("Start! 🚀")
````

### Příklad 2: Sčítání čísel od uživatele

Program se opakovaně ptá na číslo a přičítá ho k součtu. Skončí, když uživatel zadá 0:

````python
soucet = 0
cislo = int(input("Zadej číslo (0 = konec): "))

while cislo != 0:
    soucet = soucet + cislo
    cislo = int(input("Zadej číslo (0 = konec): "))

print(f"Součet všech zadaných čísel: {soucet}")
````

### Příklad 3: Collatzův problém (3n+1)

Slavný matematický problém — vezmeme libovolné přirozené číslo a opakujeme:
- je-li **sudé**, vydělíme ho dvěma
- je-li **liché**, vynásobíme třemi a přičteme jedna

Hypotéza říká, že **vždy nakonec dojdeme k číslu 1** (nikdo to zatím nedokázal dokázat ani vyvrátit!).

Více na [Wikipedii](https://cs.wikipedia.org/wiki/Collatz%C5%AFv_probl%C3%A9m).

````python
x = 6
print("Počáteční hodnota:", x)

while x > 1:                    # opakuj, dokud jsme nedosáhli jedničky
    if x % 2 == 0:              # sudé číslo → vyděl dvěma
        x = x // 2
    else:                        # liché číslo → vynásob 3 a přičti 1
        x = 3 * x + 1
    print("→", x)

print("Hotovo! Došli jsme k 1.")
````

# Důležité detaily a časté chyby

## 1. Klíčové slovo `pass`

Uvnitř těla cyklu musí být vždy **alespoň jeden příkaz**. Pokud chceme cyklus zatím nechat prázdný (jako zástupný kód), použijeme klíčové slovo `pass`:

```python
while podminka:
    pass    # nic nedělej — jen zástupný příkaz (v budoucnu tu něco bude, ale zatím nevím co)
```

## 2. Nekonečný cyklus ⚠️

**Nejčastější chyba při práci s `while`!**

Pokud se podmínka **nikdy nestane `False`**, cyklus poběží donekonečna a program "zamrzne":

```python
# ❌ ŠPATNĚ — nekonečný cyklus!
x = 0
while x < 10:
    print(x)
    # Chybí: x = x + 1  → x je pořád 0, podmínka je pořád True
```

**Jak se tomu vyhnout:**
- Vždy se ujistěte, že **v těle cyklu měníte proměnnou**, která je v podmínce
- Ověřte, že se hodnota proměnné **postupně přibližuje** k nesplnění podmínky

> 💡 **Tip:** Pokud vám program "zamrzne" — pravděpodobně jste vytvořili nekonečný cyklus. V Jupyter notebooku můžete buňku zastavit tlačítkem stop (⏹), v terminálu kombinací `Ctrl + C`.

## 3. Podmínka se kontroluje PŘED průchodem

Pokud je podmínka `False` hned na začátku, tělo cyklu se **neprovede ani jednou**:

```python
x = 100
while x < 10:      # 100 < 10 je False → tělo se přeskočí
    print(x)
    x = x + 1
print("Konec")      # Vypíše se rovnou "Konec"
```

## 4. Zkrácený zápis operací

Při práci s cykly často měníme hodnotu proměnné. Python nabízí zkrácené operátory:

| Zkrácený zápis | Ekvivalent |
|----------------|------------|
| `x += 1` | `x = x + 1` |
| `x -= 1` | `x = x - 1` |
| `x *= 2` | `x = x * 2` |
| `x //= 2` | `x = x // 2` |

Příklad:
```python
pocitadlo = 0
while pocitadlo < 5:
    print(pocitadlo)
    pocitadlo += 1      # totéž jako: pocitadlo = pocitadlo + 1
```

## 5. `while` vs `for` — kdy co použít?

| Situace | Vhodný cyklus | Proč |
|---------|--------------|------|
| Projít seznam studentů | `for` | Víme, kolik jich je |
| Opakovat kód 10× | `for` + `range` | Známe přesný počet |
| Ptát se na heslo, dokud není správné | `while` | Nevíme kolikrát uživatel chybuje |
| Hledat číslo splňující podmínku | `while` | Nevíme, kdy ho najdeme |
| Hra běží, dokud hráč žije | `while` | Délka hry je nepředvídatelná |
