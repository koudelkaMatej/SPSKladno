# 📅 Datum a čas

## Modul `datetime`

Python má vestavěný modul `datetime` pro práci s **datem a časem**. Je to jeden z nejpoužívanějších modulů – datum a čas potřebujete téměř v každé aplikaci.

### Hlavní třídy

| Třída | Co reprezentuje | Příklad |
|:---|:---|:---|
| `date` | Pouze **datum** (rok, měsíc, den) | `2026-03-31` |
| `time` | Pouze **čas** (hodina, minuta, sekunda) | `14:30:00` |
| `datetime` | **Datum i čas** dohromady | `2026-03-31 14:30:00` |
| `timedelta` | **Rozdíl** mezi dvěma časy | `5 days, 3:00:00` |

````python
from datetime import date, time, datetime, timedelta

# Aktuální datum a čas
ted = datetime.now()
print("Teď je:", ted)

# Pouze datum
dnes = date.today()
print("Dnešní datum:", dnes)

# Vytvoření konkrétního data
narozeniny = date(2005, 6, 15)
print("Narozeniny:", narozeniny)
````

---

## 🔍 Přístup k jednotlivým složkám

Z objektu `datetime` nebo `date` můžeme získat jednotlivé části:

````python
from datetime import datetime

ted = datetime.now()

print("Rok:", ted.year)
print("Měsíc:", ted.month)
print("Den:", ted.day)
print("Hodina:", ted.hour)
print("Minuta:", ted.minute)
print("Sekunda:", ted.second)
print("Den v týdnu:", ted.weekday())   # 0 = pondělí, 6 = neděle
````

## 📝 Formátování – `strftime()`

Metoda `strftime()` (string format time) převede datum/čas na **formátovaný řetězec**:

| Kód | Význam | Příklad |
|:---|:---|:---|
| `%d` | Den (01–31) | `31` |
| `%e` | Den bez úvodní nuly (1–31) | `3` |
| `%m` | Měsíc (01–12) | `03` |
| `%B` | Název měsíce (plný) | `March` |
| `%b` | Název měsíce (zkrácený) | `Mar` |
| `%Y` | Rok (4 číslice) | `2026` |
| `%y` | Rok (2 číslice) | `26` |
| `%H` | Hodina 24h (00–23) | `14` |
| `%I` | Hodina 12h (01–12) | `02` |
| `%p` | AM / PM | `PM` |
| `%M` | Minuta (00–59) | `30` |
| `%S` | Sekunda (00–59) | `00` |
| `%A` | Název dne v týdnu (plný) | `Tuesday` |
| `%a` | Název dne v týdnu (zkrácený) | `Tue` |
| `%j` | Den v roce (001–366) | `090` |
| `%W` | Číslo týdne v roce (00–53) | `13` |

````python
from datetime import datetime

ted = datetime.now()

# Český formát
print(ted.strftime("%d.%m.%Y"))           # 31.03.2026
print(ted.strftime("%d.%m.%Y %H:%M"))     # 31.03.2026 14:30
print(ted.strftime("%H:%M:%S"))            # 14:30:00
print(ted.strftime("%A, %d. %B %Y"))       # Tuesday, 31. March 2026
````

## 📥 Parsování – `strptime()`

Metoda `strptime()` (string parse time) převede **řetězec** na objekt `datetime`:

```python
from datetime import datetime

text = "25.12.2025 18:00"
dt = datetime.strptime(text, "%d.%m.%Y %H:%M")
print(dt)        # 2025-12-25 18:00:00
print(dt.year)   # 2025
```

> ⚠️ Formát v `strptime()` musí **přesně odpovídat** formátu vstupního řetězce!

````python
from datetime import datetime

# Parsování českého formátu
datum_text = "15.06.2005"
datum = datetime.strptime(datum_text, "%d.%m.%Y")
print("Parsované datum:", datum)
print("Rok:", datum.year, "Měsíc:", datum.month)
````

## ⏳ Timedelta – počítání s časem

`timedelta` reprezentuje **časový rozdíl**. Můžeme s ním **přičítat** a **odečítat** od data:

````python
from datetime import datetime, timedelta, date

dnes = date.today()

# Přičtení dnů
za_tyden = dnes + timedelta(days=7)
print("Za týden:", za_tyden)

# Odečtení dnů
pred_mesic = dnes - timedelta(days=30)
print("Před měsícem:", pred_mesic)

# Rozdíl dvou dat
vanoce = date(2026, 12, 25)
rozdil = vanoce - dnes
print(f"Do Vánoc zbývá {rozdil.days} dní")

# Timedelta s hodinami a minutami
ted = datetime.now()
za_2_hodiny = ted + timedelta(hours=2, minutes=30)
print("Za 2.5 hodiny:", za_2_hodiny.strftime("%H:%M"))
````

---

## 🔄 Porovnávání dat

Data a časy lze **porovnávat** pomocí operátorů `<`, `>`, `==`, `!=`:

```python
from datetime import date

d1 = date(2026, 1, 1)
d2 = date(2026, 12, 31)

print(d1 < d2)    # True  (leden je před prosincem)
print(d1 == d2)   # False
print(d1 > d2)    # False
```
