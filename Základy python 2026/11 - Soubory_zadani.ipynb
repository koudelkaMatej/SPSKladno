# 📁 Práce se soubory

## Proč potřebujeme soubory?

Veškerá data, se kterými Python pracuje, existují **pouze v operační paměti (RAM)**. Jakmile program skončí, **všechna data zmizí**.

```
┌──────────────────────────────────┐
│        Operační paměť (RAM)      │
│                                  │
│  proměnné, seznamy, výsledky...  │
│                                  │
│  ⚠️ Po ukončení programu = PRYČ  │
└──────────────────────────────────┘
          │ zápis          ▲ čtení
          ▼                │
┌──────────────────────────────────┐
│          Disk (SSD/HDD)          │
│                                  │
│  soubory: .txt, .csv, .json...   │
│                                  │
│  ✅ Data přežijí restart          │
└──────────────────────────────────┘
```

Soubory nám umožňují:

| Účel | Příklad |
|:---|:---|
| **Uložit výsledky** | Zapsat výpočty do `.txt` nebo `.csv` |
| **Načíst vstupní data** | Přečíst seznam jmen, tabulku hodnot |
| **Logování** | Zaznamenat průběh programu |
| **Konfigurace** | Načíst nastavení programu |

> 💡 V tomto materiálu se zaměříme na **textové soubory** (`.txt`, `.csv`). Pro specializované formáty (`.xlsx`, `.pdf`, `.json`) existují knihovny.

## 🔓 Otevírání souborů – `open()`

Soubor otevřeme funkcí `open()`, které předáme **cestu** a **režim**:

```python
soubor = open("nazev.txt", "r")   # otevře soubor
# ... práce se souborem ...
soubor.close()                      # NUTNÉ uzavřít!
```

### Režimy otevření

| Režim | Název | Co dělá |
|:---|:---|:---|
| `"r"` | **read** | Čtení – soubor musí existovat |
| `"w"` | **write** | Zápis – **smaže** obsah, vytvoří nový |
| `"a"` | **append** | Připojení – přidá na konec souboru |
| `"r+"` | read + write | Čtení i zápis (soubor musí existovat) |
| `"w+"` | write + read | Zápis i čtení (smaže obsah) |

> ⚠️ **Pozor na `"w"`!** Režim `"w"` **smaže veškerý obsah** souboru při otevření. Pokud chcete přidat data, použijte `"a"`.

---

## 🛡️ `with` – bezpečné otevírání souborů

Ruční zavírání souboru (`close()`) se snadno **zapomene**, zvlášť pokud nastane chyba. Proto Python nabízí příkaz `with`, který soubor **automaticky zavře**:

```python
with open("soubor.txt", "r") as f:
    obsah = f.read()
    # ... práce s obsahem ...
# ← Zde je soubor AUTOMATICKY zavřen (i při chybě!)
```

```
┌─────────────────────────────┐
│  with open(...) as f:       │
│      ┌─────────────────┐    │
│      │ soubor otevřen  │    │
│      │ práce s daty    │    │
│      └─────────────────┘    │
│  ← soubor automaticky zavřen│
└─────────────────────────────┘
```

> 🎯 **Pravidlo:** Vždy používejte `with` pro práci se soubory. Je to bezpečnější a přehlednější.

## ✏️ Zápis do souboru

### `write()` – zapíše text

Metoda `write()` zapíše **přesně** to, co jí předáte. **Nepřidává automaticky nový řádek** – musíte ho přidat sami (`\n`):

````python
# === Zápis do souboru (write) ===
with open("ukazkovy.txt", "w") as soubor:
    soubor.write("První řádek\n")
    soubor.write("Druhý řádek\n")
    soubor.write("Třetí řádek\n")

print("✅ Soubor 'ukazkovy.txt' vytvořen!")
````

### `writelines()` – zapíše seznam řádků

````python
# === Zápis seznamu řádků ===
radky = ["Anna\n", "Bára\n", "Cyril\n", "Dana\n"]

with open("jmena.txt", "w") as soubor:
    soubor.writelines(radky)

print("✅ Soubor 'jmena.txt' vytvořen!")
````

> ⚠️ `writelines()` také **nepřidává `\n`** automaticky – musíte ho mít v každém řetězci.

### Připojení na konec souboru (`"a"`)

````python
# === Append – přidání na konec ===
with open("jmena.txt", "a") as soubor:
    soubor.write("Emil\n")
    soubor.write("Franta\n")

print("✅ Přidáno na konec souboru 'jmena.txt'!")
````

## 📖 Čtení ze souboru

Python nabízí **tři způsoby** čtení:

| Metoda | Co vrátí | Kdy použít |
|:---|:---|:---|
| `read()` | Celý obsah jako **jeden string** | Malé soubory |
| `readline()` | **Jeden řádek** (při opakovaném volání další) | Postupné čtení |
| `readlines()` | **Seznam** všech řádků | Když potřebujete seznam |

````python
# === read() – celý soubor najednou ===
with open("ukazkovy.txt", "r") as f:
    obsah = f.read()
    print("=== read() ===")
    print(obsah)
    print(f"Typ: {type(obsah)}")  # <class 'str'>
````

````python
# === readline() – jeden řádek ===
with open("ukazkovy.txt", "r") as f:
    print("=== readline() ===")
    prvni = f.readline()
    druhy = f.readline()
    print(f"1. řádek: {prvni.strip()}")
    print(f"2. řádek: {druhy.strip()}")
````

````python
# === readlines() – seznam řádků ===
with open("ukazkovy.txt", "r") as f:
    radky = f.readlines()
    print("=== readlines() ===")
    print(f"Typ: {type(radky)}")    # <class 'list'>
    print(f"Počet řádků: {len(radky)}")
    print(f"Obsah: {radky}")
````

---

### Čtení řádek po řádku (cyklem `for`)

Nejelegantnější způsob – soubor můžeme přímo **procházet cyklem**:

````python
# === Čtení v cyklu for ===
with open("jmena.txt", "r") as f:
    for i, radek in enumerate(f, 1):
        jmeno = radek.strip()  # odstraní \n na konci
        print(f"  {i}. {jmeno}")
````

> 💡 **`strip()`** je klíčová metoda při čtení souborů – odstraní **bílé znaky** (mezery, `\n`, `\t`) ze začátku a konce řetězce.

> ⚠️ Každý řádek přečtený ze souboru **obsahuje `\n` na konci** (znak nového řádku). Bez `strip()` budete mít prázdné řádky navíc.

## 📌 Kurzor v souboru (file pointer)

Když Python otevře soubor, udržuje si **pozici kurzoru** – místo, odkud bude číst dál. Představte si ho jako prst sledující text v knize:

```
Po otevření:           Po prvním read():
┌──────────────┐       ┌──────────────┐
│▸První řádek  │       │ První řádek  │
│ Druhý řádek  │       │ Druhý řádek  │
│ Třetí řádek  │       │ Třetí řádek▸ │
└──────────────┘       └──────────────┘
 ↑ kurzor na začátku    ↑ kurzor na KONCI!
```

| Operace | Co se stane s kurzorem |
|:---|:---|
| `open()` | Kurzor na **začátku** souboru |
| `read()` | Přečte vše → kurzor na **konci** |
| `readline()` | Přečte 1 řádek → kurzor za ním |
| `seek(0)` | **Vrátí** kurzor na začátek |

> ⚠️ **Častý problém:** Zavoláte `read()` dvakrát a podruhé dostanete **prázdný řetězec**. Kurzor je totiž už na konci souboru a není co číst!

````python
# === Problém: dvojité read() ===
with open("ukazkovy.txt", "r") as f:
    prvni = f.read()
    druhy = f.read()   # ← kurzor je na konci → prázdný řetězec!

print(f"1. čtení: {repr(prvni[:30])}...")
print(f"2. čtení: {repr(druhy)}")  # ''

print("\n--- S použitím seek(0): ---")

with open("ukazkovy.txt", "r") as f:
    prvni = f.read()
    f.seek(0)           # ← vrátíme kurzor na začátek
    druhy = f.read()

print(f"1. čtení: {repr(prvni[:30])}...")
print(f"2. čtení: {repr(druhy[:30])}...")  # funguje!
````

## 🔧 Zpracování dat ze souboru

Data v textovém souboru jsou **vždy text** (řetězce). Pokud chceme počítat s čísly, musíme je **převést**:

```
Soubor:           Po načtení:         Po zpracování:
┌──────────┐      ┌──────────┐        ┌──────────┐
│ "Petr"   │  →   │ str      │   →    │ str      │
│ "25"     │  →   │ str      │   →    │ int: 25  │
│ "85.5"   │  →   │ str      │   →    │ float    │
└──────────┘      └──────────┘        └──────────┘
```

### Typický postup:

1. **Přečíst** řádek
2. **`strip()`** – odstranit bílé znaky
3. **`split()`** – rozdělit na části (podle mezery, čárky, ...)
4. **Převést** typy (`int()`, `float()`, ...)

````python
# === Zpracování strukturovaných dat ===
# Nejprve vytvoříme ukázkový soubor:
with open("znamky.txt", "w") as f:
    f.write("Adam 1 2 1 3\n")
    f.write("Bára 2 1 1 2\n")
    f.write("Cyril 3 2 4 2\n")
    f.write("Dana 1 1 1 1\n")

# Nyní ho přečteme a zpracujeme:
print(f"{'Jméno':<10} {'Známky':<15} {'Průměr':>6}")
print("-" * 35)

with open("znamky.txt", "r") as f:
    for radek in f:
        casti = radek.strip().split()     # ["Adam", "1", "2", "1", "3"]
        jmeno = casti[0]                   # "Adam"
        znamky = [int(z) for z in casti[1:]]  # [1, 2, 1, 3]
        prumer = sum(znamky) / len(znamky)
        print(f"{jmeno:<10} {str(znamky):<15} {prumer:>6.2f}")
````

### `split()` s vlastním oddělovačem

Ve výchozím nastavení `split()` dělí podle **mezer**. Můžeme ale zadat **jiný oddělovač**:

```python
"Adam;25;Praha".split(";")     # ["Adam", "25", "Praha"]
"2024-01-15".split("-")        # ["2024", "01", "15"]
"a, b, c".split(", ")         # ["a", "b", "c"]
```

> 💡 Soubory s daty oddělenými středníkem nebo čárkou se nazývají **CSV** (Comma Separated Values) – velmi běžný formát.

### Praktická ukázka – parsování CSV

CSV soubory jsou nejběžnější způsob výměny tabulkových dat. V Česku se často používá **středník** jako oddělovač (protože čárka je u nás desetinný oddělovač):

```
jméno;věk;město
Adam;17;Kladno
Bára;16;Praha
Cyril;18;Beroun
```

Ruční parsování CSV souboru je přímočaré – stačí `split()` s příslušným oddělovačem:

````python
# === Práce s CSV souborem ===

# 1. Vytvoříme ukázkový CSV
with open("studenti.csv", "w", encoding="utf-8") as f:
    f.write("jméno;věk;město\n")
    f.write("Adam;17;Kladno\n")
    f.write("Bára;16;Praha\n")
    f.write("Cyril;18;Beroun\n")
    f.write("Dana;17;Slaný\n")

# 2. Přečteme a zpracujeme
with open("studenti.csv", "r", encoding="utf-8") as f:
    hlavicka = f.readline().strip().split(";")  # přeskočíme hlavičku
    print(f"Sloupce: {hlavicka}\n")

    for radek in f:
        jmeno, vek, mesto = radek.strip().split(";")
        vek = int(vek)
        print(f"  {jmeno} ({vek} let) – bydliště: {mesto}")
        if vek >= 18:
            print(f"    → {jmeno} je plnoletý/á ✅")
````

## 🛡️ Ošetření chyb při práci se soubory

Při práci se soubory může nastat řada problémů:

| Chyba | Příčina |
|:---|:---|
| `FileNotFoundError` | Soubor neexistuje (při čtení) |
| `PermissionError` | Nemáme oprávnění číst/zapisovat |
| `IsADirectoryError` | Pokusili jsme se otevřít složku místo souboru |
| `UnicodeDecodeError` | Soubor má jiné kódování |

````python
# === Bezpečné čtení souboru ===
nazev = "neexistuje.txt"

try:
    with open(nazev, "r") as f:
        obsah = f.read()
        print(obsah)
except FileNotFoundError:
    print(f"❌ Soubor '{nazev}' nenalezen!")
except PermissionError:
    print(f"❌ Nemáte oprávnění číst '{nazev}'!")
````

### Kontrola existence souboru

Místo zachytávání chyb můžeme **předem zkontrolovat**, zda soubor existuje:

````python
# === Kontrola existence souboru ===
import os

nazev = "ukazkovy.txt"

if os.path.exists(nazev):
    print(f"✅ Soubor '{nazev}' existuje")
    print(f"   Velikost: {os.path.getsize(nazev)} bajtů")
else:
    print(f"❌ Soubor '{nazev}' neexistuje")
````

## 💡 Praktický příklad – narozeniny

Ukázka kompletního zpracování: čtení souboru, parsování dat, porovnání:

````python
import os
from datetime import date

# Zápis do souboru - na jendom řádku jsou dvoje data - jméno a datum - obojí je však uloženo jako text
with open('osoby.txt', 'w') as soubor:
  soubor.write('Petr 1990-08-28\n')
  soubor.write('Jana 1985-05-20\n')
  soubor.write('Pavel 2001-12-10\n')
  soubor.write('Lucie 1978-03-08\n')
  soubor.write('Martin 1995-07-25\n')
  soubor.write('Matěj 1995-11-06\n')

# Čtení ze souboru a kontrola narozenin
dnes = date.today() # dnešní datum získáme pomocí knihovny datetime
with open('osoby.txt', 'r') as s:
  for radek in s: # for cyklem čteme soubor řádek po řádku až dokonce souboru
    data = radek.strip().split() # strip() odstraní bílé znaky ze začátku a konce textu, split() rozdělí text v místech, kde jsou mezery
    jmeno = data[0] # první část textu je jméno
    datum_narozeni = data[1] # druhá část textu je datum narození
    rok, mesic, den = map(int, datum_narozeni.split('-')) # split() rozdělí datum v místech, kde jsou pomlčky, map() převede každou část na integer (celé číslo)
    if date(dnes.year, mesic, den) == dnes: # ze tří integerů (celých čísel) vytvoříme skutečné "datum" (nikoliv jen text), které se umí porovnat s jiným datumem
      print(f"Blahopřeji k narozeninám, {jmeno}!")
````

## 🌍 Kódování (encoding)

Textové soubory mohou být uloženy v **různých kódováních**. Pokud čeština vypadá „rozsypaně" (znaky jako `Ã¡`, `Å¾`), je to problém s kódováním.

| Kódování | Popis |
|:---|:---|
| `utf-8` | Moderní standard – podporuje **všechny** znaky (doporučeno) |
| `cp1250` | Starší Windows kódování pro češtinu |
| `latin-1` | Západoevropské jazyky |

```python
# Explicitní nastavení kódování:
with open("soubor.txt", "r", encoding="utf-8") as f:
    obsah = f.read()
```

> 💡 **Tip:** Pokud máte problémy s diakritikou, zkuste přidat `encoding="utf-8"` do `open()`.

## 📋 Shrnutí

### Otevírání souborů

```python
with open("soubor.txt", "režim", encoding="utf-8") as f:
    # práce se souborem
# soubor se automaticky zavře
```

### Režimy

| Režim | Akce | Poznámka |
|:---|:---|:---|
| `"r"` | Čtení | Soubor musí existovat |
| `"w"` | Zápis | **Smaže** původní obsah! |
| `"a"` | Připojení | Přidá na konec |

### Metody pro čtení

| Metoda | Vrátí | Příklad |
|:---|:---|:---|
| `read()` | Celý obsah (str) | `obsah = f.read()` |
| `readline()` | Jeden řádek (str) | `radek = f.readline()` |
| `readlines()` | Seznam řádků (list) | `radky = f.readlines()` |
| `for radek in f` | Řádek po řádku | Nejčastější způsob |

### Metody pro zápis

| Metoda | Co dělá | Přidá `\n`? |
|:---|:---|:---|
| `write(text)` | Zapíše text | ❌ Ne |
| `writelines(list)` | Zapíše seznam | ❌ Ne |

### Zpracování řádku

```python
radek = "Adam 1 2 3\n"
radek = radek.strip()           # "Adam 1 2 3"
casti = radek.split()           # ["Adam", "1", "2", "3"]
jmeno = casti[0]                # "Adam"
cisla = [int(x) for x in casti[1:]]  # [1, 2, 3]
```

### Pravidla

| Pravidlo | |
|:---|:---|
| Vždy používejte `with` | Automaticky zavře soubor |
| `strip()` po čtení řádku | Odstraní `\n` na konci |
| Pozor na `"w"` režim | Smaže vše! Pro přidání použijte `"a"` |
| `encoding="utf-8"` | Řeší problémy s diakritikou |
| `try-except` u čtení | Soubor nemusí existovat |

> 🎯 **Zapamatuj si:** Soubory = trvalé úložiště dat. Otevři (`with open`), zpracuj (`read/write`), zavři (automaticky). Vždy `strip()` při čtení!
