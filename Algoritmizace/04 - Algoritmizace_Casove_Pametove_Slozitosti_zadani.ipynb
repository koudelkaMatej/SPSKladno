# 📖 Teorie – Co je složitost algoritmu?

Když porovnáváme dva algoritmy řešící stejný problém, ptáme se:
- Jak rychle algoritmus poroste s velikostí vstupu?
- Kolik paměti spotřebuje?

**Složitost algoritmu** popisuje, jak se mění čas nebo paměť v závislosti na velikosti vstupu `n`.

---

## Notace Big O – O(…)

**Big O** vyjadřuje **horní odhad** složitosti – popisuje nejhorší možný případ.

Zjednodušující pravidla:
- Konstanty ignorujeme: `O(2n)` → `O(n)`
- Menší členy ignorujeme: `O(n² + n)` → `O(n²)`
- Zajímá nás pouze dominantní člen

---

## Přehled nejčastějších složitostí (od nejlepší k nejhorší)

| Notace | Název | Příklad |
|--------|-------|--------|
| **O(1)** | Konstantní | Přístup k prvku pole `pole[i]` |
| **O(log n)** | Logaritmická | Binární vyhledávání |
| **O(n)** | Lineární | Lineární vyhledávání, jeden for cyklus |
| **O(n log n)** | Lineárně-logaritmická | Merge Sort, Quick Sort (průměr) |
| **O(n²)** | Kvadratická | Bubble Sort, dva vnořené for cykly |
| **O(2ⁿ)** | Exponenciální | Rekurzivní Fibonacci bez memoizace |
| **O(n!)** | Faktoriální | Brute-force permutace |

---

## Vizuální porovnání

```
Počet operací
▲
│                                          O(n!)
│                                    O(2ⁿ)
│                              O(n²)
│                        O(n log n)
│                  O(n)
│          O(log n)
│  O(1)
└─────────────────────────────────────────────► n (velikost vstupu)
```

# 🔬 Části 1–7: Ukázky jednotlivých složitostí

## 1. O(1) – Konstantní složitost

Čas výpočtu **nezávisí** na velikosti vstupu. Bez ohledu na velikost pole trvá operace stejně dlouho.

**Příklady:**
- Čtení prvku pole na indexu `i`
- Přidání prvku na konec seznamu (amortizovaně)
- Aritmetická operace

```python
def prvni_prvek(pole):
    return pole[0]   # vždy 1 operace, bez ohledu na délku pole
```

> ✅ Ideální složitost – algoritmus je stejně rychlý pro n=10 i n=1 000 000.

````python
import time

def prvni_prvek(pole):
    return pole[0]

# Testování pro různé velikosti vstupu
for velikost in [100, 10_000, 1_000_000]:
    pole = list(range(velikost))
    zacatek = time.perf_counter()
    prvni_prvek(pole)
    konec = time.perf_counter()
    print(f"n = {velikost:>10} | čas: {(konec - zacatek) * 1e6:.4f} µs")
````

---

## 2. O(log n) – Logaritmická složitost

Počet operací roste **logaritmicky** – s každým krokem se problém zhruba **halvuje**.

**Příklady:**
- Binární vyhledávání
- Vyhledávání v binárním vyhledávacím stromě

```
n = 1 000 000 → log₂(1 000 000) ≈ 20 kroků
n = 1 000 000 000 → log₂(1 000 000 000) ≈ 30 kroků
```

````python
import time
import math

def binarni_vyhledavani(pole, cil):
    leva, prava = 0, len(pole) - 1
    kroky = 0
    while leva <= prava:
        kroky += 1
        stred = (leva + prava) // 2
        if pole[stred] == cil:
            return stred, kroky
        elif pole[stred] < cil:
            leva = stred + 1
        else:
            prava = stred - 1
    return -1, kroky

for velikost in [100, 10_000, 1_000_000]:
    pole = list(range(velikost))
    _, kroky = binarni_vyhledavani(pole, velikost - 1)  # hledáme poslední prvek
    print(f"n = {velikost:>10} | kroky: {kroky:>3} | log₂(n) ≈ {math.log2(velikost):.1f}")
````

---

## 3. O(n) – Lineární složitost

Počet operací roste **přímo úměrně** s velikostí vstupu.

**Příklady:**
- Lineární vyhledávání
- Průchod celým polem (jeden for cyklus)
- Výpočet součtu všech prvků

````python
import time

def soucet_pole(pole):
    soucet = 0
    for prvek in pole:   # n operací
        soucet += prvek
    return soucet

for velikost in [1_000, 100_000, 10_000_000]:
    pole = list(range(velikost))
    zacatek = time.perf_counter()
    soucet_pole(pole)
    konec = time.perf_counter()
    print(f"n = {velikost:>12} | čas: {(konec - zacatek) * 1000:.2f} ms")
````

---

## 4. O(n log n) – Lineárně-logaritmická složitost

Typická pro **efektivní třídicí algoritmy** – kombinace lineárního a logaritmického přístupu.

**Příklady:**
- Merge Sort
- Quick Sort (průměrný případ)
- Heap Sort

````python
import time
import random

def merge_sort(pole):
    if len(pole) <= 1:
        return pole
    stred = len(pole) // 2
    leva = merge_sort(pole[:stred])
    prava = merge_sort(pole[stred:])
    return merge(leva, prava)

def merge(leva, prava):
    vysledek = []
    i = j = 0
    while i < len(leva) and j < len(prava):
        if leva[i] <= prava[j]:
            vysledek.append(leva[i]); i += 1
        else:
            vysledek.append(prava[j]); j += 1
    vysledek.extend(leva[i:])
    vysledek.extend(prava[j:])
    return vysledek

for velikost in [1_000, 10_000, 100_000]:
    pole = [random.randint(0, 10000) for _ in range(velikost)]
    zacatek = time.perf_counter()
    merge_sort(pole)
    konec = time.perf_counter()
    print(f"n = {velikost:>8} | čas: {(konec - zacatek) * 1000:.2f} ms")
````

---

## 5. O(n²) – Kvadratická složitost

Typická pro **dva vnořené cykly** – pro každý prvek projdeme celé pole znovu.

**Příklady:**
- Bubble Sort
- Selection Sort
- Insertion Sort
- Porovnání každé dvojice prvků

> ⚠️ Pro n = 10 000 je to 100 000 000 operací – výrazně pomalejší než O(n log n)!

````python
import time
import random

def bubble_sort(pole):
    pole = pole.copy()
    n = len(pole)
    for i in range(n):          # vnější cyklus: n průchodů
        for j in range(n - i - 1):  # vnitřní cyklus: n-i-1 porovnání
            if pole[j] > pole[j + 1]:
                pole[j], pole[j + 1] = pole[j + 1], pole[j]
    return pole

for velikost in [500, 2_000, 5_000]:
    pole = [random.randint(0, 10000) for _ in range(velikost)]
    zacatek = time.perf_counter()
    bubble_sort(pole)
    konec = time.perf_counter()
    print(f"n = {velikost:>6} | čas: {(konec - zacatek) * 1000:.2f} ms")
````

---

## 6. O(2ⁿ) – Exponenciální složitost

Počet operací se s každým přidaným prvkem **zdvojnásobí**. Nepoužitelné pro velké vstupy.

**Příklady:**
- Rekurzivní Fibonacci bez memoizace
- Prohledávání všech podmnožin

```
n = 10  → ~1 000 operací
n = 30  → ~1 000 000 000 operací
n = 50  → ~1 000 000 000 000 000 operací
```

````python
import time

pocitadlo = 0

def fib_rekurze(n):
    global pocitadlo
    pocitadlo += 1
    if n <= 1:
        return n
    return fib_rekurze(n - 1) + fib_rekurze(n - 2)

for n in [10, 20, 30]:
    pocitadlo = 0
    zacatek = time.perf_counter()
    vysledek = fib_rekurze(n)
    konec = time.perf_counter()
    print(f"fib({n:>2}) = {vysledek:>8} | volání funkce: {pocitadlo:>10} | čas: {(konec - zacatek) * 1000:.2f} ms")
````

## 7. Porovnání složitostí – graf

Vizualizujeme růst jednotlivých složitostí pomocí matplotlib.

# 💾 Paměťová složitost

**Paměťová složitost** (prostorová složitost) popisuje, kolik **paměti navíc** algoritmus potřebuje v závislosti na velikosti vstupu.

| Notace | Popis | Příklad |
|--------|-------|--------|
| **O(1)** | Konstantní paměť | Bubble Sort – třídíme na místě, jen pár proměnných |
| **O(log n)** | Logaritmická | Quick Sort – hloubka rekurzivního zásobníku |
| **O(n)** | Lineární | Merge Sort – pomocné pole stejné velikosti |
| **O(n²)** | Kvadratická | Matice sousednosti grafu |

> 💡 Časová a paměťová složitost spolu **nesouvisí přímočaře** – algoritmus může být rychlý, ale spotřebovávat hodně paměti (a naopak).

### Příklad: Merge Sort vs Bubble Sort

| Algoritmus | Časová složitost | Paměťová složitost |
|------------|-----------------|-------------------|
| Bubble Sort | O(n²) | **O(1)** – třídí na místě |
| Merge Sort | **O(n log n)** | O(n) – potřebuje pomocné pole |

# 🚀 Optimalizace – Memoizace

**Memoizace** ukládá výsledky dříve vypočítaných volání funkce, aby se nemusely počítat znovu.

Rekurzivní Fibonacci má složitost **O(2ⁿ)**, protože stejné podproblémy počítá mnohokrát.
S memoizací klesne na **O(n)**.

````python
import time
from functools import lru_cache

# Bez memoizace – O(2ⁿ)
def fib_pomaly(n):
    if n <= 1:
        return n
    return fib_pomaly(n - 1) + fib_pomaly(n - 2)

# S memoizací – O(n)
@lru_cache(maxsize=None)
def fib_rychly(n):
    if n <= 1:
        return n
    return fib_rychly(n - 1) + fib_rychly(n - 2)

n = 35

zacatek = time.perf_counter()
vysledek1 = fib_pomaly(n)
konec = time.perf_counter()
print(f"fib_pomaly({n}) = {vysledek1} | čas: {(konec - zacatek) * 1000:.2f} ms  [O(2ⁿ)]")

zacatek = time.perf_counter()
vysledek2 = fib_rychly(n)
konec = time.perf_counter()
print(f"fib_rychly({n}) = {vysledek2} | čas: {(konec - zacatek) * 1000:.4f} ms  [O(n)]")
````
